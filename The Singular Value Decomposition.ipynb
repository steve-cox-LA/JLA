# The Singular Value Decomposition


The singular value decomposition is, in a sense, the spectral
representation of a rectangular matrix.
Of course if $A$ is $m$-by-$n$ and $m\ne n$ then it does not make
sense to speak of the eigenvalues of $A$. We may, however, rely on
the previous chapter to give us relevant spectral representations of
the two symmetric matrices 

\begin{equation*}
  A^TA \and AA^T.
\end{equation*}

That these two matrices together indeed tell us 'everything' about 
$A$ can be gleaned from

\begin{equation}
  \eqalign{  \N(A^TA)&=\N(A), \hskip 0.25in \N(AA^T)=\N(A^T), \cr
  \Ra(A^TA) &= \Ra(A^T), \and \Ra(AA^T)=\Ra(A). \cr} \label{eq:spacM}
\end{equation}

You have proven the first of these in Exer.~\secref{null}.\ref{nc3}.
The proof of the second is identical. The row and column space results
follow from the first two via orthogonality. In light of (\ref{eq:spacM})
we will see that the singular value decomposition of $A$ delivers orthonormal
bases for the four fundamental subspaces of $A$.

Beyond the satisfaction of reinforcing the Fundamental Theorem of Linear
Algebra the singular value decomposition will also shed new light on both 
least squares and the pseudoinverse. These results together indicate that
the SVD may be a useful means for constructing accurate low rank approximations
of large matrices. Our experimental investigation of this hunch
in \S\subsecref{svdic} will be followed by theoretical confirmation in 
\S\subsecref{lra}.
In addition, it is a fundamental
tool in information and data sciences. 

## The Decomposition  

On the spectral side, we shall now see that the eigenvalues of $AA^T$ and 
$A^TA$ are nonnegative and that their nonzero eigenvalues coincide.
Let us first confirm this on the $A$ matrix associated with
the unstable swing (see Figure \ref{fig:frame})

\begin{equation}
  A = \begin{bmatrix}0 & 1 & 0 & 0 \cr -1 & 0 & 1 & 0 \cr 0 & 0 & 0 & 1 \end{bmatrix}.
  \label{eq:swa}
\end{equation}

The respective products are 

\begin{equation*}
  AA^T = \begin{bmatrix}1 & 0 & 0 \cr 0 & 2 & 0 \cr 0 & 0 & 1\end{bmatrix}
  \and
  A^TA = \begin{bmatrix}1 & 0 & -1 & 0 \cr
                0 & 1 & 0 & 0 \cr
                -1 & 0 & 1 & 0 \cr
                0 & 0 & 0 & 1 \end{bmatrix}.
\end{equation*}

Analysis of the first is particularly simple. Its null space is
clearly just the zero vector while $\la_1=2$ and $\la_2=1$ are its
eigenvalues. Their geometric multiplicities are $n_1=1$ and $n_2=2$.
In $A^TA$ we recognize the $S$ matrix from exercise 
\secref{sym}.\ref{symexe1} and recall
that its eigenvalues are $\la_1=2,\ \la_2=1$, and $\la_3=0$ with
multiplicities $n_1=1,\ n_2=2$, and $n_3=1$.
Hence, at least for this $A$, the eigenvalues of $AA^T$ and $A^TA$ are 
nonnegative and their nonzero eigenvalues coincide. In addition, the
geometric multiplicities of the nonzero eigenvalues sum to 3, the rank
of $A$.
 
\begin{proposition}\label{prop:svd1}. The eigenvalues of $AA^T$ and $A^TA$ are 
nonnegative. Their nonzero eigenvalues, including geometric multiplicities,
coincide. The geometric multiplicities of the nonzero eigenvalues sum to 
the rank of $A$.

\end{proposition}

**Proof:** If $A^TAx=\la x$ then $x^TA^TAx=\la x^Tx$, i.e., $\|Ax\|^2=\la\|x\|^2$ and
so $\la\ge 0$. A similar argument works for $AA^T$. 

Now suppose that $\la_j>0$ and that $\{x_{j,k}\}_{k=1}^{n_j}$ constitutes an
orthogonal basis for the eigenspace $\Ra(P_j)$.
Starting from 

\begin{equation}
  A^TAx_{j,k}=\la_j x_{j,k} \label{eq:eigjk}
\end{equation} 
 
we find, on multiplying through (from the left) by $A$ that 

\begin{equation*}
  AA^TAx_{j,k}=\la_j Ax_{j,k},
\end{equation*} 
 
i.e., $\la_j$ is an eigenvalue of $AA^T$ with eigenvector $Ax_{j,k}$, 
so long as $Ax_{j,k}\ne 0$. It follows from the first paragraph of this proof
that $\|Ax_{j,k}\|=\sqrt\la_j$, which, by hypothesis, is nonzero. Hence,

\begin{equation}
  y_{j,k} \equiv \frac{Ax_{j,k}}{\sqrt\la_j}, \hskip 0.25in 1\le k\le n_j
  \label{eq:yjk}
\end{equation}

is a collection of unit eigenvectors of $AA^T$ associated with $\la_j$.
Let us now show that these vectors are orthonormal for fixed $j$.

\begin{equation*}
  y_{j,i}^Ty_{j,k} = \frac{1}{\la_j}x_{j,i}^TA^TAx_{j,k}
                   = x_{j,i}^Tx_{j,k} = 0.
\end{equation*}

We have now demonstrated that if $\la_j>0$ is an eigenvalue of $A^TA$ of
geometric multiplicity $n_j$ then it is an eigenvalue of $AA^T$ of
geometric multiplicity at least $n_j$. Reversing the argument, i.e.,
generating eigenvectors of $A^TA$ from those of $AA^T$ we find that the
geometric multiplicities must indeed coincide.

Regarding the rank statement, we discern from (\ref{eq:eigjk}) that if
$\la_j>0$ then $x_{j,k}\in\Ra(A^TA)$. The union of these vectors indeed
constitutes a basis for $\Ra(A^TA)$, for anything orthogonal to each of
these $x_{j,k}$ necessarily lies in the eigenspace corresponding to a
zero eigenvalue, i.e., in $\N(A^TA)$. As $\Ra(A^TA)=\Ra(A^T)$ it follows
that ${\rm dim}\,\Ra(A^TA)=r$ and hence the $n_j$, for $\la_j>0$, sum to
$r$. **End of Proof.**

\bigskip

Let us now gather together some of the separate pieces of the proof.
For starters, we order the eigenvalues of $A^TA$ from high to low,

\begin{equation*}
  \la_1>\la_2>\cdots>\la_h
\end{equation*}

and write

\begin{equation}
  A^TA = X\Lambda_n X^T \label{eq:specX}
\end{equation}

where

\begin{equation*}
  X = [X_1 \cdots X_h], \where X_j = [x_{j,1}\cdots x_{j,n_j}]
\end{equation*}

and $\Lambda_n$ is the $n$-by-$n$ diagonal matrix with $\la_1$ in the
first $n_1$ slots, $\la_2$ in the next $n_2$ slots,  *etc* .
Similarly

\begin{equation}
  AA^T = Y\Lambda_m Y^T \label{eq:specY}
\end{equation}

where

\begin{equation*}
  Y = [Y_1 \cdots Y_h], \where Y_j = [y_{j,1}\cdots y_{j,n_j}].
\end{equation*}

and $\Lambda_m$ is the $m$-by-$m$ diagonal matrix with $\la_1$ in the
first $n_1$ slots, $\la_2$ in the next $n_2$ slots,  *etc* .
The $y_{j,k}$ were defined in (\ref{eq:yjk}) under the assumption that
$\la_j>0$. If $\la_j=0$ let $Y_j$ denote an orthonormal basis for
$\N(AA^T)$. Finally, call

\begin{equation*}
  \sigma_j = \sqrt\la_j 
\end{equation*}

and let $\Sigma$ denote the $m$-by-$n$ matrix diagonal
matrix with $\s_1$ in the first $n_1$ slots and $\s_2$ in the next $n_2$ 
slots,  *etc* . Notice that

\begin{equation}
  \Sigma^T\Sigma = \Lambda_n \and \Sigma\Sigma^T = \Lambda_m. \label{eq:sig2}
\end{equation}

Now recognize that (\ref{eq:yjk}) may be written

\begin{equation*}
  Ax_{j,k} = \s_j y_{j,k}
\end{equation*}

and that this is simply the column by column rendition of

\begin{equation*}
  AX = Y\Sigma.
\end{equation*}

As $XX^T=I$ we may multiply through (from the right) by $X^T$ and arrive at
the  **singular value decomposition**  of $A$,

\begin{equation}
  \boxed{A = Y\Sigma X^T.} \label{eq:svd}
\end{equation}

Let us confirm this on the $A$ matrix in (\ref{eq:swa}).
We have

\begin{equation*}
  \Lambda_4 = \begin{bmatrix}2 & 0 & 0 & 0 \cr
                       0 & 1 & 0 & 0 \cr
		       0 & 0 & 1 & 0 \cr
		       0 & 0 & 0 & 0 \end{bmatrix}  \and
  X = \frac{1}{\sqrt 2}\begin{bmatrix}-1 & 0		&  0 	& 1 \cr
				0  & \sqrt 2	& 0	& 0 \cr
                                1  &  0		& 0	& 1 \cr
                                0  &  0         & \sqrt 2 & 0\end{bmatrix}
\end{equation*}

and

\begin{equation*}
  \Lambda_3 = \begin{bmatrix}2 & 0 & 0 \cr
                       0 & 1 & 0 \cr
                       0 & 0 & 1 \end{bmatrix} \and
  Y = \begin{bmatrix} 0 & 1 & 0 \cr
                1 & 0 & 0 \cr
                0 & 0 & 1 \end{bmatrix}.
\end{equation*}

Hence,

\begin{equation}
  \Sigma = \begin{bmatrix}\sqrt 2 & 0 & 0 & 0\cr
                    0 & 1 & 0 & 0 \cr
                    0 & 0 & 1 & 0 \end{bmatrix} \label{eq:swsig}
\end{equation}

and so $A=Y\Sigma X^T$ says that $A$ should coincide with

\begin{equation*}
\begin{bmatrix} 0 & 1 & 0 \cr
                1 & 0 & 0 \cr
                0 & 0 & 1 \end{bmatrix}
\begin{bmatrix}\sqrt 2 & 0 & 0 & 0\cr
                    0 & 1 & 0 & 0 \cr
                    0 & 0 & 1 & 0 \end{bmatrix}
\begin{bmatrix}-1/\sqrt 2 & 0 & 1/\sqrt 2 & 0 \cr
         0 & 1 & 0 & 0 \cr
         0 & 0 & 0 & 1 \cr
         1/\sqrt 2 & 0 & 1/\sqrt 2 & 0 \end{bmatrix}.
\end{equation*}

This indeed agrees with $A$. It also agrees (up to sign changes in the
columns of $X$) with what one receives upon typing  *[Y,SIG,X]=svd(A)*  
in \Matlab.

You now ask what we get for our troubles. I express the first dividend
as a proposition that looks to me like a quantitative version of the
fundamental theorem of linear algebra.

\begin{proposition}\label{prop:svd2}. If $Y\Sigma X^T$ is the singular value
decomposition of $A$ then

 (i) The rank of $A$, call it $r$, is the number of nonzero elements in
$\Sigma$.

 (ii) The first $r$ columns of $X$ constitute an orthonormal basis for
$\Ra(A^T)$. The $n-r$ last columns of $X$  constitute an orthonormal basis for
$\N(A)$.

 (iii) The first $r$ columns of $Y$ constitute an orthonormal basis for
$\Ra(A)$. The $m-r$ last columns of $Y$  constitute an orthonormal basis for
$\N(A^T)$.

\end{proposition}

Let us now 'solve' $Ax=b$ with the help of the pseudo-inverse of $A$.
You know the 'right' thing to do, namely reciprocate all of the nonzero
singular values. Because $m$ is not necessarily $n$ we must also be
careful with dimensions. To be precise, let $\Sigma^+$ denote the
$n$-by-$m$ matrix  whose first $n_1$
diagonal elements are $1/\s_1$, whose next $n_2$ diagonal elements are $1/\s_2$
and so on. In the case that $\s_h=0$, set the final $n_h$ diagonal elements of
$\Sigma^+$ to zero. Now, one defines the  **pseudo-inverse**  of $A$ to be

\begin{equation*}
  A^+ \equiv X\Sigma^+Y^T.
\end{equation*}

Taking the $A$ of (\ref{eq:swa}) we find

\begin{equation*}
  \Sigma^+ = \begin{bmatrix}1/\sqrt 2 & 0 & 0 \cr 0 & 1 & 0 \cr 0 & 0 & 1 \cr
     0 & 0 & 0 \end{bmatrix}
\end{equation*}

and so

\begin{equation*}
  A^+ =  \begin{bmatrix}-1/\sq2 & 0 & 0 & 1/\sq2 \cr
                  0 & 1 & 0 & 0 \cr
                   1/\sq2 & 0 & 0 & 1/\sq2 \cr
                  0 & 0 & 1 & 0 \end{bmatrix}
  \begin{bmatrix}1/\sqrt 2 & 0 & 0 \cr 0 & 1 & 0 \cr 0 & 0 & 1 \cr
     0 & 0 & 0 \end{bmatrix}
\begin{bmatrix} 0 & 1 & 0 \cr
                1 & 0 & 0 \cr
                0 & 0 & 1 \end{bmatrix} 
   = \begin{bmatrix}0 & -1/2 & 0 \cr 1 & 0 & 0 \cr 0 & 1/2 & 0 \cr
               0 & 0 & 1 \end{bmatrix}, 
\end{equation*}

in agreement with what appears from  *pinv(A)*  . Let us now investigate
the sense in which $A^+$ is the inverse of $A$. Suppose that $b\in\bR^m$
and that we wish to solve $Ax=b$. We suspect that $A^+b$ should be a good
candidate. Observe now that

\begin{equation*}
  \eqalign{(A^TA)A^+b 
  &= X\Lambda_n X^TX\Sigma^+Y^Tb \hskip 0.25in\hbox{by (\ref{eq:specX}})\cr
  & = X\Lambda_n\Sigma^+Y^Tb \hskip 0.25in\hbox{because}\ X^TX=I \cr
  &= X\Sigma^T\Sigma\Sigma^+Y^Tb \hskip 0.25in\hbox{by (\ref{eq:sig2}}) \cr
  &= X\Sigma^T Y^Tb \hskip 0.25in\hbox{because}\ \Sigma^T\Sigma\Sigma^+=\Sigma^T\cr
  &= A^Tb \hskip 0.25in\hbox{by (\ref{eq:svd}}, \cr})
\end{equation*}

that is, $A^+b$ satisfies the  **least-squares problem**  $A^TAx=A^T b$.

$x^+=A^+b$ and suppose $x$ satisfies $A^TAx=A^Tb$. Prove that

## The SVD in Image Compression  

Most applications of the SVD are manifestations of the folk theorem, 

*The singular vectors associated with the largest singular values capture
the essence of the matrix.*

This is most easily seen when applied to
gray scale images. For example, the  *jpeg*   associated with the image
in the top left of Figure \ref{fig:jbred} is a 262-by-165 matrix. 

```python
# John Brown Image and its Singular Values

import numpy as np
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2)    # create a figure with 2 axes (plot windows)
plt.subplots_adjust(right=2, top=1, wspace=0.25)  # adjust them for the right look

A = plt.imread('Images/JohnBrown.jpg')
ax1.imshow(A, cmap='gray', vmin=0, vmax=255)
ax1.axis('off')

Y, sigma, X = np.linalg.svd(A, full_matrices=True)
ax2.semilogy(sigma,'ko', markersize=3)
ax2.set_ylim([1,1e5])
ax2.grid('on')
ax2.set_xlabel('index', fontsize=12)
ax2.set_ylabel('singular value', fontsize=12);
```

## Low Rank Approximation  

In this section we establish a precise version of last section's
folk theorem. To prepare the way we note that 
if $B=B^T\in\maR^{m\times m}$ and we list its eigenvalues, including multiplicities, 
$\{\la_1,\la_2,\ldots,\la_m\}$ then, from \cor\procref{eig7c} it follows that

\begin{equation}
  \tr(B) = \sum_{i=1}^m \la_j(B).  \label{eq:trspec}
\end{equation}

This together with our trace formulation of the Frobenius norm of
$A\in\maR^{m\times n}$, recall (\ref{eq:frobdef2}), yields

\begin{equation}
  \|A\|_F^2 = \tr(AA^T) = \sum_{i=1}^m \la_i(AA^T) = \sum_{i=1}^m \sigma_i^2,
    \label{eq:Fnormspec}
\end{equation}

i.e., the Frobenius norm of a matrix is the square root of the sum of the
squares of its singular values. We will also need the fact that if
$Q$ is square and $Q^TQ=I$ then

\begin{equation}
  \|QA\|_F^2 = \tr(QAA^TQ^T) = \tr(AA^TQ^TQ) = \tr(AA^T)
   =\|A\|_F^2. \label{eq:frobinv}
\end{equation}

The same argument reveals that $\|AQ\|_F = \|A\|_F$.
We may now establish

\begin{proposition}\label{prop:svd3} Given an $m$-by-$n$ matrix $A$ (with SVD
$A=Y\Sigma X^T$) and a whole number $k\le {\rm min} (m,n)$ then the
best (in terms of Frobenius distance) rank $k$ approximation of $A$ is

\begin{equation*}
  A_k = Y(:,1:k)\Sigma(1:k,1:k)X(:,1:k)^T.
\end{equation*}

The square of the associated approximation error is

\begin{equation*}
  \|A-A_k\|_F^2 = \min_{{\rm rank}(B)=k}\|A-B\|_F^2 = \sum_{j>k}\sigma_j^2.
\end{equation*}

\end{proposition}

**Proof:** If $B$ is $m$-by-$n$ then

\begin{equation*}
  \|A-B\|_F^2 = \|Y\Sigma X^T-B\|_F^2 = \|Y(\Sigma - Y^TBX)X^T\|_F^2
   = \|\Sigma - Y^TBX\|_F^2.
\end{equation*}

If $B$ is to be chosen, among matrices of rank $k$, to minimize this
distance then, as $\Sigma$ is diagonal, so too must $Y^TBX$. If we
denote this diagonal matrix by $S$ then

\begin{equation*}
  Y^TBX = S \quad\hbox{implies}\hskip 0.25in B = YSX^T,
\end{equation*}

and so

\begin{equation*}
  \|A-B\|_F^2 = \sum_j(\sigma_j-s_j)^2.
\end{equation*}

As the rank of $B$ is $k$ it follows that the best choice of the
$s_j$ is $s_j=\sigma_j$ for $j=1:k$ and $s_j=0$ there after. **End of Proof.**

The expression, (\ref{eq:Fnormspec}), of the Frobenius norm in terms of
singular values leads naturally to the question of like bounds on the
associated inner product. 
Upon associating to each unitary matrix, $U$, a doubly stochastic
matrix $D$, via $D_{i,j}=|U_{i,j}|^2$, 
we note that our Mixing Theorem, 
\prop\procref{fund5}, is well suited to this task. In particular,


\begin{proposition}\label{prop:svd4}. If $A$ and $B$ lie in $\maC^{m\times n}$ then

\begin{equation}
  |\tr(A^*B)| \le \s(A)^T\s(B). \label{eq:vNti}
\end{equation}

\end{proposition}


**Proof:** Given the two singular value decompositions
$A = USV^*$ and $B=WTX^*$ we express

\begin{equation*}
  \tr (A^*B) = \tr (VSU^*WTX^*) = \tr (X^*VSU^*WT) = \tr (Q^*SPT)
\end{equation*}

where $P\equiv U^*W$ and $Q=V^*X$ are unitary and so the matrices with
elements $|p_{ij}|^2$ and $|q_{ij}|^2$ are doubly stochastic and so

\begin{equation*}
  \eqalign{ |\tr(A^*B)|
  & = |\tr((SQ)^*PT)| \cr
  &\le \sum_{i=1}^n\sum_{j=1}^n s_it_j|\ov q_{ij}p_{ij}| \cr
  &\le \frac{1}{2}\sum_{i=1}^n\sum_{j=1}^n s_it_j|q_{ij}|^2 
    + \frac{1}{2}\sum_{i=1}^n\sum_{j=1}^n s_it_j|p_{ij}|^2 \cr
  &\le s^Tt. \cr}
\end{equation*}

The first inequality is the triangle inequality.
The second is $2|\ov z_1 z_2|\le(|z_1|^2+|z_2|^2)$ and the third
is our Mixture conclusion, (\ref{eq:bestmix}), on realizing that
the matrix with elements $(|p_{ij}|^2+|q_{ij}|^2)/2$ is also
doubly stochastic. **End of Proof.**

## Principal Component Analysis

Our modeling strategy works where it works but there are many fields where
data has outstripped first principals and we seek a way to patterns in data as
a means to finding order in the data.

PCA is a tool for multi-variate analysis. Given a data set set
$\{x_1,x_2,\ldots,x_m\}$ where each $x_j\maR^n$. The simplest place to
start is the mean. The next simplest is to find the directions in $\maR^n$ in
which the data varies most.


The statistical interpretation of least squares can be extended to a similar framework for PCA. Recall that we first considered a statistical interpretation for least squares where an unknown line $y=mx+b$ generated the observed points, 
in the sense that each observed point $i$ was obtained by perturbing the point 
$mx_i +b$ vertically by a random amount from a single Gaussian distribution with mean 0 and variance $\sigma^2$. 
PCA can be understood analogously by replacing  *vertically} by {\it orthogonally* . 
However this approach is not completely satisfactory as the orthogonality of the perturbation is is not readily interpretable. Stated differently, it is not obvious what physical processes would generate points orthogonal to an affine subspace by perturbations that are always orthogonal to the subspace. In the case of least squares, the vertical perturbation corresponds to noise in one measurement (represented by one coordinate). The problem is in naturally interpreting orthogonal perturbations in terms of a noise model for measurements. This difficulty is resolved by a model called probabilistic PCA (pPCA) that is illustrated in the figure below, and that we now explain:

In the pPCA model there is an (unknown) line (affine space in higher dimension) on which (hidden) points (blue) are generated at random according to a Gaussian distribution (represented by gray streak in the figure above, where the mean of the Gaussian is the green point). Observed points (red) are then generated from the hidden points by addition of isotropic Gaussian noise (blue smear), meaning that the Gaussian has a diagonal covariance matrix with equal entries. 
Formally, this is a linear Gaussian model described as follows:

Observed random variables $t$ are given by $t = Wx + \mu + \epsilon$ where 
$x$ are latent (hidden) random variables, $W$ is a matrix describing a subspace 
and $Wx+\mu$ are the latent points on an affine subspace ($\mu$ corresponds to a translation). Finally, $\epsilon$ is an error term, given by a Gaussian random 
variable with mean 0 and covariance matrix $\sigma^2 I$. The parameters of the model are $W$, $\mu$ and $\sigma^2$. Equivalently, the observed random 
variables are themselves Gaussian, described by the distribution 
$t \sim \mathcal{N}(\mu,WW^T + \psi)$ where 
$\psi \sim \mathcal{N}(0,\sigma^2I)$. Tipping and Bishop prove an analogy of the Gauss-Markov theorem, namely that the affine subspace given by the maximum likelihood estimates of $W$ and $\mu$ is the PCA subspace (the proof is not difficult but I omit it and refer interested readers to their paper, or Bishops Pattern Recognition and Machine Learning book).

It is important to note that although the maximum likelihood estimates of $W,\mu$ 
in the pPCA model correspond to the PCA subspace, only posterior distributions can be obtained for the latent data (points on the subspace). Neither the mode nor the mean of those distributions corresponds to the PCA points (orthogonal projections of the observations onto the subspace). However what is true, is that the posterior distributions converge to the PCA points as $\sigma^2 \rightarrow 0$. In other words, the relationship between pPCA and PCA is a bit more subtle than that between least squares and regression.


## Independent Component Analysis


The data is $n$ samples of $m$ signals, $X\in\maR^{m\times n}$.
To compute the first component we maximize the normalized kurtosis of
$w^TX$.

\begin{equation*}
  \cK(w) = \frac{E[(w^TX)^4]}{E^2[(w^TX)^2]} = \frac{n\sum_{j=1}^n\left(\sum_{i=1}^m w_iX_{i,j}\right)^4}
   {\left(\sum_{j=1}^n\left(\sum_{i=1}^m w_iX_{i,j}\right)^2\right)^2}
\end{equation*}

We first record its gradient   

\begin{equation*}
  \frac{\p\cK(w)}{\p w_k} 
   = 4n\frac{\sum_{j=1}^n\left(\sum_{i=1}^m w_iX_{i,j}\right)^2
    \sum_{j=1}^n\left(\sum_{i=1}^m w_iX_{i,j}\right)^3 X_{k,j}-
   \sum_{j=1}^n\left(\sum_{i=1}^m w_iX_{i,j}\right)^4\sum_{j=1}^n\sum_{i=1}w_iX_{i,j}X_{k,j}}
{\left(\sum_{j=1}^n\left(\sum_{i=1}^m w_iX_{i,j}\right)^2\right)^3}
\end{equation*}

Hence,

\begin{equation*}
  \nabla\cK(w) =\frac{4n}{(y^Ty)^2}X(y^3) -
      \frac{4n(y^2)^T(y^2)}{(y^Ty)^3}Xy, 
    \where y=X^Tw
\end{equation*}

and $y^m$ denotes the element wise product  *y.\textasciicircum m*  .
we then make the step $w=w+t\nabla\cK(w)$ where $t$ is chosen to maximize 

\begin{equation*}
  f(t) \equiv \cK(w+t\nabla\cK(w)).
\end{equation*}

To reveal its dependence on $t$ we set

\begin{equation*}
  z = X^T\nabla\cK(w) \quad\hbox{and express}\hskip 0.25in f(t) = n\frac{p(t)}{q^2(t)}-3
\end{equation*}

where $p$ and $q$ are the polynomials in $t$

\begin{equation*}
  p(t) = \sum_{k=0}^4p_kt^k \and q(t) = \sum_{k=0}^2 q_kt^k
\end{equation*}

where

\begin{equation*}
  \eqalign{
  p_4 &= (z^2)^Tz^2, \hskip 0.25in p_3=4(z^3)^Ty, \hskip 0.25in p_2 =  6(y^2)^Tz^2, \quad
    p_1 = 4(y^3)^Tz, \hskip 0.25in p_0 = (y^2)^Ty^2 \cr
  q_2 &=z^Tz, \hskip 0.25in q_1 =2y^Tz, \hskip 0.25in q_0=y^Ty. \cr} 
\end{equation*}

It follows that the critical points of $f$ are the roots of

\begin{equation*}
  r(t) = q(t)p'(t)-2p(t)q'(t) = \sum_{k=0}^4 r_kt^k
\end{equation*}

where

\begin{equation*}
  \eqalign{r_4 &= 2p_4q_1 - p_3q_2, \hskip 0.25in r_3 = p_3q_1 - 2p_2q_2 + 4p_4q_0, \quad
                           r_2 = 3p_3q_0 - 3p_1q_2, \cr
              r_1 &= 2p_2q_0 - p_1q_1 - 4p_0q_2, \hskip 0.25in r_0 = p_1q_0  - 2p_0q_1.\cr}
\end{equation*} 
 


\begin{exercise}\label{ex:fermi0} A matrix $A\in\maC^{2n\times 2n}$ is called  **Fermi**  when

\begin{equation}
  A^2=0 \and A^*A + AA^* = I. \label{eq:fermi1}
\end{equation}

(i) Show that $A^*A$ is an orthogonal projection and deduce that
the singular values of $A$ are 0 and 1, each with multiplicity $n$.


 (ii) If $B\in\maC^{2n\times 2n}$ is a second Fermi matrix and

\begin{equation}
  AB+BA = 0 \and AB^*+B^*A = 0 \label{eq:fermi2}
\end{equation}

show that the eigenvalues of $C\equiv A^*A + B^*B - I$ are either 1, -1, 0, or 0
with equal multiplicities. Hint:

\begin{equation*}
  A^*AC = A^*AB^*B, \hskip 0.25in B^*BC = B^*BA^*A \and A^*AB^*B = B^*BA^*A = B^*A^*AB.
\end{equation*}

So if $Cx=\la x$ then

\begin{equation*}
  A^*ACx = A^*AB^*Bx = A^*A\la x \and B^*BCx = B^*BA^*Ax = B^*B\la x
\end{equation*}

it follows that $A^*A\la x=B^*B\la x$ and so, if $\la\ne 0$,

\begin{equation*}
  Cx = 2A^*Ax - x = \la x
\end{equation*}

i.e, $A^*Ax = (\la+1)/2\, x$ and so $\la+1 = \{0,2\}$ and so $\la=\pm 1$ with
equal multiplicity.

Careful with multiplicities: if $x$ is and evec of $C$ for $\la\ne 0$ then $x$ is and evec of $A$ and $B$.
\end{exercise}


\begin{exercise}\label{ex:LSI}
We use the singular value decomposition to improve our ability
to search for terms across documents. The SVD will make associations
that are latent or implicit. We denote by $A$ our term by document
matrix and its full and reduced SVD by

\begin{equation*}
  A = Y\Sigma X^T \and A_k = Y_k\Sigma_k X_k^T.
\end{equation*}

We view a document vector as a column of $A$ or $A_k$ and record

\begin{equation*}
  A_k(:,j) = Y_k\Sigma_k X_k^T(:,j)
\end{equation*}

and so, given a query vector $q$ we consider

\begin{equation*}
  \hat q \equiv \Sigma_k^{-1}Y_k^T q
\end{equation*}

and compute the cosines

\begin{equation*}
  \cos(\theta_j) = \frac{\hat q^T X_k^T(:,j)}{\|\hat q\|\|X_k^T(:,j)\|}
\end{equation*}
\end{exercise}
