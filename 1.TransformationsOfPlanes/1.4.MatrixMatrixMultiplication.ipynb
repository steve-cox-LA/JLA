## Matrix-Matrix Multiplication

We explore the rich tapestry of products of our individual transformations. We add **reflections** to this set and note that transformations may be naturally divided into sets call **Matrix Groups**. We demonstrate that this group structure is both the consequence of geometric symmetries and the source of beautiful graphs.


### Composing Dilations, Shears, and Rotations

We recall, from our previous notebook, the basic transformations

\begin{equation} \label{eq:iwasawa}
  \mathbf{dilations}:\  A(\alpha) = \begin{bmatrix} \alpha & 0 \\ 0 & 1/\alpha \end{bmatrix} \hskip 0.25in
  \mathbf{shears}:\  N(\s) = \begin{bmatrix} 1 & \s \\ 0 & 1 \end{bmatrix} \and
  \mathbf{rotations}:\  K(\theta) = \begin{bmatrix} \cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) \end{bmatrix}
\end{equation}

and ask what happens if we follow a dilation with a shear or rotation, or *vice versa.*
For example, now that we understand matrix-vector multiplication we can simply multiply the vector $N(\s)x$ by $A(\alpha)$;

\begin{equation}
  A(\alpha)N(\s)x = \begin{bmatrix} \alpha & 0 \\ 0 & 1/\alpha \end{bmatrix} 
    \begin{bmatrix} 1 & \s \\ 0 & 1 \end{bmatrix}
    \begin{bmatrix} x[0] \\ x[1] \end{bmatrix} = \begin{bmatrix} \alpha & 0 \\ 0 & 1/\alpha \end{bmatrix}
    \begin{bmatrix} x[0]+\s x[1] \\ x[1] \end{bmatrix} =
    \begin{bmatrix} \alpha (x[0]+\s x[1]) \\ x[1]/\alpha \end{bmatrix} =
    \begin{bmatrix} \alpha & \alpha\s \\ 0 & 1/\alpha \end{bmatrix}
    \begin{bmatrix} x[0] \\ x[1] \end{bmatrix}
\end{equation}

from which we deduce that

\begin{equation}\label{eq:AxN}
  A(\alpha)N(\s) = \begin{bmatrix} \alpha & \alpha\s \\ 0 & 1/\alpha \end{bmatrix}
\end{equation}

Similarily, if we reverse the order of our composition

\begin{equation}
  N(\s)A(\alpha)x = 
    \begin{bmatrix} 1 & \s \\ 0 & 1 \end{bmatrix}\begin{bmatrix} \alpha & 0 \\ 0 & 1/\alpha \end{bmatrix} 
    \begin{bmatrix} x[0] \\ x[1] \end{bmatrix} = \begin{bmatrix} 1 & \s \\ 0 & 1 \end{bmatrix}
    \begin{bmatrix} \alpha x[0] \\ x[1]/\alpha \end{bmatrix} =
    \begin{bmatrix} \alpha x[0]+\s x[1]/\alpha \\ x[1]/\alpha \end{bmatrix} =
    \begin{bmatrix} \alpha & \s/\alpha \\ 0 & 1/\alpha \end{bmatrix}
    \begin{bmatrix} x[0] \\ x[1] \end{bmatrix}
\end{equation}

from which we deduce that

\begin{equation}\label{eq:NxA}
  N(\s)A(\alpha) = \begin{bmatrix} \alpha & \s/\alpha \\ 0 & 1/\alpha \end{bmatrix}
\end{equation}

and note that multiplication is **not commutative**, i.e., $A(\alpha)N(\s)\ne N(\s)A(\alpha)$. 
These two calculations illustrate the basic rule of **matrix-multiplication**, if $A$ and $B$ are 2-by-2 then their
product, $C$, is the 2-by-2 collection of inner products of the rows of $A$ and columns of $B$;

\begin{equation}
  C[i,j] = A[i,0]B[0,j] +  A[i,1]B[1,j]
\end{equation}

where $i$ and $j$ are each either 0 or 1. For example,

\begin{equation}
  \begin{bmatrix} 3 & 1 \\ 4 & 2 \end{bmatrix}\begin{bmatrix} 5 & 6 \\ 8 & 7 \end{bmatrix} =
  \begin{bmatrix} 3\cdot 5 + 1\cdot 8 & 3\cdot 6 + 1\cdot 7 \\ 4\cdot 5 + 2\cdot 8 & 4\cdot 6 + 2\cdot 7  \end{bmatrix}  = \begin{bmatrix} 23 & 25 \\ 36 & 38 \end{bmatrix} 
\end{equation}

This mechanical process, though easy to implement, often conceals meaningful patterns - that we now tease out in the next six exercises.

### Inverses, Eigenvalues, Determinants and Traces

We know how to invert dilations, shears, and rotations. The general case requires

\begin{exercise} \label{ex:inv2by2} Please show that if $\det(M)\ne 0$ then

\begin{equation} \label{eq:Minv}
  M^{-1} = \frac{1}{\det(M)}\begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
\end{equation}

is the **inverse** of $M$ in the sense that $MM^{-1}=M^{-1}M=I$.

\end{exercise}

From here it not hard to see that each 2-by-2 matrix obeys a quadratic equation.

\begin{exercise} \label{ex:MpMinv} Use (\ref{eq:Minv}), assuming $\det(M)\ne 0$,  to achieve

\begin{equation} \label{eq:MpMinv}
  M + \det(M) M^{-1} = (a+d)I
\end{equation}

This sum, $a+d$, of the diagonal elements occurs so frequently that we give it a name. We call it the **trace** and abbreviate it by $\tr$. To wit, $\tr(M)=a+d$. Use this notion as you multiply (\ref{eq:MpMinv}) by $M$ to achieve the
quadratic equation

\begin{equation} \label{eq:Mquad}
  M^2 - \tr(M)M + \det(M) I = 0
\end{equation}

Please explore the implications of this identity when $M$ is $A(\alpha)$, $N(\s)$, $K(\theta)$, or $H(\theta)$.

\end{exercise}

From here we learn that that the eigenvalues of each 2-by-2 matrix obey the same quadratic equation.

This says that $M$ acts like a scalar, $\la$, for vectors on the line through $q$. As there is a common $q$ on each side of (\ref{eq:eigeq0}) it makes sense to collect terms and land at

The eigenvectors of a typical symmetric matrix $M$ are revealed by the **Power Method**, i.e., by following the orbit of normalized products

\begin{equation}\label{eq:powmet}
    \frac{M^n x}{\|M^nx\|} \hskip 0.25in n=1,2,3,\ldots
\end{equation}

for any random unit vector $x\in\maR^2$. We compute these iterates, starting from 100 different $x$, for the $M$ in (\ref{eq:mat2112}) in the code cell below.

```python
# illustrate the power method, for better discrimination
# plot its orbits on the unit circle by stacking circles by iteration number, n

%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(' ')
ax = plt.axes(projection="3d")    # set up for 3D work

M = np.array([[2,-1],[-1,2]])

t = np.linspace(0,2*np.pi,100)  # all angles
z = np.ones(100)

for run in range(100):   # run 100 different orbits
    
    ax.plot3D(np.cos(t), np.sin(t), 0*z, 'k', linewidth=0.75)   # plot unit circle at height 0

    a = t[np.random.randint(0,100)]          # pick a random angle
    x = np.array([np.cos(a), np.sin(a)])     # specify starting vector

    for n in range(1,8):                      # its enough to follow 7 iterates
        y = M@x                               # get new direction
        y = y/np.sqrt(y[0]**2+y[1]**2)        # normalize
        ax.plot3D(np.cos(t), np.sin(t), n*z, 'k', linewidth=0.75)   # plot unit circle at height n
        ax.plot3D([x[0], y[0]],[x[1], y[1]], [n-1,n])  # connect subsequent points in orbit
        x = y                                 # old vec equals new vec
        
    ax.set_prop_cycle(None)    # reset colors

ax.set_xlabel('x[0]');
ax.set_ylabel('x[1]');
ax.set_zlabel('iteration');
```

We view the Power Method as a discreet dynamical system on the unit circle. If $\hat x$ is the limit of an orbit for matrix $M$ then $\hat x$ is an eigenvector of $M$ with associated eigenvalue, say $\la$. That is, $M\hat x=\la\hat x$. As $\|\hat x\|=1$ it follows immediately that

\begin{equation}\label{eq:preQ}
    \hat x^TM\hat x = \la
\end{equation}

This suggests that we explore the quadratic form

\begin{equation} \label{eq:Qxt}
  Q(\theta) \equiv x^T(\theta)Mx(\theta)  \hskip 0.25in \hbox{over unit vectors}\hskip 0.25in
  x(\theta)  \equiv \begin{bmatrix}\cos(\theta) \\ \sin(\theta) \end{bmatrix}, \hskip 0.25in 0\le\theta\le 2\pi
\end{equation}

Although running $\theta$ from $0$ to $2\pi$ covers the unit circle, in fact it covers it twice.

As a result we need only explore $Q(\theta)$ on the interval $[0,\pi]$. We do this in the code cell below for the $M$ in (\ref{eq:mat2112}).

```python
# explore the quadratic form, Q 

import numpy as np
import matplotlib.pyplot as plt

M = np.array([[2,-1],[-1,2]])
theta = np.linspace(0,np.pi,100)
Q = np.zeros(100)

for n in range(100):
    x = np.array([np.cos(theta[n]), np.sin(theta[n])])
    Q[n] = np.dot(x,M@x)

plt.plot(theta,Q)
plt.grid('on')
plt.xlabel('$\\theta$')
plt.ylabel('Q');
```

\end{equation}

As the determinant and trace are such "natural" descriptors we explore how they behave when we compose, i.e., multiply, matrices. First, as the determinant of a matrix measures the degree to which it scales areas we may guess that the change in area following a composition is the product of their individual area changes. To check this guess we set

\begin{equation}
  M = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \and N = \begin{bmatrix} s & t \\ u & v \end{bmatrix} 
\end{equation}

and compute their product

\begin{equation}
  MN = \begin{bmatrix} a & b \\ c & d \end{bmatrix}\begin{bmatrix} s & t \\ u & v \end{bmatrix} =
  \begin{bmatrix} as+bu & at+bv \\ cs+du & ct+dv \end{bmatrix}
\end{equation}

and note that it's determinant

\begin{equation} \label{eq:detprod}
  \det(MN) = (as+bu)(ct+dv) - (at+bv)(cs+du) = (ad-bc)(sv-tu) = \det(M)\det(N),
\end{equation}

as we had guessed. This deserves a box,

\begin{equation} \label{eq:detprodbox}
  \boxed{\det(MN) = \det(M)\det(N)}
\end{equation}

Do you see how this produces $\det(M^{-1})=1/\det(M)$?
Regarding the behavior of the trace under composition 
we note although $A(\alpha)N(\s)\ne N(\s)A(\alpha)$ we do observe that
$\tr(A(\alpha)N(\s))= \tr(N(\s)A(\alpha))$. To see that this is no accident, 
we compute the *other* product

\begin{equation}
  NM = \begin{bmatrix} s & t \\ u & v \end{bmatrix}\begin{bmatrix} a & b \\ c & d \end{bmatrix} =
  \begin{bmatrix} as+ct & sb+td \\ ua+vc & bu+dv \end{bmatrix}
\end{equation}

and note that summing the diagonal terms in $MN$ and $NM$ reveals

\begin{equation} \label{eq:trprod}
  \boxed{\tr(MN) = \tr(NM)}
\end{equation}

### Matrix Groups

The identities in (\ref{eq:RefRotT}) are specially helpful in cataloging and connecting the symmetries of planar objects. In Figure \ref{fig:eqtri} below we mark the reflections enjoyed by the equilateral triangle.

```python
# equilateral triangle and its Cayley graph

%matplotlib inline
import numpy as np    
import matplotlib.pyplot as plt 

fig, (ax1, ax2) = plt.subplots(1, 2)    # create a figure with 2 axes (plot windows)
plt.subplots_adjust(right=2, top=1, wspace=0.25)  # adjust them for the right look

def cmark2(xc, yc, rad, ang):     # draw custom marker 
    t = np.linspace(0, 2*np.pi, 100)
    x = xc + rad*np.cos(t)
    y = yc + rad*np.sin(t)
    ax2.fill(x, y, 'c')
    tL = np.linspace(ang, ang + np.pi, 100)
    x = xc + rad*np.cos(tL)
    y = yc + rad*np.sin(tL)
    ax2.fill(x, y, 'm')
    
a = np.pi/3

ax1.plot([0,2],[0,0],'b',linewidth=2)   # link 0
ax1.plot([1,1],[0,2*np.sin(a)],'k--')
ax1.text(1,1.8,'$H(\pi/2)$', fontsize=12)

ax1.plot([2, 2*np.cos(a)],[0, 2*np.sin(a)],'b',linewidth=2)   # link 1
ax1.plot([0,1+np.cos(a)],[0,np.sin(a)],'k--')
ax1.text(1+np.cos(a)+0.1,np.sin(a),'$H(\pi/6)$', fontsize=12)
          
ax1.plot([2*np.cos(a),0],[2*np.sin(a),0],'b',linewidth=2)   # link 2
ax1.plot([2,1-np.cos(a)],[0,np.sin(a)],'k--')
ax1.text(1-np.cos(a)-0.5,np.sin(a),'$H(5\pi/6)$', fontsize=12)
ax1.text(1.1,0.85,'$\pi/3$',fontsize=12)

ax1.axis('equal')
ax1.axis('off')

s3 = np.sqrt(3);

ax2.text(0.1,1,'I',fontsize=12)
ax2.arrow(0, 1, s3/2, -1/2-1, color = 'b', width=0.025, length_includes_head = True)
cmark2(-0.15, 1.1, 0.1, 3*np.pi/2  )
ax2.arrow(s3/2, -1/2, -2*s3/2, 0, color = 'b', width=0.025, length_includes_head = True)
cmark2(s3/2-0.15, -.7, 0.1, 3*np.pi/2 + 2*np.pi/3 )
ax2.arrow(-s3/2, -1/2, s3/2, 3/2, color = 'b', width=0.025, length_includes_head = True)
cmark2(-s3/2-0.15, -.4, 0.1, 3*np.pi/2 + 4*np.pi/3 )

ax2.text(s3/2-0.7,-1/2+.1,'$K(2\pi/3)$',fontsize=12)
ax2.text(-s3/2,-1/2-.2,'$K(4\pi/3)$',fontsize=12)

ax2.plot([0, 0],[1, 2],'r',linewidth=2)
ax2.plot([s3/2, s3],[-1/2,-1],'r',linewidth=2)
ax2.plot([-s3/2, -s3],[-1/2, -1],'r',linewidth=2)

ax2.arrow(0, 2, -s3, -1-2, color = 'b', width=0.025, length_includes_head = True)
ax2.arrow(-s3, -1, 2*s3, 0, color = 'b', width=0.025, length_includes_head = True)
ax2.arrow(s3, -1, -s3, 3, color = 'b', width=0.025, length_includes_head = True)

ax2.text(0.1,2,'$H(\pi/2)$',fontsize=12)
cmark2(-0.2, 2, 0.1, np.pi/2 )

ax2.text(s3,-1+.1,'$H(\pi/6)$',fontsize=12)
cmark2(s3+.2, -1.1, 0.1, -np.pi/6 )

ax2.text(-s3,-1-.2,'$H(5\pi/6)$',fontsize=12)
cmark2(-s3-.2, -1.1, 0.1, -5*np.pi/6 )
ax2.axis('equal')
ax2.axis('off');
```

\begin{definition}\label{def:MG}
A collection $G$ of 2-by-2 matrices is called a **matrix group** when

(1) If $M\in G$ then $M^{-1}\in G$, and

(2) If $M\in G$ and $N\in G$ then $MN\in G$.
\end{definition}

### Groups of Integer Matrices

This last exercise, together with Exercises \ref{ex:Mob1} and \ref{ex:gps}, implies that $\SL_2(\maR)$ is a product of the subgroups of dilations, shears, and rotations. We will have cause to study quite a few other subgroups of $\SL_2(\maR)$. For example, if we restrict ourselves to matrices with integer entries and determinant 1, then it follows from (\ref{eq:Minv}) that these matrices have integer inverses with determinant 1. Denoting the integers by $\maZ$ we see that $\SL_2(\maZ)$ forms a subgroup of $\SL_2(\maR)$. We note that $I$ is the only dilation in $\SL_2(\maZ)$ while it contains the group of four rotations $\{K(n\pi/2):n=1,2,3,4\}$ and the infinite group of shears

\begin{equation}\label{eq:zshear}
  \begin{bmatrix} 1 & n \\ 0 & 1 \end{bmatrix}, \hskip 0.25in n\in\maZ.
\end{equation}

We next show that this **particular** rotation and **particular** shear 

\begin{equation}\label{eq:zgen}
  A=K(\pi/2)=\begin{bmatrix} 0 & -1 \\ 1 & 0 \end{bmatrix} \and
  B=N(1)=\begin{bmatrix} 1 & 1 \\ 0 & 1 \end{bmatrix}
\end{equation}

together with their inverses, 

\begin{equation}\label{eq:zgeninv}
  A^{-1}=\begin{bmatrix} 0 & 1 \\ -1 & 0 \end{bmatrix} = -A \and
  B^{-1}=\begin{bmatrix} 1 & -1 \\ 0 & 1 \end{bmatrix}
\end{equation}

generate all of $\SL_2(\maZ)$. This kind of statement will become a theme throughout the coming notebooks, where, e.g., we will explore the tiling or *tesselation* of the plane by particular transformations and we will also show that all qubit *gates* can be effectively generated by a particular pair of transformations.

\begin{proposition}\label{prop:genz} Each $M\in\SL_2(\maZ)$ can be written as a finite product of $A,B, A^{-1}$ and $B^{-1}$ from (\ref{eq:zgen}-\ref{eq:zgeninv}).
\end{proposition}

**Proof:** We write $M\in\SL_2(\maZ)$

\begin{equation}\label{eq:genz1}
  M = \begin{bmatrix} a & b \\ c & d \end{bmatrix}
\end{equation}

and note that if $c=0$ then, as $ad-bc=1$, $M$ must be either

\begin{equation}\label{eq:genz2}
  M_+ =\begin{bmatrix} 1 & b \\ 0 & 1 \end{bmatrix} \hskip 0.25in \hbox{or}\hskip 0.25in
  M_- =\begin{bmatrix} -1 & b \\ 0 & -1 \end{bmatrix} 
\end{equation}

but $M_+=B^b$ while $M_-=-B^{-b}=A^2B^{-b}$ are both products of powers of $A$ and $B$. Hence, to complete the proof we must show that $\{A,B,A^{-1},B^{-1}\}$ can used to zero out the lower left element in $M$. To begin, we note that

\begin{equation}\label{eq:genz3}
  AM = \begin{bmatrix} -c & -d \\ a & b \end{bmatrix} \and
  B^nM = \begin{bmatrix} a+nc & b+nd \\ c & d \end{bmatrix} 
\end{equation}

and so will use $B^n$ to modify the upper left element and then use $A$ to exchange the upper left and the lower left elements.
In particular,  if $|a|\ge |c|$, divide $a$ by $c$, i.e., express $a = cq + r$ with integer quotient $q$ and remainder $r$ and 
$0 \le r < |c|$. Now, by (\ref{eq:genz3}), 

\begin{equation}\label{eq:genz4}
  B^{-q}M = \begin{bmatrix} a-qc & b-qd \\ c & d \end{bmatrix} = \begin{bmatrix} r & b-qd \\ c & d \end{bmatrix} \and
  AB^{-q}M = \begin{bmatrix} -c & -d \\ r & b-qd \end{bmatrix} 
\end{equation}

and indeed $0 \le r < |c|$. Now, if $r\ne 0$ we simply repeat this process. As it is guaranteed to diminish the magnitude of the lower left entry by at least one at each iteration then after a finite number of steps we must reach a zero remainder.  **ALMOST End of Proof.**

We have coded this *triangulator* in the cell below on the matrix

\begin{equation}\label{eq:ConZ}
 M = \begin{bmatrix} 17 & 29 \\ 7 & 12\end{bmatrix} 
\end{equation}

and so find $M=B^2A^{-1}B^{-3}A^{-1}B^{-2}A^{-1}B^{-2}AB$. The intermediate steps of the *triangulator* are illustrated in Figure \ref{fig:trimaker}.

```python
# SL2(Z) triangulator,
# given M express it as product of powers of the generators A and B by driving
# its lower left element to zero

import numpy as np
import matplotlib.pyplot as plt

A = np.array([[0,-1],[1,0]])

M = np.array([[17,29],[7,12]])
#M = np.array([[10,11],[9,10]])
#M = np.array([[7,3],[2,1]])
word = 'M'
if ( abs(M[1,0]) > abs(M[0,0])):
    M = A@M
    word = 'A' + word

iter = 0
cdat = abs(M[1,0])
plt.plot(iter, cdat, 'ro')
plt.text(iter, cdat+0.2, '$' + word + '$', rotation=90)

while ( abs(M[1,0]) > 0.1 and iter < 25):    # go until M[1,0] vanishes
    a = M[0,0]
    c = M[1,0]
    r = a % c            # remainder
    q = int((a-r)/c)     # quotient
    M = A@np.array([[1,-q],[0, 1]])@M
    word = 'AB^{' + str(-q) + '}' + word
    iter = iter + 1
    cdat = np.append(cdat, abs(M[1,0]))   
    plt.plot(iter, abs(M[1,0]), 'ro')                 # mark the c level
    plt.text(iter, abs(M[1,0])+0.2, '$' + word + '$', rotation=90)   # label it with present word
    
plt.plot(cdat,'k')
    
word = word + ' = '  # prepare the right hand side
a = M[0,0]
b = int(M[0,1])
if a < 0:
    word = word + 'A^2'
    b = -b
word = word + 'B^{' + str(b) + '}'

plt.title('$' + word + '$')
plt.grid('on')
plt.box(False)
plt.xlabel('iteration')
plt.ylabel('|c value|');
```

### Congruence Subgroups

**These appear naturally as kernel of modular reduction, see wiki page**

We obtain subgroups of $\SL_2(\maZ)$ by restricting to matrices that are congruent to the identity mod $N$ for integer $N\ge 1$. 

\begin{exercise}\label{ex:Congruent} Define, for integer $N\ge 1$,

\begin{equation}\label{eq:cong}
  \Gamma(N) \equiv \{M\in\SL_2(\maZ): M\equiv I\modu N\}.
\end{equation}

Show that if $A\in\Gamma(N)$ and $B\in\Gamma(N)$ then $AB\in\Gamma(N)$ and conclude that $\Gamma(N)$ is a subgroup of $\SL_2(\maZ)$.
\end{exercise}

To see that $\Gamma(N)$ is an infinite group note that the shear

\begin{equation}\label{eq:zmshear}
  \begin{bmatrix} 1 & mN \\ 0 & 1 \end{bmatrix}
\end{equation}

belongs to $\Gamma(N)$ for every $m\in\maZ$. We obtain **finite** subgroups of $\SL_2(\maZ)$ by replacing "regular" arithmetic with **modular** arithmetic. Recall that two integers, $m$ and $n$, are said to be **congruent** modulo $p$, written $m\equiv n\modu p$, when their difference is a multiple of $p$. In the simplest case, $p=2$, we note that $1+1=2\equiv 0\modu 2$ and $-1\equiv 1 \modu 2$. Then $\SL_2(2)$, the 2-by-2 matrices with elements from $\{0,1\}$ and determinant 1 under mod 2 arithmetic are simply

\begin{equation}\label{eq:sl22}
\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}, \hskip 0.25in
\begin{bmatrix} 1 & 1 \\ 0 & 1 \end{bmatrix}, \hskip 0.25in
\begin{bmatrix} 1 & 0 \\ 1 & 1 \end{bmatrix}, \hskip 0.25in
\begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}, \hskip 0.25in
\begin{bmatrix} 1 & 1 \\ 1 & 0 \end{bmatrix}, \and
\begin{bmatrix} 0 & 1 \\ 1 & 1 \end{bmatrix}
\end{equation}

\begin{exercise}\label{ex:SL2z2} Show that the determinant of each matrix in (\ref{eq:sl22}) is 1 mod 2. 
Show that $\SL_2(2)$ is a group and, moreover, that it has the same multiplication table and Cayley graph as $\Dih_3$ with

\begin{equation}
\begin{bmatrix} 1 & 1 \\ 1 & 0 \end{bmatrix} \and \begin{bmatrix} 1 & 1 \\ 0 & 1 \end{bmatrix}
\end{equation}

"acting" like $K(2\pi/3)$ and $H(\pi/2)$, respectively.
\end{exercise}

On moving up to mod 3 arithmetic we may use the fact that $2\cdot 2\equiv 1\modu 3$ to see that each nonzero in $\{0,1,2\}$ has a mod 3 multiplicative inverse. As a result, in building a matrix in 
$\SL_2(3)$ we note that there are $3^2-1$ ways of choosing the first (nonzero) column, say $(a,b)$. As the second column, $(c,d)$, must be chosen so that $ad-bc=1 \modu 3$ we note that if $a\ne 0$ then $d=(1-bc)a^{-1}$ where $c$ is arbitrary, while if $a=0$ then $c=-1b^{-1}$ and $d$ is arbitrary. Hence, regardless of $a$, there are 3 choices for the second column. As a result, there are 24 matrices in $\SL_2(3)$. To move beyond mod 3 we need multiplicative inverses to accomodate the determinant equation. Do you see why this requires prime congruences? Please generalize the argument above to confirm that.

\begin{exercise}\label{ex:SL2p} For prime $p$, $\SL_2(p)$, the 2-by-2 matrices with elements from $\{0,1,\ldots,p-1\}$ and determinant 1 under mod p arithmetic, is a group with $p(p^2-1)$ matrices.
\end{exercise}

\begin{proposition}\label{prop:genSL2p}
The group $\SL_2(p)$ is generated by

\begin{equation}\label{eq:genSL2p}
  U = \begin{bmatrix} 1 & 1 \\ 0 & 1 \end{bmatrix} \and L = \begin{bmatrix} 1 & 0 \\ 1 & 1 \end{bmatrix}
\end{equation}

\end{proposition}

**Proof:** Writing $M\in \SL_2(p)$ as

\begin{equation}
  M = \begin{bmatrix} a & b \\ c & d \end{bmatrix}
\end{equation}

we note first that if $c\ne 0$ then

\begin{equation}\label{eq:triple}
  \begin{bmatrix} 1 & (a-1)/c \\ 0 & 1 \end{bmatrix} \begin{bmatrix} 1 & 0 \\ c & 1 \end{bmatrix}
  \begin{bmatrix} 1 & (d-1)/c \\ 0 & 1 \end{bmatrix} = \begin{bmatrix} a & (d-1)/c + d(a-1)/c \\ c & d \end{bmatrix} = M
\end{equation}

because $(d-1)/c + d(a-1)/c=(ad-1)/c=(ad-(ad-bc))/c = b$. Note that, in terms of $U$ and $L$ this reads

\begin{equation}\label{eq:trip1}
  M = U^{(a-1)/c}L^cU^{(d-1)/c}
\end{equation}

Now if $c=0$ then $ad=1$ and so

\begin{equation}\label{eq:trip2}
   \begin{bmatrix} a+b & b \\ d & d \end{bmatrix} \in \SL_2(p)
\end{equation}

with a nonzero lower left entry and so can be written with $U$ and $L$ as above. It remains to note that this latter matrix is easily transformed to $M$;

\begin{equation}\label{eq:trip3}
  \begin{bmatrix} a+b & b \\ d & d \end{bmatrix}\begin{bmatrix} 1 & 0 \\ p-1 & 1 \end{bmatrix}  
    = \begin{bmatrix} a & b \\ 0 & d \end{bmatrix} = M
\end{equation}

That is, $M = U^{(a+b-1)/d}L^dU^{(d-1)/d}L^{p-1}$ when $c=0$. **End of proof.**
