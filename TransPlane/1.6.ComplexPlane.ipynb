## The Complex Plane

In our study of M&ouml;bius transformations, 

\begin{equation}\label{eq:mob0}
  \mu(t) \equiv \frac{at+b}{ct+d}, \where a,b,c,d,t\in \maR \and ad-bc = 1,
\end{equation}

of the Projective Line we found that the nature of their fixed points, i.e., solutions to $\mu(t)=t$, was determined by the trace of the associated matrix

\begin{equation}\label{eq:mob1}
  M \equiv \begin{pmatrix} a & b \\ c & d \end{pmatrix}
\end{equation}

As $\mu(t)=t$ is nothing other that the quadratic equation

\begin{equation}\label{eq:muq}
    ct^2 + (d-a)t  - b = 0
\end{equation}

we learned that when $c\ne 0$ that $\mu$ has

 (a) 2 fixed points when $|\tr(M)| > 2$

 (b) 1 fixed point when $|\tr(M)| = 2$, and

 (c) 0 fixed points when $|\tr(M)| < 2$

You likely have already realized that condition (c) is a consequence of our focus on *real* fixed points. The culprit is that the radicand $(d+a)^2-4$ is negative when $|\tr(M)|<2$. Fortunately, *all* negative radicands are accomodated by the declaration of the single imaginary unit

\begin{equation}\label{eq:deci}
  \boxed{i \equiv \sqrt{-1}}
\end{equation}

Given two real numbers, $x$ and $y$, we attach $i$ to $y$ and generate the **complex number**

\begin{equation}\label{eq:z}
  z = x + iy
\end{equation}

and call $x$ the  **real**  part and $y$ the **imaginary**  part of $z$. 
In what follows we explore the rich arithmetic and geometry associated with this move.

### Complex Arithmetic

Regarding arithmetic we name two complex numbers 

\begin{equation*}
  z_1 = x_1 + iy_1 \and z_2 = x_2 + iy_2
\end{equation*}

and **define** their sum and difference to be

\begin{equation}\label{eq:zsd}
  z_1+z_2 \equiv (x_1+x_2)+i(y_1+y_2)  \and z_1-z_2 \equiv (x_1-x_2)+i(y_1-y_2) 
\end{equation}

and their product to be

\begin{equation}\label{eq:zprod}
  z_1z_2 \equiv (x_1 + iy_1)(x_2 + iy_2) = 
                       (x_1x_2-y_1y_2)+i(x_1y_2+x_2y_1).
\end{equation}

The fact that the product of $z =x+iy$ and its **conjugate**, $\ov z = x-iy$, is

\begin{equation}\label{eq:zpc}
  z\ov z \equiv (x + iy)(x - iy) = x^2 + y^2
\end{equation}

is real, suggests a strategy for computing the quotient of

\begin{equation}\label{eq:zq}
    \frac{z_1}{z_2} = \frac{z_1}{z_2}\frac{\ov z_2}{\ov z_2}
                  = \frac{(x_1x_2+y_1y_2)+i(x_2y_1-x_1y_2)}{x_2^2+y_2^2}
\end{equation}

The fact that (\ref{eq:zpc}) is not only real but positive (not to mention pythagorean) also suggests that we **define** the **magnitude** of $z$ to be

\begin{equation}\label{eq:zmag}
              |z| \equiv \sqrt{z\ov z} = \sqrt{x^2+y^2}.
\end{equation}

In addition to the **Cartesian representation**, $z=x+iy$, one also has the
**polar form**

\begin{equation}
  z = |z|(\cos\theta+i\sin\theta), \label{eq:polarform}
\end{equation}

where $|z|$ is the magnitude of $z$ and $\theta$ is the angle that it make
with the positive real axis. We have illustrated these various representations in Figure \ref{fig:cplane}(Left).

The polar form is especially convenient with regards to multiplication. More precisely,

\begin{equation*}
  \eqalign{z_1z_2 &= |z_1||z_2|\{(\cos\theta_1\cos\theta_2-\sin\theta_1\sin\theta_2)+
                              i(\cos\theta_1\sin\theta_2+\sin\theta_1\cos\theta_2)\} \cr
                  &= |z_1||z_2|\{\cos(\theta_1+\theta_2)+i\sin(\theta_1+\theta_2)\}. \cr}
\end{equation*}

As a result, for integer values of $n$, we find

\begin{equation*}
  z^n = |z|^n(\cos(n\theta)+i\sin(n\theta)).
\end{equation*}

This formula dictates that taking powers of a complex number simultaneously 
scales its magnitude and rotates its argument. The spiral in Figure \ref{fig:cplane}(Right)
offers a concrete illustration.

The polar form, (\ref{eq:polarform}), takes on even simpler guise once Euler's formula

\begin{equation}\label{eq:EuF}
  \boxed{\exp(i\theta) = \cos(\theta)+i\sin(\theta)}, \hskip 0.25in \theta\in\maR.
\end{equation}

This comes from the power series definition, for complex $z$,

\begin{equation}\label{eq:ExPow}
  \exp(z) = \sum_{n=0}^\infty \frac{z^n}{n!}
\end{equation}

For if $z=i\theta$ then

\begin{equation}\label{eq:ExPow1}
  \exp(i\theta) = \sum_{n=0}^\infty \frac{i^n\theta^n}{n!}
\end{equation}

which Exercise \ref{ex:ipow} reveals to be

\begin{equation}\label{eq:ExPow2}
  \eqalign{
  \exp(i\theta) &= \sum_{n=0}^\infty \frac{i^{2n}\theta^{2n}}{(2n)!} + \frac{i^{2n+1}\theta^{2n+1}}{(2n+1)!} \cr
  &= \sum_{n=0}^\infty \frac{(-1)^n\theta^{2n}}{(2n)!} + \frac{i(-1)^n\theta^{2n+1}}{(2n+1)!} \cr
  &= \cos(\theta) + i\sin(\theta) }
\end{equation}

as claimed. With (\ref{eq:EuF}) now comes the compact polar form

\begin{equation}\label{eq:PF2}
  z = |z|\exp(i\theta)
\end{equation}

### M&ouml;bius Transformations

We now pursue **complex** M&ouml;bius Transformations

\begin{equation}\label{eq:mobz}
  \mu(z) \equiv \frac{az+b}{cz+d}, \where a,b,c,d,\in \maC \and ad-bc = 1
\end{equation}

for **complex** argument $z$. We saw in the previous notebook every real M&ouml;bius Transformation could be expressed as a compositions of a dilation, $t\mapsto\alpha^2t$, translation, $t\mapsto t+\sigma$, and rotation by $\pi/2$, $t\mapsto -1/t$. These retain their status in the complex setting, with real dilation branching into dilation of magnitude and translation of phase. 

\begin{proposition}\label{prop:mobC} Each complex M&ouml;bius Transformation can be expressed as a composition of the four

\begin{equation}\label{eq:Mob3T}
  \eqalign{
  A_\alpha(z) &\equiv \alpha z, \hskip 0.25in \alpha>0 \cr
  R_\theta(z) &\equiv \exp(i\theta)z, \hskip 0.25in \theta\in\maR \cr
  N_\sigma(z) &\equiv z + \s z, \hskip 0.25in \s\in\maC \cr
  K_{\pi/2}(z) &\equiv -1/z \cr}
\end{equation}

elementary transformations.
\end{proposition}

**Proof:** If $c=0$ then

\begin{equation}\label{eq:Mobd1}
   \frac{az+b}{cz+d} = (a/d)z + (b/d) = N_{b/d}(R_\theta(A_\alpha(z))), \where a/d = \alpha\exp(i\theta)
\end{equation}

If $c\ne 0$ then

\begin{equation}\label{eq:Mobd2}
   \frac{az+b}{cz+d} = \frac{a}{c} - \frac{1}{c(cz+d)} = N_{a/c}(K_{\pi/2}(N_{cd}(R_\theta(A_\alpha(z)))) \where c^2 = \alpha\exp(i\theta)
\end{equation}

**End of Proof.**

Even better, $\cM$ is transitive on $\cU$: if $z_1\in\cU$ and $z_2\in\cU$ then there exists a $\mu\in\cM$ such that $\mu(z_1)=z_2$. In particular, suppose that $z_k=\alpha_k i+\beta_k\in\cU$, then $\alpha_k>0$ and (\ref{eq:MobAff}) provides 

\begin{equation}\label{eq:MobAffk}
\mu_k(z) \equiv \alpha z + \beta, \where \mu_k(i) = z_k
\end{equation}

As a result $\mu(z)\equiv \mu_2(\mu_1^{-1}(z))$ takes $z_1$ to $i$ and then $i$ to $z_2$.

We now return to our classification

 (a) When $|\tr(M)| = 2$ then $\mu$ has one real fixed point and is call **parabolic**. If $\alpha$ is the fixed point of $\mu$ then, $S(z) = 1/(z-\alpha)$ maps $\alpha$ to $\infty$ and so $W=S\mu S^{-1}$ is parabolic and takes $\infty$ to
$\infty$ and so $W(z) = z + t$ for some nonzero $t\in\maR$. With $V(z)=z/|t|$ we find that $VWV^{-1}$ is $z\pm 1$. We will see in the exercise below that $z+1$ is *not* conjugate to $z-1$.
 
 (b) When $|\tr(M)| > 2$ then $\mu$ has two real fixed points and is call **hyperbolic**. If $\alpha$ and $\beta$ are the two fixed points we note that $S(z) \equiv (z-\alpha)/(z-\beta)$ obeys $S(\alpha)=0$ and $S(\beta)=\infty$ and so $W=S\mu S^{-1}$ is hyperbolic and fixes $0$ and $\infty$ and so $W(z)=tz$ for some $t\in\maR\setminus\{0,1\}$. With $B(z)=-1/z$ then $(BWB^{-1})(z) = (1/t)z$. We will see in the exercise below that $tz$ is *not* conjugate to $sz$ unless $t=s$ or $t=1/s$.

 (c) When $|\tr(M)| < 2$ then $\mu$ has two complex (conjugate) fixed points and is call **elliptic**.


```python
# Mobius widget on SL_2(R), plot phase of mu(z)-z to see fixed points

%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import *

fig = plt.figure(' ')
x = np.linspace(-4, 4, 1000)
X, Y = np.meshgrid(x, x)
Z = X + 1j*Y
data = np.angle(Z)

im = plt.imshow(data, interpolation='nearest', 
                            origin='lower', 
                            extent = [-4,4,-4,4], 
                            vmin=np.min(data),
                            vmax=np.max(data), 
                            cmap='hsv')

cb = plt.colorbar(im)

def update( sigma=widgets.FloatSlider(min=-10, max=10, step=0.2, value=0, description='shear: $\sigma$'), 
            alpha=widgets.FloatSlider(min=0.1, max=10, step=0.1, value=1.5, description='dilation: $\\alpha$'), 
            theta=widgets.FloatSlider(min=0, max=2*np.pi, step=np.pi/50, value=0, description='rotation: $\\theta$') ):
    
    N = np.array([[1,sigma],[0,1]])
    A = np.array([[alpha,0],[0, 1/alpha]])
    K = np.array([[np.cos(theta), -np.sin(theta)],[np.sin(theta), np.cos(theta)]])
    M = N@A@K
    mu = (M[0,0]*Z+M[0,1])/(M[1,0]*Z+M[1,1])
    im.set_data(np.angle(mu-Z))
    fig.canvas.draw_idle()

interact(update);
```

We can now return to unfinished business, namely the two complex fixed points of 

\begin{equation}\label{eq:Ktz}
  K_\theta(z) = \frac{\cos(\theta)z - \sin(\theta)}{\sin(\theta)z + \cos(\theta)}
\end{equation}

This has the fixed points $\pm i$ and so

\begin{equation}\label{eq:eta2}
  \eta(z) \equiv \frac{z-i}{z+i}
\end{equation}

maps the fixed points to $0$ and $\infty$ and so $\xi(z)\equiv \eta\circ K_\theta \circ\eta^{-1}$ fixes both $0$ and $\infty$.

Can we prove that the trajectory is a circle?  If $z=ai$ then

\begin{equation}\label{eq:c1}
  \mu_n = \frac{aic_n - s_n}{ais_n + c_n} = \frac{(aic_n - s_n)(-ais_n + c_n)}{c_n^2+a^2s_n^2}
    = \frac{s_nc_n(a^2-1) + ia}{c_n^2+a^2s_n^2} = ib + r\exp(it)
\end{equation}

when $s_n=0$ we know $c_n^2=1$ so

\begin{equation}\label{eq:c2}
    ia = ib + r\exp(it)
\end{equation}

and so $\cos(t)=0$ and $\sin(t)=\pm 1$ so $a=b\pm r$.

similarly, when $c_n=0$ we get $s_n^2=1$ so

\begin{equation}\label{eq:c3}
    i/a = ib + r\exp(it')
\end{equation}

and so $\cos(t')=0$ and $\sin(t')=\pm 1$ so $1/a=b\mp r$.  Together this gives, assuming $a>1$ that

\begin{equation}\label{eq:c4}
    a = b+r,\ 1/a = b-r
\end{equation}

adding them up gives $a+1/a=2b$ so $b=(a+1/a)/2$ and so $r=(a-1/a)/2$. Let's now check the magnitude of

\begin{equation}\label{eq:c5}
  \mu_n - ib = \frac{s_nc_n(a^2-1) + i(a-(a+1/a)(c_n^2+a^2s_n^2)/2)}{c_n^2+a^2s_n^2}
\end{equation}

well

\begin{equation}\label{eq:c6}
  |\mu_n - ib|^2 = \frac{s_n^2c_n^2(a^2-1) + (a-(a+1/a)(c_n^2+a^2s_n^2)/2)^2}{(c_n^2+a^2s_n^2)^2}
\end{equation}

in the special case $s_n^2=c_n^2=1/2$ this becomes

\begin{equation}\label{eq:c7}
  \eqalign{|\mu_n - ib|^2 
   &= \frac{(a^2-1)/4 + (a-(a+1/a)(1+a^2)/4)^2}{(1+a^2)^2/4} \cr
   &= \frac{(a^2-1)/4 + (a-(a+a^3+1/a+a)/4)^2}{(1+a^2)^2/4} \cr
   &= \frac{(a^2-1)/4 + (a-(a+1/a)(1+a^2)/4)^2}{(1+a^2)^2/4} \cr
   &= \frac{(a^2-1)/4 + (a/2-1/(4a)-a^3/4)^2}{(1+a^2)^2/4} \cr
   &= \frac{(a^2-1) + (a-1/(2a)-a^3/2)^2}{(1+a^2)^2} \cr
   &= \frac{a^2-1 + a^2 - 2 - a^4 + a^2/2 + 1/(4a^2) + a^6/4}{(1+a^2)^2} \cr
   &= \frac{(5/2)a^2-3 - a^4 + 1/(4a^2) + a^6/4}{(1+a^2)^2}}
\end{equation}

is this 

$$
  r^2 = (a-1/a)^2/4 = (a^2-2-1/a^2)/4
$$
