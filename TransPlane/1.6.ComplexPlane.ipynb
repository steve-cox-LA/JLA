# 1.6 The Complex Plane

In our study of M&ouml;bius transformations, 

\begin{equation}\label{eq:mob0}
  \mu(t) \equiv \frac{at+b}{ct+d}, \where a,b,c,d,t\in \maR \and ad-bc = 1,
\end{equation}

of the Projective Line we found that the nature of their fixed points, i.e., solutions to $\mu(t)=t$, was determined by the trace of the associated matrix

\begin{equation}\label{eq:mob1}
  M \equiv \begin{pmatrix} a & b \\ c & d \end{pmatrix}
\end{equation}

As $\mu(t)=t$ is nothing other that the quadratic equation

\begin{equation}\label{eq:muq}
    ct^2 + (d-a)t  - b = 0
\end{equation}

we learned that $\mu$ has

 (a) 2 fixed points when $|\tr(M)| > 2$

 (b) 1 fixed point when $|\tr(M)| = 2$, and

 (c) 0 fixed points when $|\tr(M)| < 2$

You likely have already realized that condition (c) is a consequence of our focus on real fixed points. The culprit is that the radicand $(d+a)^2-4$ is negative when $|\tr(M)|<2$. All negative radicands are accomodated by the declaration of the imaginary unit

\begin{equation}\label{eq:deci}
  \boxed{i \equiv \sqrt{-1}}
\end{equation}

Given two real numbers, $x$ and $y$, we attach $i$ to $y$ and generate the **complex number**

\begin{equation}\label{eq:z}
  z = x + iy
\end{equation}

In case we call $x$ the  **real**  part and $y$ the **imaginary**  part of $z$. 
In what follows we explore the rich arithmetic and geometry associated with this move.

Regarding arithmetic we name two complex numbers 

\begin{equation*}
  z_1 = x_1 + iy_1 \and z_2 = x_2 + iy_2
\end{equation*}

and **define** their sum and difference to be

\begin{equation}\label{eq:zsd}
  z_1+z_2 \equiv (x_1+x_2)+i(y_1+y_2)  \and z_1-z_2 \equiv (x_1-x_2)+i(y_1-y_2) 
\end{equation}

and their product to be

\begin{equation}\label{eq:zprod}
  z_1z_2 \equiv (x_1 + iy_1)(x_2 + iy_2) = 
                       (x_1x_2-y_1y_2)+i(x_1y_2+x_2y_1).
\end{equation}

The fact that the product of $z =x+iy$ and its **conjugate**, $\ov z = x-iy$, is

\begin{equation}\label{eq:zpc}
  z\ov z \equiv (x + iy)(x - iy) = x^2-y^2
\end{equation}

is real, suggests a strategy for computing the quotient of

\begin{equation}\label{eq:zq}
    \frac{z_1}{z_2} = \frac{z_1}{z_2}\frac{\ov z_2}{\ov z_2}
                  = \frac{(x_1x_2+y_1y_2)+i(x_2y_1-x_1y_2)}{x_2^2+y_2^2}
\end{equation}

The fact that (\ref{eq:zpc}) is not only real but positive (not to mention pythagorean) also suggests that we **define** the **magnitude** of $z$ to be

\begin{equation}\label{eq:zmag}
              |z| \equiv \sqrt{z\ov z} = \sqrt{x^2+y^2}.
\end{equation}

In addition to the **Cartesian representation**, $z=x+iy$, one also has the
**polar form**

\begin{equation}
  z = |z|(\cos\theta+i\sin\theta), \label{eq:polarform}
\end{equation}

where $|z|$ is the magnitude of $z$ and $\theta$ is the angle that it make
with the positive real axis. We have illustrated these various representations in Figure \ref{fig:cplane}(Left).

```python
# cplane

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig, (ax1, ax2) = plt.subplots(1, 2)
plt.subplots_adjust(right=2, top=1, wspace=0.25)

#ax1.figure(figsize=(6,6))
ax1.plot([-.25, 1.25],[0, 0],'k')
ax1.plot([0, 0],[-1.25, 1.25],'k')
ax1.plot([0, 1],[0, 1],'k','linewidth',1.5)
ax1.plot(1,1,'k.',markersize=12)
ax1.plot([0, 1],[0, -1],'k')
ax1.plot(1,-1,'k.',markersize=12)
ax1.plot([0, 1],[1, 1],'k--')
ax1.plot([0, 1],[-1, -1],'k--')
ax1.plot([1, 1],[-1, 1],'k--')
ax1.text(1.05,0.075,'$x$',fontsize=14)
ax1.text(-.25,1,'$iy$',fontsize=14)
ax1.text(-.35,-1,'$-iy$',fontsize=14)
ax1.text(.25,.1,'$\\theta$',fontsize=14)
ax1.text(1.05,1.05,'$z$',fontsize=14)
ax1.text(1.05,-1.05,'$\overline{z}$',fontsize=14)
ax1.text(.3,.6,'$|z|$',fontsize=14)

ax1.axis('equal')
ax1.axis('off')

z = (2/3)*(1+1j)
z0 = z
for n in range(2,62):
    z1 = z0*z
    ax2.plot([z0.real, z1.real],[z0.imag, z1.imag],'r-+',linewidth=1.5)
    z0 = z1

ax2.axis('equal')
ax2.grid('on')
ax2.set_xlabel('real part',fontsize=14)
ax2.set_ylabel('imaginary part',fontsize=14);
```

The polar form is especially convenient with regards to multiplication. More precisely,

\begin{equation*}
  \eqalign{z_1z_2 &= |z_1||z_2|\{(\cos\theta_1\cos\theta_2-\sin\theta_1\sin\theta_2)+
                              i(\cos\theta_1\sin\theta_2+\sin\theta_1\cos\theta_2)\} \cr
                  &= |z_1||z_2|\{\cos(\theta_1+\theta_2)+i\sin(\theta_1+\theta_2)\}. \cr}
\end{equation*}

As a result, for integer values of $n$, we find

\begin{equation*}
  z^n = |z|^n(\cos(n\theta)+i\sin(n\theta)).
\end{equation*}

This formula dictates that taking powers of a complex number simultaneously 
scales its magnitude and rotates its argument. The spiral in Figure \ref{fig:cplane}(Right)
offers a concrete illustration.

We may now return to our original fixed-point quadratic, (\ref{eq:muq}), and track its complex roots in a situation where $|\tr(M)|<2$. We take the concrete case

\begin{equation}\label{eq:mobcon}
 M = \begin{pmatrix} 1/2 & b \\ -1/(2b) & 1 \end{pmatrix}
\end{equation}

and track the roots of (\ref{eq:muq}) for $-1\le b\le 1$.

```python
# track the roots of cz^2 + (d-a)z - b for fixed a and d as b spans an interval

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()

a = 0.5
d = 1
B = np.linspace(-1,1,20)

for b in B:
    c = (a*d-1)/b     # choose c to guarantee   ad - bc = 1
    r = np.roots([c, d-a, -b])  # 2 roots of the quadratic 
    rp = r[0]
    rm = r[1]
    plt.plot(rp.real, rp.imag, 'k+')
    plt.plot(rm.real, rm.imag, 'ro')
    
plt.grid('on')
plt.xlabel('real part',fontsize=14)
plt.ylabel('imaginary part',fontsize=14);
```

A complex vector (matrix) is simply a vector (matrix) of complex numbers.
Vector and matrix addition proceed, as in the real case, from element-wise
addition. The inner product of two complex vectors requires, however,
some care. This is evident when we try to use the old notion to
define the length of complex vector. To wit, note that if

\begin{equation*}
  z = \begin{bmatrix}1+i \\ 1-i \end{bmatrix}
\end{equation*}

then

\begin{equation*}
  z^Tz = (1+i)^2 + (1-i)^2 = 1+2i-1+1-2i-1 = 0.
\end{equation*}

Now norm  **should**  measure the distance from a point to the origin and
should only be zero for the zero vector. The fix, as you have probably guessed,
is to sum the squares of the  **magnitudes**  of the components of $z$. This
is accomplished by simply conjugating one of the vectors. Namely, we define
the norm of a complex vector via  \index{norm!vector!complex}

\begin{equation}
  \|z\| \equiv \sqrt{\ov z^Tz}. \label{eq:cnorm}
\end{equation}

In the example above this produces

\begin{equation*}
  \sqrt{|1+i|^2+|1-i|^2} = \sqrt 4 = 2.
\end{equation*}

As each real number is the conjugate of itself, this new definition subsumes 
its real counterpart. The double symbol, conjugate transpose, occurs so
often - for both vectors and matrices - that it has been contracted to 
a single symbol. Namely  \index{conjugate transpose}

\begin{equation}
  \boxed{Z^* \equiv {\ov Z}^T, \hskip 0.25in Z\in\maC^{m\times n}.} \label{eq:conjtran}
\end{equation}

The notion of magnitude in (\ref{eq:cnorm}) also gives us a way to 
define limits and hence will
permit us to introduce complex calculus. We say that the sequence of complex
numbers, $\{z_n:n=1,2,\ldots\}$, converges to the complex number $z_0$ and
write

\begin{equation*}
  z_n\to z_0 \hskip 0.25in\hbox{or}\hskip 0.25in z_0 = \lim_{n\to\infty}z_n,
\end{equation*}

when, presented with any $\eps>0$ one can produce an integer $N$ for which
$|z_n-z_0|<\eps$ when $n\ge N$. As an example, we note that $(i/2)^n\to 0$,
for given an $\eps$ we note that if $n>N=\log_2(1/\eps)$ then $|(i/2)^n|<\eps$.
Similarly, the series

\begin{equation*}
  \sum_{n=1}^\infty z_n 
\end{equation*}

is said to converge to the number $Z$ if the sequence of partial sums

\begin{equation*}
  Z_n \equiv \sum_{m=1}^nz_m
\end{equation*}

converge to $Z$. We shall make use of


\begin{proposition}\label{prop:cv11}
If $z_j\to z$ and

\begin{equation*}
  c_n \equiv \frac{1}{n}\sum_{j=1}^n z_j
\end{equation*}

then $c_n\to z$.

\end{proposition}

**Proof:** Given $\eps>0$ as $z_j\to z$ there exists an $N>0$ such
that $|z_j-z|<\eps/2$ when $j\ge N$. Now, for $n>N$ write

\begin{equation*}
  \eqalign{|z-c_n| 
  = \left|z-\frac{1}{n}\sum_{m=1}^nz_m\right| 
  = \left|\frac{1}{n}\sum_{m=1}^n(z-z_m)\right| 
   &\le  \frac{1}{n}\sum_{m=1}^n|z-z_m| \cr
   &= \frac{1}{n}\sum_{m=1}^N|z-z_m|  +
   \frac{1}{n}\sum_{m=N+1}^n|z-z_m| \cr
   &\le \frac{\eps}{2} + \frac{1}{n}\sum_{m=1}^N|z-z_m| \cr}
\end{equation*}

and so $|z-c_n|<\eps$ when $n>N$ and 

\begin{equation*}
  n > \frac{2}{\eps}\sum_{m=1}^N|z-z_m|.
\end{equation*}

**End of Proof.**
