## Statistical Mechanics 

We begin with equilibrium configuration of finite systems via Information, Entropy and Lagrange Multipliers, solve the 1-d Ising problem, and close with continuous distributions for Hamiltonians and Liouville's Theorem.

We seek to quantify the amount of information associated with the occurrence of an event. In symbols, if $p$ is the probability of the event we seek constraints on the  **information**, $I(p)$, its occurrence brings. First off, its seems
natural that an event that is sure to happen, $p=1$, yields no new information, 

\begin{equation*}
  I(1) = 0.
\end{equation*}

Similarly, the occurrence of an impossible event, $p=0$, yields an infinite amount of information (or surprise or novelty),

\begin{equation*}
  I(0) = \infty.
\end{equation*}

Between these extremes it seems reasonable to expect that $I(p)$ smoothly decreases as we increase $p$, i.e.,

\begin{equation*}
  p\mapsto I(p) \quad\hbox{is smoothly monotonically decreasing.}
\end{equation*}

The key structural property determines how it behaves when faced with the occurrence of two such independent events. Being independent, the probability of their joint occurrence is simply $p^2$. It seems reasonable to claim that this carries twice the information of a single occurrence, i.e.,

\begin{equation}
  I(p^2) = 2I(p).  \label{eq:Idouble}
\end{equation}

(a) Deduce from (\ref{eq:Idouble}) that

\begin{equation}
  I(p^n) = nI(p).  \label{eq:Ipower}
\end{equation}

for positive integer $n$. Next use the tautology
$I(p) = I((p^{1/m})^m)$ and (\ref{eq:Ipower}) to deduce that

\begin{equation}
  I(p^{n/m}) = (n/m)I(p)  \label{eq:Irat}
\end{equation}

for all positive integers $m$ and $n$.

 (b) As every $a\in\maR$ is the limit of rationals and $I$ is presumed smooth conclude from (\ref{eq:Irat}) that

\begin{equation}
  I(p^a) = aI(p) \quad\forall\, a>0.  \label{eq:Iapow}
\end{equation}

 (c) Rewrite (\ref{eq:Iapow}) as $I(\exp(a\log p)) = aI(p)$.
Differentiate this with respect to $a$ and set $a=1$ to arrive at

\begin{equation*}
  I'(p)p\log p = I(p).
\end{equation*}

Rearrange this to read

\begin{equation*}
  (\log I(p))' = (\log \log p )'
\end{equation*}

and so conclude that

\begin{equation*}
  I(p) = c\log(p)
\end{equation*}

and the earlier conditions require $c<0$. We take $c=-1$. 

If an event may have $n$ outcomes, each with probability $p_j$
and information $-\log(p_j)$ then its **entropy** is the average information

\begin{equation}
  S(p) \equiv -\sum_{j=1}^n p_j\log(p_j).  \label{eq:EntD}
\end{equation}

As one of these outcomes must occur at each instance it follows that
 
\begin{equation} 
  \sum_{j=1}^n p_j = 1. \label{eq:sumpj}
\end{equation}

Show that the uniform distribution has the greatest entropy over all those $p$
obeying (\ref{eq:sumpj}).

 (a) We will append the constraint, (\ref{eq:sumpj}), to the entropy via
a Lagrange Multiplier, $\la$. Namely

\begin{equation*}
  L(p,\la) = -\sum_{j=1}^n p_j\log(p_j) + \la\left(1-\sum_{j=1}^n p_j\right)
    = \la - \sum_{j=1}^n p_j(\la+\log(p_j)).
\end{equation*}

Show that

\begin{equation*}
  \frac{\p L(p,\la)}{\p p_i} = \la - 1 + \log(p_i).
\end{equation*}

 (b) We expect a maximizer of $S$ to be a critical point of $L$ and so find
$\log(p_i) = 1 - \la$. Conclude that each $p_i=1/n$ and $\la = 1 + \log(n)$ and
$S(p) = \log(n)$.

 (c) Which has greater entropy, the flipping of a coin or the roll of a die?

 (d) Maximize $S$ subject to (\ref{eq:sumpj}) and $p$ on the hyperplane $p^Th=E$. The Lagrangian now takes the form

\begin{equation*}
  \eqalign{ L(p,\la) 
  &= -\sum_{j=1}^n p_j\log(p_j) + \la_1\left(1-\sum_{j=1}^n p_j\right) + \la_2\left(E-\sum_{j=1}^n p_jh_j\right) \cr
   & = \la_1 + E\la_2 - \sum_{j=1}^n p_j(\la_1+\la_2 h_j+\log(p_j)). \cr}
\end{equation*}

Show that

\begin{equation*}
  \frac{\p L(p,\la)}{\p p_i} = \la_1 - 1 + h_i\la_2 + \log(p_i).
\end{equation*}

and so

\begin{equation*}
  p_j = \exp(1 - \la_1 - h_j\la_2)
\end{equation*}

and now the constraints reveal $\la_1$ and $\la_2$.

\begin{equation*}
  1 = \exp(1-\la_1)\sum_{j=1}^n\exp(-\la_2 h_j) \and
 E = \exp(1-\la_1)\sum_{j=1}^n h_j\exp(-\la_2 h_j) = 
   \frac{h^T\exp(-\la_2 h)}{\maOne^T\exp(-\la_2 h)}
\end{equation*}

Show that $\la_2\mapsto h^T\exp(-\la_2 h)/\maOne^T\exp(-\la_2 h)$
is monotone increasing.

Numerator of derivative is

\begin{equation*}
  \eqalign{
  & -\maOne^T\exp(-\la_2 h)(h^2)^T\exp(-\la_2 h) + h^T\exp(-\la_2 h)h^T\exp(-\la_2 h) \cr
  &= (h^T\exp(-\la_2 h))^2 - \maOne^T\exp(-\la_2 h)(h^2)^T\exp(-\la_2 h) \cr}
\end{equation*}

Now use, for $y\ge 0$,

\begin{equation*}
  x^Ty = (x.*\sqrt y)^T\sqrt y \le \|(x.*\sqrt y)\|\|\sqrt y\|.
\end{equation*}

 (f) Show that 

\begin{equation*}
  S(p) = -\exp(1-\la_1)\sum_{j=1}^n\exp(\la_2 h_j)(1 - \la_1 - h_j\la_2)
  = \la_1 - 1 + \la_2 E = \la_2 E + \log Z(\la_2).
\end{equation*}


### The Ising Problem 

We consider a ring of $n$ particles with each particle able to take on two possible values, $u_i=\pm 1$.
There are then $2^n$ possible states or outcomes. We denotes a state $u^{(s)}\in\maZ_2^n$.
If $J$ is the coupling constant then the nearest neighbor interaction energy of state $s$ is

\begin{equation*}
  h_s \equiv E(u^{(s)}) = -J\sum_{i=1}^n u_i^{(s)}u_{i+1}^{(s)}, \hskip 0.25in u_{n+1}^{(s)}=u_1^{(s)}.
\end{equation*}

We denote the probability of state $s$ by $p_s$ and maximize the entropy

\begin{equation*}
  -\sum_{s=1}^{2^n} p_s\log(p_s) \quad\hbox{subject to}\hskip 0.25in \sum_{s=1}^{2^n} p_s h_s = {\ov E}
  \and \sum_{s=1}^{2^n} p_s = 1.
\end{equation*}

We denote the solution by $g$ after Gibbs,

\begin{equation*}
  g_s = \exp(-\la_2 h_s)/Z_{1,n}, % \where Z(\la_2) = \sum_{s=1}^{2^n} \exp(-\la_2 h_s)
\end{equation*}

where $Z_{1,n}$ is the partition function (on the 1-dimensional ring with $n$ players)

\begin{equation*}
  Z_{1,n} = \sum_{s=1}^{2^n}\exp(-\beta E(u^{(s)})) 
     = \sum_{s=1}^{2^n}\exp\left(\nu\sum_{i=1}^n u_i^{(s)}u_{i+1}^{(s)}\right)
\end{equation*}

where $\nu=\beta J$. The key to evaluating this is to set

\begin{equation}
  L_1(u_i,u_{i+1}) = \exp(\nu u_iu_{i+1})   \label{eq:L1i}
\end{equation}

and note that $Z_{1,n}$ reads

\begin{equation}
  Z_{1,n} = \sum_{s=1}^{2^n}L_1(u_1^{(s)},u_2^{(s)})L_1(u_2^{(s)},u_3^{(s)})\cdots L_1(u_n^{(s)},u_1^{(s)}).  \label{eq:Z1n0}
\end{equation}

As each $u_j$ can take on but two values we see that there are 4 possible pairs of $(u_i,u_{i+1})$.
If we collect these into the 2-by-2 matrix

\begin{equation}
   L_1 = \begin{bmatrix} L_1(1,1) & L_1(1,-1) \\ L_1(1,-1) & L_1(-1,-1) \end{bmatrix} \label{eq:L1mat}
\end{equation}

and page back to (\ref{eq:trAn} we recognize in \ref{eq:Z1n0}) the trace of the $n$th power of $L_1$,

\begin{equation}
  Z_{1,n} = \tr(L_1^n). \label{eq:Z1n00}
\end{equation}

On substituting (\ref{eq:L1i} in \ref{eq:L1mat}) we find

\begin{equation*}
   L_1 = \begin{bmatrix} \re^\nu & \re^{-\nu} \\ \re^{-\nu} & \re^\nu \end{bmatrix}
  = \re^\nu I + \re^{-\nu}\tau_1, \where 
    \tau_1 = \begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}.
\end{equation*}

Now using $\tau_1^2=I$ we conclude that
$\exp(t\tau_1) = \cosh(t)I + \sinh(t)\tau_1 = \cosh(t)(I+\tanh(t)\tau_1)$. Hence

\begin{equation*}
  L_1 = \re^\nu(I+\re^{-2\nu}\tau_1) = \re^\nu\exp(t\tau_1)/\cosh(t)
\end{equation*}

when

\begin{equation*}
  \tanh(t) = \re^{-2\nu} \quad\hbox{or, equivalently}\hskip 0.25in \sinh(2t)\sinh(2\nu) = 1.
\end{equation*}

In this case

\begin{equation*}
  L_1 = (2\sinh(2\nu))^{1/2}\exp(t\tau_1)
\end{equation*}

and so

\begin{equation*}
   L_1^n = (2\sinh(2\nu))^{n/2}\exp(nt\tau_1)
\end{equation*}

and (using that $\pm 1$ are the eigenvalues of $\tau_1$)

\begin{equation*}
  \tr(L_1^n) = (2\sinh(2\nu))^{n/2}(\re^{nt}+\re^{-nt}) 
\end{equation*}

and so the free energy per spin in the thermodynamic limit

\begin{equation*}
     \lim_{n\to\infty}\frac{1}{n}\log(Z_{1,n}) = (1/2)\log(2\sinh(2\nu)) + t
\end{equation*}

and yet

\begin{equation*}
  \log(2\sinh(2\nu)) = \log(4\sinh(\nu)\cosh(\nu)) = \log(2\sinh(\nu)) + \log(2\cosh(\nu))
\end{equation*}

and

\begin{equation*}
  \eqalign{
  t &= \tanh^{-1}(\re^{-2\nu}) \cr
     &= \frac{1}{2}\log\left(\frac{1+\re^{-2\nu}}{1-\re^{-2\nu}}\right) \cr
     &= \frac{1}{2}\log\left(\frac{2\cosh(\nu)}{2\sinh(\nu)}\right) \cr
    &= (1/2)\log(2\cosh(\nu)) - (1/2)\log(2\sinh(\nu)) \cr}
\end{equation*}

hence

\begin{equation*}
  \boxed{\lim_{n\to\infty}\frac{1}{n}\log(Z_{1,n}) = \log(2\cosh(\nu)).}
\end{equation*}

### Liouville's Theorem

With regard to the $n$-dimensional Hamiltonian system

\begin{equation*}
  x'=\nabla_p H(p,x) \hskip 0.25in p'=-\nabla_x H(p,x)
\end{equation*}

let $\Omega$ be a subset of configuration space, $\maR^{2n}$, and define 

\begin{equation*}
  P_\Omega(t) \equiv \int_\Omega \rho(p,x,t)\,dp\,dx
\end{equation*}

to be the probability that the system lies in $\Omega$ at time $t$. In this case, the associated 
probability density function, $\rho$, obeys an associated transport equation.


\begin{proposition}\label{prop:Liouville} **Liouville's Theorem**. The density $\rho$ obeys

\begin{equation}
  \frac{\p\rho}{\p t} = \nabla_x H\cdot\nabla_p \rho - \nabla_p H\cdot\nabla_x \rho. \label{eq:LioThm}
\end{equation}

\end{proposition}

**Proof:** Let $\Omega$ denote a nice chunk of state-space, $\maR^{2n}$, with boundary $\p\Omega$.
Now, the rate of change states in $\Omega$ is the rate at which states cross $\p\Omega$, i.e.,

\begin{equation*}
  \frac{d}{dt}\int_\Omega \rho(p,q,t)\,dp\,dq =
  \int_{\p\Omega} \rho v\cdot n \, ds
\end{equation*}

where $v=(p',q')=(-\nabla_q H, \nabla_p H)$ is the state-space velocity 
Now the divergence Theorem permits

\begin{equation*}
  \int \p_t\rho(p,q,t)\,dp\,dq = \int_\Omega \nabla\cdot(\rho v)\,dp\,dq
\end{equation*}

It follows that

\begin{equation*}
  -\p_t\rho = \nabla\cdot(\rho v) = -\nabla_p\rho\cdot \nabla_q H + \nabla_q\rho\cdot \nabla_p H
  + (\nabla_p\cdot\nabla_x H-\nabla_x\cdot\nabla_p H)\rho.
\end{equation*}

As this latter term vanishes (as we have simply changed the order of partial differentiation) we arrive at
(\ref{eq:LioThm}). **End of Proof.**

  **Example:** We return to the harmonic oscillator. It's physics is captured in the $B$ matrix and
Hamiltonian then dictates that its associated density obeys

\begin{equation}
  \frac{\p\rho}{\p t} = kx\frac{\p\rho}{\p p} - \frac{p}{m}\frac{\p\rho}{\p x}, \where \rho(x,p,0) = \rho_0(x,p), 
      \label{eq:LioCHO} % = \exp(-(q-1)^2/2 - p^2/2)
\end{equation}

is its initial density. As the dynamics are governed by the matrix exponential, $\exp(Bt)$, as in (\ref{eq:BonemassExp}), we see
that to reach state $(x,p)$ at time $t$ it must have begun at $\rho_0((\exp(Bt))^{-1}(x,p/m)^T)$. More precisely, on
establishing that

\begin{equation*}
  (\exp(Bt))^{-1} = \begin{bmatrix} \cos(\om t) & -\sin(\om t)/\om \\ 
      \om\sin(\om t) & \cos(\om t) \end{bmatrix}, \hskip 0.25in \om^2 = k/m,
\end{equation*}

we find the density

\begin{equation}

                  &= \rho_0(x\cos(\om t)-(p/(m\om))\sin(\om t), x\om\sin(\om t)+(p/m)\cos(\om t)) \cr} \label{eq:rhoharm}
\end{equation}

## Quantum Mechanics 

In the Rutherford Hydrogen atom the electron radiates energy and its orbit shrinks, contrary to experiment. 
When a high-voltage electrical discharge is passed through a sample of hydrogen gas (H2) at low pressure, the result is individual isolated hydrogen atoms that emit a red light. When the emitted light is passed through a slit and a prism, only a images of the slit of particular wavelengths, called a line spectrum, are observed rather than a continuous range of wavelengths, Figure \ref{fig:Balmer}The light emitted by hydrogen atoms is red because, of its four characteristic lines, the most intense line in its spectrum is in the red portion of the visible spectrum, at 656 nm.

His student, Neils Bohr, observing the Balmer series, Figure \ref{fig:Balmer},
postulated that

> "An atom can exist in only certain discrete stationary states in which it does not radiate. Light will be emitted on transition from one state into another; the energy which it loses in this transition, the difference $E_1-E_2$ of its energy in the two states, will be transformed into a photon of energy $h\nu$, where the frequency is determined by $h\nu = E_1-E_2$. In this equation $E_1$ and $E_2$ may be any two discrete energy levels, so long as $E_1>E_2$. Conversely, in absorption a photon raises the 
atom from the energy level $E_1$ to a higher $E_2$ by giving up its energy $h\nu$ to the atom."


The right fix to the theory took some time to come. The fix is radical, yet simple
to execute, through a recipe known as  **canonical quantization** . Rather than
track the particle's exact position, $x(t)$, and momentum, $p(t)$, we track their density through
a complex function, $\Psi$, of space, $x$, and time, $t$. To be exact, given the classical system with Hamiltonian

\begin{equation}
  H(p,x) = \frac{p^Tp}{2m} + V(x)     \label{eq:hamo}
\end{equation}

replace each $p_j$ with $i\hslash\p_{x_j}$,  where $\hslash\equiv h/(2\pi)$ and $h=6.62607015\times 10^{-34}\,Js$ is Planck's constant, so that 

\begin{equation}
  p^Tp = p_1^2+p_2^2+p_3^2 \quad\hbox{becomes}\hskip 0.25in -\hslash^2(\p_{x_1}^2+\p_{x_2}^2+\p_{x_3}^2) 
    = -\hslash^2 \Delta,    \label{eq:CQp}
\end{equation}

where $\Delta$ is the **Laplacian** in the three spatial variables. 
This operator acts on the putative density function, $\Psi$, through differentiation, while the
potential $V(x)$ acts on $\Psi$ through multiplication. Again, to be exact, one replaces (\ref{eq:hamo}) with

\begin{equation}
  \cH(x)\Psi(x,t) = \frac{-\hslash^2}{2m} \Delta\Psi(x,t) + V(x)\Psi(x,t).  \label{eq:cHamPsi}
\end{equation}

Finally, the dynamics of the classical Hamiltonian system

\begin{equation}
   H(p(t),x(t)) = E, \label{eq:HpxE}
\end{equation}

in terms of the energy $E$ is quantized by noting that space-time coordinates, $(x,t)$, give rise to momentum-energy coordinates, $(p,E)$.
As such we replace $E$, in a manner consistent with (\ref{eq:CQp}), with the scaled time derivative $i\hslash\p_t$ applied to $\Psi$. This 
substitution in (\ref{eq:cHamPsi} and \ref{eq:HpxE}) brings the **Schr&ouml;dinger Equation**

\begin{equation}
 \boxed{ \frac{-\hslash^2}{2m} \Delta\Psi(x,t) + V(x)\Psi(x,t) = i\hslash\p_t\Psi(x,t),} \label{eq:schroeq}
\end{equation}

for the  **wave function**, $\Psi$. We view
(\ref{eq:schroeq}) as an evolution equation given a known initial wave function 

\begin{equation*}
  \Psi(x,0) = \Psi_0(x).
\end{equation*}

To do justice to the statistical analogy underlying this quantization we must make clear the sense in which $\Psi$ acts
as a probability density for position and momentum. As $\Psi$ is complex valued we will write it in terms of its real
magnitude and phase

\begin{equation}
  \Psi(x,t) = |\Psi(x,t)|\exp(i\theta(x,t)).  \label{eq:PsiPolar}
\end{equation}

We will first show that $x\mapsto|\Psi(x,t)|^2$ behaves like a probability distribution for position.
We show that the integral over all positions does not vary with time. To do this in an orderly
fashion we begin by integrating over the disk, $\|x\|<R$,
and eventually let $R\to\infty$. 
We show that

\begin{equation*}
  A(t) \equiv \int_{\|x\|<R} |\Psi(x,t)|^2\,dx = \int_{\|x\|<R} \Psi(x,t)\Psi^*(x,t)\,dx
\end{equation*}

is constant. To see that we'll need that $\Psi^*(x,t)$ obeys

\begin{equation}
 -\frac{\hslash^2}{2m}\Delta\Psi^*(x,t) + V(x)\Psi^*(x,t)
  = -i\hslash\p_t \Psi^*(x,t) \label{eq:ccschroeq}
\end{equation}

when $V$ is real. Now

\begin{equation*}
  \eqalign{A'(t) &= \int_{\|x\|<R} (\Psi_t(x,t)\Psi^*(x,t)+\Psi(x,t)\Psi^*_t(x,t))\,dx \cr
  &= \frac{1}{i\hslash}\int_{\|x\|<R} (V(x)\Psi(x,t)-\hslash^2\Delta\Psi(x,t)/(2m))\Psi^*(x,t) -
   \Psi(x,t)(V(x)\Psi^*(x,t)-\hslash^2\Delta\Psi^*(x,t)/(2m))\,dx \cr
 &= \frac{i\hslash}{2m}\int_{\|x\|<R} (\Psi^*(x,t)\Delta\Psi(x,t) - \Psi(x,t)\Delta\Psi^*(x,t))\,dx \cr
 &= \frac{i\hslash}{2m}\int_{\|x\|<R} \nabla\cdot (\Psi^*(x,t)\nabla\Psi(x,t) - \Psi(x,t)\nabla\Psi^*(x,t))\,dx \cr
 &= \frac{i\hslash}{2m}\int_{\|x\|=R} (\Psi^*(x,t)\nabla\Psi(x,t) - \Psi(x,t)\nabla\Psi^*(x,t))\cdot\frac{x}{R}\,dx \cr}
\end{equation*}

Hence, so long as $\Psi(x,t)\nabla\Psi^*(x,t)$ approaches 0 at large $\|x\|$ we see that

\begin{equation*}
  \int_{\maR^3} |\Psi(x,t)|^2\,dx = \int_{\maR^3} |\Psi_0(x)|^2\,dx 
\end{equation*}

for all time.
It follows that $x\mapsto|\Psi(x,t)|^2$ may be considered a probability density. In that sense, given a set
$\Omega\subset\maR$ we can interpret

\begin{equation*}
  \int_\Omega|\Psi(x,t)|^2\,dx
\end{equation*}

to be the probability that the particle lies in $\Omega$ at time $t$. We may then, per (\ref{eq:normmvar}), go on
to write the average position at time $t$ as

\begin{equation}
  \mean( |\Psi(\cdot,t)|^2) = \int_{\maR^3} x|\Psi(x,t)|^2\,dx  \label{eq:avgposQM}
\end{equation}

In the physics literature this is often abbreviated to $\langle x\rangle$. 

Using the fact that momentum
$p=m\,dx/dt$, we may then expect the average $j$th momentum to be

\begin{equation*}
  \eqalign{\langle p_j\rangle &= m(d/dt)\langle x_j\rangle \cr
  &= m(d/dt)\int_{\maR^3} x_j|\Psi(x,t)|^2\,dx \cr
  &= \frac{i\hslash}{2}\int_{\maR^3} x_j\nabla\cdot (\Psi^*\nabla\Psi - \Psi\nabla\Psi^*)\,dx \cr
  &= \frac{i\hslash}{2}\int_{\maR^3} x_j\p_{x_j} (\Psi^*\p_{x_j}\Psi - \Psi\p_{x_j}\Psi^*)\,dx \cr
  &= \frac{i\hslash}{2}\int_{\maR^3} (\Psi\p_{x_j}\Psi^* - \Psi^*\p_{x_j}\Psi)\,dx \cr
  &= \frac{i\hslash}{2}\int_{\maR^3} 2i\Im(\Psi\p_{x_j}\Psi^*) \,dx \cr
  &= -\hslash\int_{\maR^3} \Im(|\Psi|\exp(i\theta)\exp(-i\theta)(\p_{x_j}|\Psi|-i|\Psi|\p_{x_j}\theta)) \,dx \cr
  &= \hslash\int_{\maR^3}|\Psi|^2\p_{x_j}\theta \,dx \cr}
\end{equation*}

which we interpret as saying that the average of $\hslash \p_{x_j}\theta(\cdot,t)$ is the average of the $j$th momentum at time $t$.


And when, $V=V(x)$, we may separate variables, namely

\begin{equation}
  \Psi(x,t) = \psi(x)\eta(t)   \label{eq:SepVarPsi}
\end{equation}

and find

\begin{equation*}
 -\frac{\hslash^2}{2m}\Delta\psi(x)\eta(t) + V(x)\psi(x)\eta(t)
  = i\hslash\psi(x)\eta'(t) 
\end{equation*}

where prime denotes differentiation with respect to the underlying independent variable. On dividing this last equation by $\psi(x)\eta(t)$ we arrive at

\begin{equation*}
 -\frac{\hslash^2}{2m}\Delta\psi(x) + V(x) = i\hslash\eta'(t)/\eta(t)
\end{equation*}

which is remarkable in that it equates a function of space with a function of time. This can occur only when both sides are equal to a constant, let us call it, $E$. In this case we find that $\psi$ must obey

\begin{equation}
 -\frac{\hslash^2}{2m}\Delta\psi(x) + V(x)\psi(x) = E\psi(x)   \label{eq:SchroSoV1}
\end{equation}

while $\eta$ must obey

\begin{equation}
  i\hslash\eta'(t) = E\eta(t) \label{eq:SchroSoV2}
\end{equation}

The has the explicit  solution

\begin{equation*}
  \eta(t) = c\exp(-iEt/\hslash)
\end{equation*}

while (\ref{eq:SchroSoV1} is the source of all the action. In particular, (\ref{eq:SchroSoV1}), as an eigenvalue problem with an infinite number of
eigenvalues, $E_n$, and eigenfunctions, $\psi_n$, where $n$ steps through the counting numbers. As such, we must also index $\eta$ by
$n$ and replace our naive guess (\ref{eq:SepVarPsi}) with the richer

\begin{equation}
  \boxed{\Psi(x,t) = \sum_{n=0}^\infty c_n\exp(-iE_nt/\hslash)\psi_n(x),}   \label{eq:SchroSolEx}
\end{equation}

where

\begin{equation}
  c_n = \int_{\maR^3}\Psi_0(x)\psi_n(x)\,dx.  \label{eq:cnInit}
\end{equation}

To make sense of these we will assemble all pieces in the two standard contexts.

### The Quantum Harmonic Oscillator



The Schr&ouml;dinger eigenvalue problem 
for the one-dimensional quantum harmonic oscillator, where $V(x) = kx^2/2=m\om^2x^2/2$ is 

\begin{equation}
  -\frac{\hslash^2}{2m}\psi''(x) + \frac{m\om^2}{2}x^2\psi(x) = E\psi(x).  \label{eq:qharmosc}
\end{equation}

We now perform a number of simple substitutions to turn this into a previously solved problem.
We first multiply through by $2m/\hslash^2$ and obtain 

\begin{equation*}
 -\psi''(x) +a^2x^2\psi(x) = \la\psi(x), \where a=m\om/\hslash \and \la = 2mE/\hslash^2.
\end{equation*}

We then note for large $x$ that $\psi(x)=\exp(\pm a x^2/2)$ balances the left hand side. Choosing the decaying exponential
we suppose $\psi(x)=\exp(-ax^2/2)\phi(x)$ and find that $\phi$ must obey

\begin{equation*}
  -\phi''(x) + 2ax\phi'(x) + a\phi(x) = \la\phi(x). 
\end{equation*}

On changing to the non-dimensional variable, $y=\sqrt a x$, we find that $H(y)\equiv\phi(x)$  obeys

\begin{equation*}
  H''(y) - 2yH'(y) + (\la/a - 1)H(y) = 0.
\end{equation*}

We recognize this as (\ref{eq:HermODE}), the Hermite differential equation, with solutions

\begin{equation*}
  H(y) = H_n(y) \and \la=\la_n = (2n+1)a, \hskip 0.25in n=0,1,2,\ldots
\end{equation*}

Hence, remaining with $y$ for the moment

\begin{equation}
  \psi_n(y) = (2^n n! \sqrt\pi)^{-1/2}\exp(-y^2/2)H_n(y), \hskip 0.25in E_n = \hslash\om(n+1/2).  \label{eq:QHOEfunc}
\end{equation}

We have graphed the first 4 of these eigenfunction in Figure \ref{fig:Qharm0}(A).
We note for small $dx$ that $2dx\psi_n^2(x)$ is the probability of finding a particle, with energy $E_n$, within $dx$ of $x$. We now ask
how this prediction compares with the known amplitude density, computed in Exer.~\ref{CHOdensity}, 

\begin{equation}
  \rho_A(x) = \frac{1}{A_0}\frac{1}{\sqrt{1-(x/A_0)^2}},  \label{eq:rhoCHO}
\end{equation}

of the classical harmonic operator. Here $A_0$ is the amplitude of the classical oscillation, see (\ref{eq:AmpPhs}), and $x$ takes values between
$-A_0$ and $A_0$. To reconcile the energy of the quantum oscillator with the amplitude of the classical oscillator we learn from (\ref{eq:AHrelCHO})
that $2E_n=kA_0^2$, i.e.,

\begin{equation}
  2\hslash \om (n+1/2) = kA_0^2.  \label{eq:QCrecon0}
\end{equation}

Finally, to remain in the non-dimensional setting we suppose $a=1$, i.e., $m\om = \hslash$. On substituting this into (\ref{eq:QCrecon0}) we find

\begin{equation}
  A_0 = \sqrt{2n+1}.  \label{eq:QCrecon1}
\end{equation}

In the case that $n=0$ we find $A_0=1$ and illustrate the resulting $\rho$ and $\psi_0^2$ in Figure \ref{fig:Qharm0}(B).

### The Hydrogen Atom

We consider an electron "orbiting" a fixed proton under the influence of the Coulomb potential

\begin{equation*}
  V(x) = -e^2/\|x\|.
\end{equation*}

The Schr&ouml;dinger eigenvalue problem, (\ref{eq:SchroSoV1}), now takes the form

\begin{equation}
 -\frac{\hslash^2}{2\mu}\Delta\psi(x,t) -\frac{e^2}{\|x\|}\psi(x,t) = E\psi(x)   \label{eq:HyEig0}
\end{equation}

To begin, we note that
as the potential $V(x)$ depends only on distance we may have an easier time working in spherical
coordinates, $(r,\theta,\phi)$, where

\begin{equation}
 \eqalign{0<r<\infty, \cr
         0<\theta<\pi, \cr
         0<\phi<2\pi,\cr} \and
 \eqalign{x_1 &= r\sin(\theta)\cos(\phi), \cr
               x_2 &= r\sin(\theta)\sin(\phi), \cr
               x_3 &= r\cos(\theta), \cr}  \label{eq:SphCoord}
\end{equation}

In these new variables, our eigenvalue problem becomes

\begin{equation}
  -\frac{\hslash^2}{2\mu}\left(\frac{1}{r^2}\frac{\p}{\p r}\left(\frac{r^2\p\psi}{\p r}\right)
  + \frac{1}{r^2\sin^2(\theta)}\frac{\p^2\psi}{\p\phi^2} + \frac{1}{r^2\sin(\theta)}\frac{\p}{\p\theta}\left(\frac{\sin(\theta)\p\psi}{\p \theta}\right)\right)
 - (e^2/r)\psi = E\psi.   \label{eq:SchroSoV1Hp}
\end{equation}

We also expect our eigenfunction to ignore the artificial boundary established by the new coordinates. That is, we expect that

\begin{equation}
  \psi(r,\theta,0) = \psi(r,\theta,2\pi) \and \p_\phi\psi(r,\theta,0) = \p_\phi\psi(r,\theta,2\pi).   \label{eq:psibc}
\end{equation}

We now attempt to separate these spatial variables via

\begin{equation}
  \psi(r,\theta,\phi) = R(r)\Theta(\theta)\Phi(\phi).  \label{eq:psiby3}
\end{equation}

On substitution into (\ref{eq:SchroSoV1Hp}) and subsequent division by $\psi$ and multiplication by $r^2$, we find

\begin{equation}
  -\frac{\hslash^2}{2\mu}\left( \frac{(r^2R'(r))'}{R(r)}
  + \frac{\Phi''(\phi)}{\sin^2(\theta)\Phi(\phi)} + \frac{(\sin(\theta)\Theta'(\theta))'}{\sin(\theta)\Theta(\theta)}\right)
 - e^2r = Er^2   \label{eq:SchroSoV1Hp2}
\end{equation}

On differentiating with respect to $\phi$ we learn that $\Phi''/\Phi$ must be constant, so $\Phi''=\alpha\Phi$ so

\begin{equation}
  \Phi(\phi) = c_1\exp(\sqrt\alpha\phi) + c_2\exp(-\sqrt\alpha\phi).  \label{eq:phiGuess}
\end{equation}

Now (\ref{eq:psibc}) requires that $\Phi$ also obey

\begin{equation}
  \Phi(0)=\Phi(2\pi) \and \Phi'(0)=\Phi'(2\pi).  \label{eq:phibc}
\end{equation}

On inserting (\ref{eq:phiGuess}) into (\ref{eq:phibc}) we find $\exp(2\pi\sqrt\alpha)=1$ and so $\sqrt\alpha=\pm im$ where
$m\in\maZ$ is referred to as the  **magnetic quantum number** . It follows that we have an infinite sequence of functions

\begin{equation}
  \boxed{\Phi_0(\phi) = \frac{1}{\sqrt{2\pi}} \and \Phi_m(\phi) = \frac{1}{\sqrt\pi}\begin{cases} \cos(m\phi), & \hbox{if}\  m=1,2,3,\ldots \\
                                \sin(m\phi), & \hbox{if}\ m=-1,-2,-3,\ldots \end{cases}}  \label{eq:HphiExact}
\end{equation}

where the scaling has been chosen so as to guarantee the normalization

\begin{equation}
  \int_0^{2\pi}\Phi_m^2(\phi)\,d\phi = 1.   \label{eq:HphiNorm}
\end{equation}

On inserting $\Phi''(\phi)=-m^2\Phi(\phi)$ into (\ref{eq:SchroSoV1Hp2}) we find

\begin{equation}
  \frac{2\mu}{\hslash^2}(e^2r+Er^2) + \frac{(r^2R'(r))'}{R(r)}
  + \frac{(\sin(\theta)\Theta'(\theta))'}{\sin(\theta)\Theta(\theta)} - \frac{m^2}{\sin^2(\theta)}
 = 0 \label{eq:SchroSoV1Hp3}
\end{equation}

On setting the expression in $r$ to $\beta$ and the expression in $\theta$ to $-\beta$ we achieve the
separate equations

\begin{equation}
  (r^2R')' + (2\mu/\hslash^2)(e^2r+Er^2)R = \beta R \label{eq:HyR}
\end{equation}

and

\begin{equation}
  -(\sin(\theta)\Theta')' + m^2\Theta/\sin(\theta) = \sin(\theta)\beta\Theta.  \label{eq:HyTheta}
\end{equation}

To solve (\ref{eq:HyTheta}) set

\begin{equation*}
  z=\cos(\theta) \and P(z)=\Theta(\theta)
\end{equation*}

and arrive at

\begin{equation*}
  ((1-z^2)P'(z))' + \left(\beta-\frac{m^2}{1-z^2}\right)P(z) = 0.
\end{equation*}

This we solved in our discussion of orthogonal polynomials. In particular, recalling (\ref{eq:asslegode}),

\begin{equation*}
  \beta = \ell(\ell+1) \and P(z) = P_{\ell,m}(z)
\end{equation*}

the associated Legendre polynomial. As these are obtained by taking $m$ derivatives of $P_\ell$ we must presume that $\ell$,
known as the  **azimuthal quantum number** , is
restricted to the integers

\begin{equation*}
  \ell = |m|, |m|+1, \ldots
\end{equation*}

Hence

\begin{equation}
  \boxed{\Theta_{\ell,m}(\theta) = \left(\frac{(2\ell+1)(\ell-|m|)!}{2(\ell+|m|)!}\right)^{1/2}P_{\ell,m}(\cos\theta).} \label{eq:HThetaExact}
\end{equation}

Our radial equation, (\ref{eq:HyR}) now takes the form

\begin{equation}
  (r^2R')' + (2\mu/\hslash^2)(e^2r+Er^2)R = \ell(\ell+1) R \label{eq:HyR2}
\end{equation}

To solve this we suppose $E<0$ and introduce the new constants

\begin{equation}
  \alpha^2 = -2\mu E/\hslash^2 \and \la = \frac{\mu  e^2}{\hslash^2\alpha}  \label{eq:Salla}
\end{equation}

and new variables,

\begin{equation*}
   \rho = 2\alpha r \and S(\rho) = R(r)
\end{equation*}

and find that $S$ obeys

\begin{equation}
  (\rho^2 S')' + (\la\rho - \ell(\ell+1)-\rho^2/4)S = 0. \label{eq:SoRode}
\end{equation}

For large $\rho$ this equation states that $4S'' = S$ and so $S$ behaves like $\exp(\pm \rho/2)$. As a result we look
for $S$ of the form $\exp(-\rho/2)F(\rho)$. On substitution into (\ref{eq:SoRode}) we find that $F$ must obey

\begin{equation}
  F'' +\left(\frac{2}{\rho}-1\right)F' + \left(\frac{\la}{\rho}-\frac{1}{\rho} - \frac{\ell(\ell+1)}{\rho^2}\right)F = 0. \label{eq:FoSode}
\end{equation}

This we clean by supposing $L(\rho)=\rho^\ell F(\rho)$ and find that $L$ must obey

\begin{equation}
    \rho L'' + (2(\ell+1)-\rho)L' + (\la-\ell-1)L = 0. \label{eq:LoFode}
\end{equation}

We obtained this differential equation in (\ref{eq:AssLaguerreOde} along with its solution, the Associated Laguerre Polynomial $L_{k,s})$
where $k=\la-\ell-1$ and $s=2\ell+1$. The first of these states that $\la=k+\ell+1$ is an integer. This is customarily denoted by $n$, and
is referred to as the  **total quantum numbe**  while $k=n-\ell-1$ is called the **radial quantum number** . As $k\ge 0$ we find
that $n\ge \ell+1$. On retracing our steps from (\ref{eq:HyR2}) we find that 

\begin{equation}
  R(r)=S(\rho)=\exp(-\rho/2)F(\rho)=\exp(-\rho/2)\rho^\ell L_{n-\ell-1}^{2\ell+1}(\rho). \label{eq:RnlPre}
\end{equation}

and

\begin{equation*}
  \rho = 2\alpha r \where \alpha = \frac{\mu e^2}{\hslash^2 n} \equiv \frac{1}{an}
\end{equation*}

where (using $\mu=m_e=9.1\times 10^{-31}\ kg$)

\begin{equation}  \label{eq:Brad}
  a = \frac{\hslash^2}{\mu e^2} = 5.2984\times 10^{-11}\ m = 0.53117\ \unicode{x212B},
\end{equation} 
 
is the **Bohr radius** in angstroms. Upon normalization (\ref{eq:RnlPre}) becomes

\begin{equation*}
  \boxed{R_{n,\ell}(r) = \left( \left(\frac{2}{an}\right)^2\frac{(n-\ell-1)!}{2n(n+\ell)!}\right)^{1/2}\frac{1}{(n+\ell)!}
      \exp(-r/(an))(2r/(an))^\ell L_{n-\ell-1}^{2\ell+1}(2r/(an))}.
\end{equation*}

It follows that the full eigenfunction is the triply indexed

\begin{equation}
  \psi_{n,\ell,m}(r,\theta,\phi) = R_{n,\ell}(r)\Theta_{\ell,m}(\theta)\Phi_m(\phi)   \label{eq:psinFull}
\end{equation}

while its associated energy level $E$, from (\ref{eq:Salla}), is

\begin{equation}
  E = E_n = -\frac{\hslash^2}{2\mu a^2n^2} = - \frac{E_0}{n^2} \where E_0 = 13.6\ eV.   \label{eq:HyEval}
\end{equation}

(Recall $1\ eV = 1.602\times 10^{-19}\ J$).
As we work back through our indices we find

\begin{equation}
  n=\ell+1,\ell+2,\ldots, \hskip 0.25in \ell = |m|, |m|+1, \ldots, \and m=0,\pm 1, \pm 2, \ldots.  \label{eq:nellm0}
\end{equation}

As $n$ is all we need to index the energy levels rather than building it with $\ell$ and $m$ we instead use it to constrain $\ell$ and $m$.
More precisely, we may rewrite (\ref{eq:nellm0}) as

\begin{equation}
  n=1,2\ldots, \hskip 0.25in \ell=0,1,\ldots,n-1, \and m = 0,\pm 1,\ldots,\pm\ell. \label{eq:nellm1}
\end{equation}

This gives a reasonable way to arrange our eigenfunctions with increasing eigenvalues, e.g., the first fourteen states are

\begin{equation}
  \eqalign{E_1, & \hskip 0.25in \psi_{1,0,0} \cr
      E_2, & \hskip 0.25in \psi_{2,0,0},\  \psi_{2,1,-1},\  \psi_{2,1,0},\  \psi_{2,1,1} \cr
      E_3, & \hskip 0.25in \psi_{3,0,0},\  \psi_{3,1,-1},\  \psi_{3,1,0},\  \psi_{3,1,1}, \  \psi_{3,2,-2},\  \psi_{3,2,-1},\  \psi_{3,2,0},\  \psi_{3,2,1},\  \psi_{3,2,2}\cr} \label{eq:1st14Efun}
\end{equation}

With this ordering we can next embark on their visualization. As the square of the eigenfunction at $(r,\theta,\phi)$ is the probability of 
finding the electron at that point we plot in Figure \ref{fig:HyEfun2} the isosurfaces

\begin{equation}
  \cS_{n,\ell,m}(c) = \{(r,\theta,\phi): \psi^2_{n,\ell,m}(r,\theta,\phi) = c\},   \label{eq:HyIsoS}
\end{equation}

at a representative $c$ for the $(n,\ell,m)$ triples in (\ref{eq:1st14Efun}).

```python
# Hydrogen Orbitals, planar sections

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp
import math

a0 = 1
n = 4
ell = 3
m = 0

# setting the grid
border = 32
accuracy = 100;
raster = np.linspace(-border, border, accuracy)
[x, y, z] = np.meshgrid(raster, raster, raster)

# conversion Cartesian to spherical coordinates
r = np.sqrt(x**2 + y**2  + z**2)
theta = np.arccos(z / r)
phi = np.arctan2(y, x)

xy2     = x**2 + y**2
r       = np.sqrt(xy2 + z**2)    
#theta   = np.arctan2(np.sqrt(xy2), z) # the polar angle in radian angles
phi     = np.arctan2(y, x)            # the azimuth angle in radian angles
phi[phi < 0] += np.pi * 2 

# build psi
                
Lag = sp.genlaguerre(n-ell-1, 2*ell+1) 
arg = 2*r/(n*a0)
sc = np.sqrt((2/n/a0)**3*math.factorial(n-ell-1)/(2*n)/math.factorial(n+ell))
R = sc*Lag(arg)*(arg**ell)*np.exp(-arg/2)
             
Leg = sp.lpmv(m, ell, np.cos(theta))
sc = (-1)**m*np.sqrt((2*ell+1)*math.factorial(ell-m)/(4*np.pi*math.factorial(ell+m)))
flm = sc*Leg

#gm = np.exp(1j*m*phi) 
gm = np.cos(m*phi) 
           
psi = R*flm*gm
psi2 = abs(psi)

fig, axs = plt.subplots(1,10)    # create a figure with 2 axes (plot windows)
plt.subplots_adjust(right=2, top=1, wspace=0.25)  # adjust them for the right look

for i in range(10):
    axs[i].imshow(psi2[:,10*i + 5,:])
    #axs[i].imshow(flm[:,:,10*i + 5])
    axs[i].axis('off')
```

```python
# Hydrogen Orbitals, 3d cloud

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp

a0 = 1
n = 4
ell = 3

# setting the grid
border = 42
accuracy = 200;
raster = np.linspace(-border, border, accuracy)
[x, y, z] = np.meshgrid((raster-border)/2, raster, raster)

# conversion Cartesian to spherical coordinates
r = np.sqrt(x**2 + y**2  + z**2)
theta = np.arccos(z / r)

# build psi
                
Lag = sp.genlaguerre(n-ell-1, 2*ell+1) 
arg = 2*r/(n*a0)
R = Lag(arg)*(arg**ell)*np.exp(-arg/2)
             
Y = sp.lpmv(0, ell, np.cos(theta))  # m=0 here
           
psi = R*Y
psi2 = abs(psi)**2
Mp = np.max(psi2)
psi2 = psi2/Mp

fig = plt.figure(' ')
ax = plt.axes(projection="3d")    # set up for 3D work

[I,J,K] = np.where(psi2 > 0.005) 
ax.scatter(x[I,J,K], y[I,J,K], z[I,J,K], s=1, c = psi2[I,J,K], cmap = 'jet', alpha = 0.1)
Zlim = ax.get_zlim()
ax.set_xlim(Zlim)
ax.set_ylim(Zlim)
ax.view_init(elev=30, azim=45)
ax.axis('off');
```

Though we neglected to plot the isosurface of $\cS_{1,0,0}$ we know that is spherical. This is already a big deal. Not a planar orbit as
in the Rutherford and Bohr models. More importantly however, we return to the probabilistic interpretation and note that

\begin{equation}
  \eqalign{D(R) &\equiv \int_0^R\int_0^\pi\int_0^{2\pi}\psi_{1,0,0}^2(r,\theta,\phi) r^2\sin(\theta)\,drd\theta d\phi \cr
                      &= 1 - (1 + (2R/a) + (2R/a)^2)\exp(-2R/a) \cr}  \label{eq:DGndP}
\end{equation}

is the probability that the electron, in its ground state, lies within $R$ of the nucleus. We graph this in Figure \ref{fig:GndP} and glean that it never gets more that 3 &Aring; from the nucleus and is 73\% sure to stay within 1 &Aring;.
