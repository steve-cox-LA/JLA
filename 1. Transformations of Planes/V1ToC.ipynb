```python
# Volume One Table of Contents

# Source - https://stackoverflow.com/a
# Posted by rAntonioH, modified by community. See post 'Timeline' for change history
# Retrieved 2026-01-05, License - CC BY-SA 3.0

import fnmatch
import os
import re

for section in range(1,10):
    print()
    fname = '1.' + str(section) + '*.ipynb' 
    for file in os.listdir('.'):
        c3 = 0
        c4 = 0
        c5 = 0
        if fnmatch.fnmatch(file, fname):
            lead = '1.' + str(section) + ' ' 
            ind = file[2:].find('.')
            ind2 = file.rfind('.')
            text = file[4:ind2]
            indices = [i for i, char in enumerate(text) if char.isupper()]
            
            if (len(indices) > 1) :
                Text = "" 
                indices.append(len(text))
                for j in range(len(indices)-1):
                    Text = Text + text[indices[j]:indices[j+1]] + " "
            else:
                Text = text
               
            print(lead + Text)
            
            with open(file, 'r', encoding='utf-8') as f:
                for line in f:
                    t3 = "### "
                    m3 = re.search(t3, line) 
                    if (m3):
                        line = line.replace("&ouml;", "o")
                        t4 = "#### "
                        m4 = re.search(t4, line) 
                        if (not m4):
                            c3 = c3 + 1
                            c4 = 0
                            c5 = 0
                            nlead = '  1.' + str(section) + '.' + str(c3) + ' '
                            ind = line.rfind('"')
                            print(nlead + line[9:ind-2]) 
                        else:
                            t5 = "##### "
                            m5 = re.search(t5, line)
                            if (not m5):
                                c4 = c4 + 1
                                c5 = 0
                                nlead = '    1.' + str(section) + '.' + str(c3) + '.' + str(c4) + ' '
                                ind = line.rfind('"')
                                print(nlead + line[10:ind-2])
                            else:
                                c5 = c5 + 1
                                nlead = '      1.' + str(section) + '.' + str(c3) + '.' + str(c4) + '.' + str(c5) + ' '
                                ind = line.rfind('"')
                                print(nlead + line[11:ind-2])
```
