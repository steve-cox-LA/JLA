## The Complex Plane

The lack of fixed points for the Elliptic M&ouml;bius transformations of the previous notebook is a consequence of the simple fact that real quadratic equations need not have real roots. The root cause being a negative radicand.
Fortunately, *all* negative radicands are accommodated by the declaration of the single imaginary unit

\begin{equation}\label{eq:deci}
  \boxed{i \equiv \sqrt{-1}}
\end{equation}

Given two real numbers, $x$ and $y$, we attach $i$ to $y$ and generate the **complex number**

\begin{equation}\label{eq:z}
  z = x + iy
\end{equation}

and call $x = \Re z$ the  **real**  part and $y=\Im z$ the **imaginary**  part of $z$. 
In what follows we explore the rich arithmetic and geometry associated with this move.

### Complex Arithmetic

Regarding arithmetic we name two complex numbers 

\begin{equation*}
  z_1 = x_1 + iy_1 \and z_2 = x_2 + iy_2
\end{equation*}

and **define** their sum and difference to be

\begin{equation}\label{eq:zsd}
  z_1+z_2 \equiv (x_1+x_2)+i(y_1+y_2)  \and z_1-z_2 \equiv (x_1-x_2)+i(y_1-y_2) 
\end{equation}

and their product, via FOIL and $i^2=-1$, to be

\begin{equation}\label{eq:zprod}
  z_1z_2 \equiv (x_1 + iy_1)(x_2 + iy_2) = 
                       (x_1x_2-y_1y_2)+i(x_1y_2+x_2y_1).
\end{equation}

The fact that the product of $z =x+iy$ and its **conjugate**, $\ov z = x-iy$, is

\begin{equation}\label{eq:zpc}
  z\ov z \equiv (x + iy)(x - iy) = x^2 + y^2
\end{equation}

is real, suggests a strategy for expressing the quotient

\begin{equation}\label{eq:zq}
    \frac{z_1}{z_2} = \frac{z_1}{z_2}\frac{\ov z_2}{\ov z_2}
                  = \frac{(x_1x_2+y_1y_2)+i(x_2y_1-x_1y_2)}{x_2^2+y_2^2}
\end{equation}

The fact that (\ref{eq:zpc}) is not only real but positive (not to mention pythagorean) also suggests that we **define** the **magnitude** of $z$ to be

\begin{equation}\label{eq:zmag}
              |z| \equiv \sqrt{z\ov z} = \sqrt{x^2+y^2}.
\end{equation}

In addition to the **Cartesian representation**, $z=x+iy$, one also has the
**polar form**

\begin{equation} \label{eq:polarform}
  z = |z|(\cos\theta+i\sin\theta) \where \theta = \angle z
\end{equation}

is the angle, from $-\pi$ to $\pi$, that $z$ makes with the positive real axis. The magnitude and angle are achieved in python via $\tt abs$ and $\tt np.angle$.

```python
import numpy as np
z = 1 + 1j
print('|1+i| = ', abs(z))
print('angle(1+i) = ', np.angle(z))
z = 1 - 1j
print('|1-i| = ', abs(z))
print('angle(1-i) = ', np.angle(z))
```

Please recognize $\sqrt 2$ and $\pm\pi/4$ in these results, experiment with complex numbers in the other two quadrants, and checkout the illustration of the Cartesian and polar representations in Figure \ref{fig:cplane}(Left).

The complex magnitude is also a remarkably compact way of expressing circles

\begin{equation}\label{eq:zcirc}
  C(z_0,r) = \{z: |z-z_0| = r\}
\end{equation}

and perpendicular bisectors

\begin{equation}\label{eq:zperp}
  L^\perp(w_1,w_2) = \{z: |z-w_1| = |z-w_2|\}.
\end{equation}

To actually draw these curves we rely on their **parametrizations**

\begin{equation}\label{eq:zcirc2}
  C(z_0,r) = \{z_0 + r\cos(t) + ir\sin(t): t\in [0,2\pi) \}
\end{equation}

and

\begin{equation}\label{eq:zperp2}
  L^\perp(w_1,w_2) = \{w + t(\cos(a)+i\sin(a)): t\in\maR\} \where w=(w_1+w_2)/2, \hskip 0.25in a = \arg(w_2-w_1) + \pi/2.
\end{equation}

We offer examples of such curves in Figure \ref{fig:cplane}(Middle).

The polar form is especially convenient with regards to multiplication. More precisely,

\begin{equation*}
  \eqalign{z_1z_2 &= |z_1||z_2|\{(\cos\theta_1\cos\theta_2-\sin\theta_1\sin\theta_2)+
                              i(\cos\theta_1\sin\theta_2+\sin\theta_1\cos\theta_2)\} \cr
                  &= |z_1||z_2|\{\cos(\theta_1+\theta_2)+i\sin(\theta_1+\theta_2)\}. \cr}
\end{equation*}

As a result, for integer values of $n$, we find

\begin{equation}\label{eq:z2n}
  z^n = |z|^n(\cos(n\theta)+i\sin(n\theta)).
\end{equation}

This formula dictates that taking powers of a complex number simultaneously 
scales its magnitude and rotates its argument. The spiral in Figure \ref{fig:cplane}(Right)
offers a concrete illustration.

The polar form, (\ref{eq:polarform}), takes on an even simpler guise once Euler's formula

\begin{equation}\label{eq:EuF}
  \boxed{\exp(i\theta) = \cos(\theta)+i\sin(\theta), \hskip 0.25in \theta\in\maR}
\end{equation}

is established. To do so we, we first recall the Taylor series representation

\begin{equation}\label{eq:ExPow}
  \exp(z) = \sum_{n=0}^\infty \frac{z^n}{n!}
\end{equation}

and note that when $z=i\theta$ we may use Exercise \ref{ex:ipow} to reveal

\begin{equation}\label{eq:ExPow1}
 \eqalign{
  \exp(i\theta) &= \sum_{n=0}^\infty \frac{i^n\theta^n}{n!} \cr
  &= \sum_{n=0}^\infty \frac{i^{2n}\theta^{2n}}{(2n)!} + \frac{i^{2n+1}\theta^{2n+1}}{(2n+1)!} \cr
  &= \sum_{n=0}^\infty \frac{(-1)^n\theta^{2n}}{(2n)!} + \frac{i(-1)^n\theta^{2n+1}}{(2n+1)!} \cr
  &= \cos(\theta) + i\sin(\theta) }
\end{equation}

Euler's formula. With (\ref{eq:EuF}) now comes the compact polar form

\begin{equation}\label{eq:PF2}
  z = |z|\exp(i\theta)
\end{equation}

From now on we will denote the set of complex numbers by $\maC$. With this we generalize (\ref{eq:z2n}) to non-integer powers, say $p$. Proceeding as in the real variable case, where log is the inverse of exp and the log of a product is the sum of the logs we find

\begin{equation}\label{eq:PF2p}
  \eqalign{
  z^p = (|z|\exp(i\theta))^p 
  &= \exp(p\log(|z|\exp(i\theta)) \cr
  &= \exp(p\{\log(|z|)+\log(\exp(i\theta))\})  \cr
  &= \exp(p\log(|z|))\exp(\log(\exp(i\theta))) \cr
  &= |z|^p\exp(ip\theta) \cr}
\end{equation}

Although we will not pause to establish these properties of the complex logarithm we can at least define it;

\begin{equation}\label{eq:clog}
  \log(z) = \log(|z|\exp(i\theta)) \equiv \log(|z|) + i\theta
\end{equation}

where the log on the right is the familiar log of a positive number.

### Lattices and Polygons

Given two points, $\om_1$ and $\om_2$, in the complex plane, not on the same line through the origin, we consider the lattice composed of their integer sums and differences;

\begin{equation}\label{eq:LatD}
  \Omega(\om_1,\om_2) \equiv \{m\om_1+n\om_2: m,n\in\maZ\}
\end{equation}

where $\maZ$ denotes the set of integers (numbers are *Zahlen* in German).
We call $\{\om_1,\om_2\}$ a **basis** for the lattice $\Om(\om_1,\om_2)$.
We depict an example in Figure \ref{fig:Lat} along with a **Fundamental Parallelogram**, $P$. 
This parallelogram is the transformation of the simple square $\{(x_1,x_2):0\le x_k\le 1\}$ by the matrix 

\begin{equation}\label{eq:Lb}
   M = \begin{pmatrix} \Re\om_1 & \Re\om_2 \\ \Im\om_1 & \Im\om_2 \end{pmatrix}
\end{equation}

and so, recalling our early work on Matrix Vector Multiplication, Area$(P)=\det(M) = \Re\om_1\Im\om_2 - \Re\om_2\Im\om_1 = |\om_1|^2\Im(\om_2/\om_1)$. To make this area positive we call $\tau\equiv\om_2/\om_1$ the **modulus** of the basis and suppose

\begin{equation}\label{eq:LatDo}
 \Im(\tau) > 0
\end{equation}

and call such bases **oriented**. 

```python
# Lattice and fundamental parallelogram

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

om1 = 1 + 0.15j   # lattice basis
om2 = 0.2 + 1j

for m in range(-2,3):    # lattice points
    for n in range(-2,3):
        plt.plot(m*om1.real+n*om2.real,m*om1.imag+n*om2.imag,'ko')
        
for m in range(-2,3):    # lattice lines
    plt.plot([m*om1.real-2*om2.real,m*om1.real+2*om2.real], [m*om1.imag-2*om2.imag,m*om1.imag+2*om2.imag],'k',alpha=0.2,linewidth=1)
    plt.plot([m*om2.real-2*om1.real,m*om2.real+2*om1.real], [m*om2.imag-2*om1.imag,m*om2.imag+2*om1.imag],'k',alpha=0.2,linewidth=1)

plt.text(-.1,-.1,'$0$')
plt.text(om1.real+.05, om1.imag-.05, '$\omega_1$')   # lattice lables
plt.text(om2.real-.15, om2.imag+.05, '$\omega_2$')
plt.text((om1+om2).real-.1, (om1+om2).imag+.1, '$\omega_1+\omega_2$')
plt.text(-om1.real+.05, -om1.imag-.05, '$-\omega_1$')   # lattice lables
plt.text(-om2.real-.2, -om2.imag+.05, '$-\omega_2$')
plt.text(-(om1+om2).real-.2, -(om1+om2).imag+.1, '$-\omega_1-\omega_2$')
plt.text((om1-om2).real-.2, (om1-om2).imag+.1, '$\omega_1-\omega_2$')
plt.text((om2-om1).real-.2, (om2-om1).imag+.1, '$\omega_2-\omega_1$')

xP = [0, om1.real, (om1+om2).real, om2.real]   # fill and label parallelogram
yP = [0, om1.imag, (om1+om2).imag, om2.imag]
plt.fill(xP, yP, alpha=0.3, edgecolor='b',linewidth=3)
plt.text(np.mean(xP),np.mean(yP),'$P$')

plt.axis([-3/2,3/2,-3/2,3/2])
plt.xlabel('real part',fontsize=14)
plt.ylabel('imaginary part',fontsize=14);
```

\begin{myfig}\label{fig:Lat} A lattice, $\Omega(\om_1,\om_2)$, and a fundamental parallelogram, $P$. \end{myfig}

We declare two Lattices, $\Om\equiv\Om(\om_1,\om_2)$ and $\hat\Om\equiv\Om(\hat\om_1,\hat\om_2)$, with oriented bases, to be **similar** when, for some $\s\in\maC$ we have $\hat\Om=\{\s\om:\om\in\Om\}$. Recalling (\ref{eq:PF2}) we see that
multiplication of a lattice by $\s=|\s|\exp(i\theta)$ scales the lattice by $|\s|$ and rotates it by $\theta$. We will now show that moduli of similar oriented lattices are related to one another by M&ouml;bius transformations of the form

\begin{equation}\label{eq:sl2z}
  \mu(\tau)=\frac{a\tau + b}{c\tau + d} \where M = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \in \SL_2(\maZ)
\end{equation}

lies in the set of 2-by-2 integer matrices with determinant one.

\begin{proposition}\label{prop:SimLat} 

Two lattices, $\Om(\om_1,\om_2)$ and $\Om(\hat\om_1,\hat\om_2)$, with oriented bases, are similar if and only if their moduli obey $\hat\tau = \mu(\tau)$ for a $\mu$ of the form (\ref{eq:sl2z}).
\end{proposition}

**Proof:** If $\Om(\om_1,\om_2)$ and $\Om(\hat\om_1,\hat\om_2)$ are similar then 

\begin{equation}\label{eq:SimLat0}
  \hat\om_2 = \s a\om_2 + \s b\om_1 \and \hat\om_1 = \s c\om_2 + \s b\om_1
\end{equation}

for some integers $a,b,c,d$ and $\s\in\maC$. Dividing $\hat\om_2$ by $\hat\om_1$ we find

\begin{equation}\label{eq:SimLat1}
  \hat\tau = \frac{\hat\om_2}{\hat\om_1} = \frac{a\om_2 + b\om_1}{c\om_2 + d\om_1} = \frac{a\tau + b}{c\tau + d}
  \equiv \mu(\tau)
  \where \tau = \om_2/\om_1.
\end{equation}

Now, as $a,b,c,d$ are real, a brief calculation reveals $\Im(\hat\tau)=(ad-bc)\Im(\tau)/|c\tau+d|^2$. As each basis is oriented it follows that $ad-bc>0$. We may now invert our starting point and write

\begin{equation}\label{eq:SimLat0h}
  \om_2 = (1/\s) \hat a\hat\om_2 + (1/\s) \hat b\hat\om_1 \and \om_1 = (1/\s) \hat c\hat\om_2 + (1/\s) \hat b\hat\om_1
\end{equation}

for some integers $\hat a,\hat b,\hat c,\hat d$. Dividing $\om_2$ by $\om_1$ we find

\begin{equation}\label{eq:SimLat1h}
  \tau = \frac{\om_2}{\om_1} = \frac{\hat a\hat\om_2 + \hat b\hat\om_1}{\hat c\hat \om_2 + \hat d\hat \om_1} 
  = \frac{\hat a\hat\tau + \hat b}{\hat c\hat\tau + \hat d} \equiv \hat\mu(\hat\tau)
\end{equation}

with $\hat a\hat d - \hat b\hat c > 0$. Combining these we find $\tau=\hat\mu(\hat\tau)=\hat\mu(\mu(\tau)))$ and so $\hat\mu$ is the inverse of $\mu$. It follows that their corresponding matrices obey $\hat M M = I$ and so 
$\det(\hat M)\det(M)=1$. As these are both integer matrices with postive determinants it follows that $\det(M)=\det(\hat M)=1$.


Conversely, if (\ref{eq:SimLat1}) holds with integral $a,b,c,d$ and $ad-bc=1$ then

\begin{equation}\label{eq:SimLat2}
  \frac{\hat\om_2}{\hat\om_1} = \frac{a\om_2 + b\om_1}{c\om_2 + d\om_1} = \frac{a\tau + b}{c\tau + d}
\end{equation}

yields

\begin{equation}\label{eq:SimLat3}
  \hat\om_2 = \frac{\hat\om_1}{c\om_2 + d\om_1}(a\om_2 + b\om_1) \and
  \hat\om_1 = \frac{\hat\om_2}{a\om_2 + b\om_1}(c\om_2 + d\om_1)
\end{equation}

and setting the common value

\begin{equation}\label{eq:SimLat4}
  \frac{\hat\om_1}{c\om_2 + d\om_1} = \frac{\hat\om_2}{a\om_2 + b\om_1}
\end{equation}

to $\s$ we see in (\ref{eq:SimLat3}) the similarity condition (\ref{eq:SimLat0}). **End of proof.**

This parallelogram, $P$, in Figure \ref{fig:Lat}, is **fundamental** in the sense that the **orbit**, $z+\Omega(\om_1,\om_2)$, of every $z\in\maC$ intersects $P$ and no 2 points in the interior of $P$ lie in the same orbit.

Although this $P$ is a natural fundamental polygon for $\Omega$ it is in some sense too natural. A construction of Dirichlet will permit the generalization to the non-Euclidean setting. The **Dirichlet Polygon** of the lattice $\Omega$ is the set of points that are as close to the origin as they are to any other lattice point. That is,

\begin{equation}\label{eq:Dir0}
  D(\Omega) \equiv \{z\in\maC: |z|\le |z-\om|\ \forall\om\in\Omega\} = \bigcap_{\om\in\Omega}\{z\in\maC: |z|\le |z-\om|\}
\end{equation}

These latter sets are half spaces bounded by the lines

\begin{equation}\label{eq:Dir1}
  S(\om) \equiv \{z\in\maC: |z|=|z-\om|\}
\end{equation}

for specific $\om\in\Omega$. On comparison with (\ref{eq:zperp}) we see that $S(\om)$ is the perpendicular bisector of the line segment from $0$ to $\om$. It follows that $D(\Omega)$ is then enclosed by the $S(\om)$ for those lattice points $\om$ near the origin. We illustrate this construction of $D(\Omega)$ in Figure \ref{fig:Dhex}, along with the visual proof that its area is the same as the area of the associated fundamental parallelogram.

```python
# Dirichlet Hexagon

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

om1 = 1 + 0.15j   # lattice basis
om2 = 0.2 + 1j

xhex = np.zeros(6)   # hexagon vertices
yhex = np.zeros(6)

for m in range(-2,3):    # lattice points
    for n in range(-2,3):
        plt.plot(m*om1.real+n*om2.real,m*om1.imag+n*om2.imag,'ko')
        
for m in range(-2,3):    # lattice lines
    plt.plot([m*om1.real-2*om2.real,m*om1.real+2*om2.real], [m*om1.imag-2*om2.imag,m*om1.imag+2*om2.imag],'k',alpha=0.2,linewidth=1)
    plt.plot([m*om2.real-2*om1.real,m*om2.real+2*om1.real], [m*om2.imag-2*om1.imag,m*om2.imag+2*om1.imag],'k',alpha=0.2,linewidth=1)
 
plt.text(om1.real+.05, om1.imag-.05, '$\omega_1$')   # lattice lables
plt.text(om2.real-.15, om2.imag+.05, '$\omega_2$')
plt.text((om1+om2).real-.1, (om1+om2).imag+.1, '$\omega_1+\omega_2$')


def Dslice(om):    # find plot and return perpendicular bisectors
    u = om.real
    v = om.imag
    b = (u**2+v**2)/(2*v)
    m = -(u/v)
    x = np.linspace(-2,2,20)
    y = m*x + b
    plt.plot(x,y,'b',linewidth=1,alpha=0.2)
    return m, b

m1, b1 = Dslice(om1)    # slice and then find intersections of perp bisectors
m2, b2 = Dslice(om2)   # m1*x + b1 = m2*x + b2
xhex[0] = (b2-b1)/(m1-m2)
yhex[0] = m1*xhex[0] + b1
plt.plot(xhex[0],yhex[0],'bh')
m1 = m2
b1 = b2

m2, b2 = Dslice(om2-om1)
xhex[1] = (b2-b1)/(m1-m2)
yhex[1] = m1*xhex[1] + b1
plt.plot(xhex[1],yhex[1],'bh')
m1 = m2
b1 = b2

m2, b2 = Dslice(-om1)
xhex[2] = (b2-b1)/(m1-m2)
yhex[2] = m1*xhex[2] + b1
plt.plot(xhex[2],yhex[2],'bh')
m1 = m2
b1 = b2

m2, b2 = Dslice(-om2)
xhex[3] = (b2-b1)/(m1-m2)
yhex[3] = m1*xhex[3] + b1
plt.plot(xhex[3],yhex[3],'bh')
m1 = m2
b1 = b2

m2, b2 = Dslice(om1-om2)
xhex[4] = (b2-b1)/(m1-m2)
yhex[4] = m1*xhex[4] + b1
plt.plot(xhex[4],yhex[4],'bh')
m1 = m2
b1 = b2

m2, b2 = Dslice(om1)
xhex[5] = (b2-b1)/(m1-m2)
yhex[5] = m1*xhex[5] + b1
plt.plot(xhex[5],yhex[5],'bh')

# draw and label partitions and thier translates

xQ1 = [0, om1.real/2, xhex[0], om2.real/2]
yQ1 = [0, om1.imag/2, yhex[0], om2.imag/2]
plt.fill(xQ1, yQ1, alpha = 0.5)
plt.text(np.mean(xQ1),np.mean(yQ1),'$P_1$')

xQ2 = np.array([0, om2.real/2, xhex[1], xhex[2], -om1.real/2])
yQ2 = np.array([0, om2.imag/2, yhex[1], yhex[2], -om1.imag/2])
plt.fill(xQ2, yQ2, 'r', alpha = 0.5)
plt.text(np.mean(xQ2),np.mean(yQ2),'$P_2$')
plt.fill(xQ2 + om1.real, yQ2 + om1.imag, 'r', alpha = 0.5)
plt.text(np.mean(xQ2)+ om1.real-.1,np.mean(yQ2) + om1.imag,'$P_2+\omega_1$')

xQ3 = np.array([0, -om1.real/2, xhex[3], -om2.real/2])
yQ3 = np.array([0, -om1.imag/2, yhex[3], -om2.imag/2])
plt.fill(xQ3, yQ3, 'g', alpha = 0.5)
plt.text(np.mean(xQ3),np.mean(yQ3),'$P_3$')
plt.fill(xQ3+om1.real+om2.real, yQ3+om1.imag+om2.imag, 'g', alpha = 0.5)
plt.text(np.mean(xQ3)+ om1.real + om2.real -.2,np.mean(yQ3) + om1.imag + om2.imag -.1,'$P_3+\omega_1+\omega_2$')

xQ4 = np.array([0, -om2.real/2, xhex[4], xhex[5], om1.real/2])
yQ4 = np.array([0, -om2.imag/2, yhex[4], yhex[5], om1.imag/2])
plt.fill(xQ4, yQ4, 'c', alpha = 0.5)
plt.text(np.mean(xQ4),np.mean(yQ4),'$P_4$')
plt.fill(xQ4 + om2.real, yQ4 + om2.imag, 'c', alpha = 0.5)
plt.text(np.mean(xQ4)+ om2.real-.1,np.mean(yQ4) + om2.imag,'$P_4+\omega_2$')

plt.axis([-3/2,3/2,-3/2,3/2])
plt.xlabel('real part',fontsize=14)
plt.ylabel('imaginary part',fontsize=14);
```

This figure illustrates the fact that the Dirichlet Hexagon has the same area as the Fundamental Parallelogram, as well as the fact that the complex plane may be tiled with translates of the Dirichlet Hexagon.  It should not be too surprising then that

\begin{proposition}\label{prop:DFD} Given a lattice, $\Om$, its Dirichlet Hexagon, $D(\Om)$, is a Fundamental Domain.
\end{proposition}

**Proof:** For arbitrary $z\in\maC$ we must first show that its orbit, $z+\Om$, has a point in $D(\Om)$. A natural candidate is one of the smallest members of $z+\Om$, say $z_0$. (To see that such a point exists: note that as the parallelograms tile $\maC$ there must be a parallelogram $\hat P$, with vertices in $\Om$, that contains $-z$. Let $\hat\om$ be the closest vertex in $\hat P$ to $-z$. Now $z_0=\hat\om+z$ will be as small as any other point in $z+\Om$.)
Now, to say that $z_0$ is as small as any other point in $z+\Om$ is to say that $|z_0|\le|z+\om|$ for every $z\in\Om$, which, recalling (\ref{eq:Dir0}) is to say that $z_0\in D(\Om)$.

We next must show that no orbit can land more that one point in the interior of $D(\Om)$. A point $\zeta$ lie in interior of $D(\Om)$ when it satisfies the strict inequality in (\ref{eq:Dir0}), i.e., when $|\zeta|<|\zeta-\om|$ for all $\om\in\Om\setminus\{0\}$. Now suppose both $\zeta_1$ and $\zeta_2$ both belong to $z+\Om$, i.e., $\zeta_1=z+\om_1$ and
$\zeta_2=z+\om_2$ for $\om_1$ and $\om_2$ in $\Om$. It follows that $\zeta_1-\zeta_2=\om_1-\om_2\in\Om$. Hence, if $\zeta_1$ lies in the interior of $D(\Om)$ then $|\zeta_1|<|\zeta_1-(\zeta_1-\zeta_2)|=|\zeta_2|$. By the same argument, if $\zeta_2$ lies in the interior of $D(\Om)$ then $|\zeta_2|<|\zeta_1|$. As these two inequaties are contradictory we find that two points of the same orbit *cannot* lie in the interior of $D(\Om)$. **End of Proof.**


### M&ouml;bius Transformations

We observed the remarkable role that M&ouml;bius Transformations of a complex variable play in identifying similar lattices. We move now from integer coefficients to complex coefficients and so study, for complex $z$,

\begin{equation}\label{eq:mobz}
  \mu(z) \equiv \frac{az+b}{cz+d} \where M \equiv \begin{pmatrix} a & b \\ c & d \end{pmatrix} \in\SL_2(\maC)
\end{equation}

lies in the set of 2-by-2 complex matrices with determinant one. As in the real case these transformations are defined on an extended domain. In the real case we appended to $\maR$ the "point at infinity" by constructing the real projective line, $\maP^1(\maR)$, of real lines through the origin in $\maR^2$. The natural home for (\ref:eq:mobz}) is the corresponding **complex projective line**, $\maP^1(\maC)$, of complex lines through the origin in $\maC^2$. As in the real case this permits us to make sense of

\begin{equation}\label{eq:mobzinf}
  \mu(-d/c) = \infty \and \mu(\infty) = a/c\where M \equiv \begin{pmatrix} a & b \\ c & d \end{pmatrix} \in\SL_2(\maC)
\end{equation}

Our proof that the composition of real M&ouml;bius Transformations correspond to the products of the associated matrices extends, without change, to the complex case.

Whereas every real M&ouml;bius Transformation could be expressed as a composition of a dilation, $t\mapsto\alpha^2t$, translation, $t\mapsto t+\sigma$, and rotation by $\pi/2$, $t\mapsto -1/t$, in the complex setting, real dilation branches into dilation of magnitude and translation of phase. 

\begin{proposition}\label{prop:mobC} Each complex M&ouml;bius Transformation can be expressed as a composition of the four

\begin{equation}\label{eq:Mob3T}
  \eqalign{
  A_\alpha(z) &\equiv \alpha z, \hskip 0.25in \alpha>0, 
        \hskip 0.25in A(\alpha) = \begin{pmatrix} \sqrt\alpha & 0 \\ 0 & 1/\sqrt\alpha \end{pmatrix}  \cr
  R_\theta(z) &\equiv \exp(i\theta)z, \hskip 0.25in \theta\in\maR,
      \hskip 0.25in R(\theta) = \begin{pmatrix} \exp(i\theta/2) & 0 \\ 0 & \exp(-i\theta/2) \end{pmatrix} \cr
  N_\sigma(z) &\equiv z + \s z, \hskip 0.25in \s\in\maC,
      \hskip 0.25in N(\s) = \begin{pmatrix} 1 & \sigma \\ 0 & 1 \end{pmatrix}\cr
  K_{\pi/2}(z) &\equiv -1/z, \hskip 0.25in K(\pi/2) = \begin{pmatrix} 0 & -1 \\ 1 & 0 \end{pmatrix} }
\end{equation}

elementary transformations.
\end{proposition}

**Proof:** If $c=0$ then

\begin{equation}\label{eq:Mobd1}
   \frac{az+b}{cz+d} = (a/d)z + (b/d) = N_{b/d}(R_\theta(A_\alpha(z))), \where a/d = \alpha\exp(i\theta)
\end{equation}

If $c\ne 0$ then

\begin{equation}\label{eq:Mobd2}
   \frac{az+b}{cz+d} = \frac{a}{c} - \frac{1}{c(cz+d)} = N_{a/c}(K_{\pi/2}(N_{cd}(R_\theta(A_\alpha(z)))) \where c^2 = \alpha\exp(i\theta)
\end{equation}

**End of Proof.**

### Circles and Lines to Circles and Lines

In Figure \ref{fig:4trans} we see evidence that our four elementary transformations take lines to lines or circles. This is self-evident for the first three, while the fourth deserves some attention. We follow the horizontal line $z=x+i$ under

\begin{equation}\label{eq:InvH}
   K_{\pi/2}(z) = \frac{-1}{z} = \frac{-1}{x+i} = \frac{i-x}{x^2+1} \equiv u + iv \where u = \frac{-x}{x^2+1} \and v = \frac{1}{x^2+1}
\end{equation}

From Figure \ref{fig:4trans} we expect this to be a circle centered on the imaginary axis in the $(u,v)$ plane and so we ask if $u$ and $v$ obey

\begin{equation}\label{eq:invH1}
 u^2 + (v-v_0)^2 = r^2, \hskip 0.25in \hbox{that is}\hskip 0.25in \frac{x^2+1+v_0^2(x^2+1)^2-2v_0(x^2+1)}{(x^2+1)^2} = r^2,
\end{equation}

from which we deduce that $v_0=r=1/2$. That is, $-1/z$ takes the horizontal line, $z=x+i$, to the circle $u^2+(v-1/2)^2=(1/2)^2$. As $-1/z$ is its own inverse it follows that $-1/z$ also maps this circle back onto the original line.

### Inversions 

We learned from our work in [1.3 Matrix Matrix Multiplication](1.3.MatrixMatrixMultiplication.ipynb) that every rotation can be expressed as a product of reflections. In this section we generalize the notion of reflection across a line to **inversion** across a circle, and show that **every** M&ouml;bius transformation is the product of an even number of inversions. Where a reflection exchanges sides, an inversion exchanges the inside with the outside and so can be thought of as turning a disk inside out.

We begin with a circle, $C=\{z:|z-z_0|=r\}$, of radius $r$ centered at $z_0$. The inversion of $z\in\maC$ across $C$ is the point $w$ on the line through $z_0$ and $z$ for which

\begin{equation}\label{eq:ICd}
  |z-z_0||w-z_0| = r^2.
\end{equation}

As $z$ and $w$ lie on the same line through $z_0$ they have same arguments with respect to $z_0$, i.e.,

\begin{equation}\label{eq:ICd2}
  \arg(z-z_0) = \arg(w-z_0).
\end{equation}

Now as, $\arg(v)=-\arg(v)$ for every $v\in\maC$, we deduce from (\ref{eq:ICd}) and (\ref{eq:ICd2}) that

\begin{equation}\label{eq:ICd3}
  \eqalign{ (\ov z-\ov z_0)(w - z_0) 
  &= |z-z_0|\exp(-i\arg(z-z_0))|w-z_0|\exp(i\arg(z-z_0)) \cr
  &= |z-z_0||w-z_0| \cr
  &= r^2.}
\end{equation}

We solve this equation for $w$, and denote the result by $I_C(z)$ to signify that it is the inversion of $z$ across $C$,

\begin{equation}\label{eq:INVdef}
  I_C(z) = \frac{r^2 - |z_0|^2 + z_0\ov z}{\ov z - \ov z_0}
\end{equation}

We illustrate the inversion of points and shapes in Figure \ref{fig:Icz}. Please note that the two Inversion equations (\ref{eq:ICd}-\ref{eq:ICd2}) are identical to the equations for the Peaucellier-Lipkin linkage established in our [Kinematics](1.4.Kinematics.ipynb) notebook.

```python
# Inversion in a circle

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2)
plt.subplots_adjust(right=2, top=1, wspace=0.25)

x0 = 1
y0 = 1
z0 = x0 + y0*1j
t = np.linspace(0, 2*np.pi, 100)
r = 0.5
x = x0 + r*np.cos(t)
y = y0 + r*np.sin(t)

ax1.plot(x,y)

N = 10
for n in range(1,N):
    z = x0 + r - (r/2)*n/N + y0*1j
    ICz = (r**2 - abs(z0)**2 + z0*np.conj(z)) / (np.conj(z-z0))
    ax1.plot([np.real(z), np.real(ICz)], [np.imag(z), np.imag(ICz)], '.', markersize=6)
    
# invert a square
Xs = np.linspace(0.9,1.1,500)
Ys = np.linspace(1.2,1.4,500)

for xs in Xs:
    zs = xs + Ys[0]*1j
    ICzs = (r**2 - abs(z0)**2 + z0*np.conj(zs)) / (np.conj(zs-z0))
    ax1.plot([np.real(zs), np.real(ICzs)], [np.imag(zs), np.imag(ICzs)], 'k.', markersize=1)
    
for ys in Ys:
    zs = Xs[-1] + ys*1j
    ICzs = (r**2 - abs(z0)**2 + z0*np.conj(zs)) / (np.conj(zs-z0))
    ax1.plot([np.real(zs), np.real(ICzs)], [np.imag(zs), np.imag(ICzs)], 'r.', markersize=1)
      
for xs in Xs:
    zs = xs + Ys[-1]*1j
    ICzs = (r**2 - abs(z0)**2 + z0*np.conj(zs)) / (np.conj(zs-z0))
    ax1.plot([np.real(zs), np.real(ICzs)], [np.imag(zs), np.imag(ICzs)], 'b.', markersize=1)
    
for ys in Ys:
    zs = Xs[0] + ys*1j
    ICzs = (r**2 - abs(z0)**2 + z0*np.conj(zs)) / (np.conj(zs-z0))
    ax1.plot([np.real(zs), np.real(ICzs)], [np.imag(zs), np.imag(ICzs)], 'g.', markersize=1)
    
ax1.grid('on')
ax1.axis('equal')


ax2.plot([np.real(z), np.real(ICz)], [np.imag(z), np.imag(ICz)], 'k.', markersize=10)
ax2.plot(x,y)
c = (ICz + z)/2 + 0.2j
rad = np.abs(c-z)
x2 = c.real + rad*np.cos(t)
y2 = c.imag + rad*np.sin(t)
ax2.plot(x2,y2)

ax2.plot([np.real(z0), np.real(c)], [np.imag(z0), np.imag(c)], 'k.-')  # connect centers
# circles intersect at 
indi = 10
indj = 10
for i in range(100):
    for j in range(100):
        if (abs(x[i]-x2[j]) + abs(y[i]-y2[j]) < 1e-2 and y[j] < 1):
            indi = i
            indj = j
            

ax2.plot([np.real(z0), x[indi]], [np.imag(z0), y[indi]], 'k.-')  # connect
ax2.plot([np.real(c), x[indi]], [np.imag(c), y[indi]], 'k.-')  # connect

ax2.text(np.real(z0), np.imag(z0)+.04, '$z_0$', fontsize=12)
ax2.text(np.real(c)+.02, np.imag(c), '$z_1$', fontsize=12)
ax2.text(x[indi], y[indi]-.06, '$z_2$', fontsize=12)
ax2.text(x[indi]-0.04, y[indi]+.04, '$\\theta$', fontsize=12)

ax2.text(np.real(z), np.imag(z)+.04, '$z$', fontsize=12)
ax2.text(np.real(ICz)-0.04, np.imag(ICz)+.04, '$w$', fontsize=12)
ax2.plot(np.real(c), np.imag(z0), 'k.',markersize=10)
ax2.text(np.real(c), np.imag(z0)+.04, '$u$', fontsize=12)

ax2.grid('on')
ax2.axis('equal');
```

\begin{myfig}\label{fig:Icz} **Left:** Illustration of inversion in the blue circle. Points retain their color when inverted. The square is mapped onto the curvilinear quadrilateral with side mappings indicated by color. We note that lengths and areas are distorted and that orientation is reversed. **Right:** The points $z$ and $w$ are inverse across the blue circle. The orange circle through these points *appears* perpendicular to the blue circle at $z_2$. To prove this we will use the ancillary point $u$, the bisector of the segment from $z$ to $w$. \end{myfig}

To establish that the blue circle is perpendicular to the orange circle at $z_2$ in Figure \ref{fig:Icz} we argue that the radial segments meeting at $z_2$ meet at a right angle, i.e., $\theta=\pi/2$. To achieve that it suffices, by the Law of Cosines, to establish the Pythagorean identity

\begin{equation}\label{eq:LoC1}
  |z_0-z_2|^2 + |z_2-z_1|^2 = |z_0-z_1|^2.
\end{equation}

To get there we note that the triangle with vertices $\{z_0,u,z_1\}$ is a right triangle and so

\begin{equation}\label{eq:LoC2}
  \eqalign{|z_0-z_2|^2 &= |z_0-z||z_0-w| \cr
    &= (|z_0-u|-|z-u|)(|z_0-u|+|z-u|) \cr
    &= |z_0-u|^2 - |z-u|^2 \cr
    &= (|z_0-z_1|^2 - |z_1-u|^2) - |z-u|^2 \cr
    &= |z_0-z_1|^2 - (|z_1-u|^2 + |z-u|^2) \cr
    &= |z_0-z_1|^2 - |z-z_1|^2 \cr
    &= |z_0-z_1|^2 - |z_2-z_1|^2 \cr}
\end{equation}

establishes (\ref{eq:LoC1}). This is Theorem 5 of https://www.malinc.se/noneuclidean/en/circleinversion.php

Although the inversion, $I_C(z)$, of (\ref{eq:INVdef}) is not quite M&ouml;bius it is the progenitor of all M&ouml;bius transformations.

\begin{proposition}\label{prop:InvPro} Each M&ouml;bius transformation is the composition of an even number of inversions and reflections.

\end{proposition}

**Proof:** We show that each elementary transformation, (\ref{eq:Mob3T}), is the composition of a pair of inversions or reflections. We begin with the translation, $N_\sigma(z) \equiv z + \s z$. With regard to the leftmost panel in Figure \ref{fig:InvPro}, given the point $\sigma$ we construct the parallel lines $L_1$ and $L_2$ perpendicular to the line through $\sigma$ at $0$ and $\sigma/2$.

We next decompose $R_\theta(z)\equiv \exp(i\theta)z$ as reflection across the real axis, $L_1$, followed by reflection across the line $L_2$, at angle $\theta/2$.

We next decompose $A_\alpha(z)\equiv \alpha z$ as inversion across the unit circle, $C_1$, followed by inversion across, $C_2$, the circle of radius $\sqrt\alpha$ centered at $0$. For $z=|z|\exp(i\theta)$, this composition reads

\begin{equation*}
  I_{C_2}(I_{C_1}(z)) = I_{C_2}(\exp(i\theta)/|z|) = \alpha|z|\exp(i\theta) = \alpha z
\end{equation*}

as claimed.
We finally decompose $K_{\pi/2}(z)\equiv -1/z$ as reflection across the imaginary axis, $L_1$, followed by inversion across the unit circle, $C_2$. Algebraically, this composition reads

\begin{equation*}
  I_{C_2}(I_{L_1}(z)) = I_{C_2}(-\ov z) = -\ov z /|z|^2 = -1/z
\end{equation*}

as claimed. **End of Proof**

The iterates of a single inversion simply bounce back and forth. The iterates of compositions of several inversions can lead to beautiful patterns. Consider, for example, three reflections across the sides of an equilateral triangle in Figure \ref{fig:TriTile}(Left). The reflections across the three sides, $L_1$, $L_2$, and $L_3$ are

\begin{equation}\label{eq:TriInv3}
  \cI_1(z) = \ov z, \hskip 0.25in \cI_2(z) = \frac{(\sqrt 3 + i)\ov z - \sqrt 3}{-\sqrt{3}+i}, 
  \hskip 0.25in \cI_3(z) = \frac{(-\sqrt 3 + i)\ov z - \sqrt 3}{\sqrt{3}+i}
\end{equation}

respectively. In Figure \ref{fig:TriTile}(Center) we present a graphical enumeration of the first seven levels of composition of these three inversions.


```python
# Breadth First Search on Tiling by reflections of equilateral triangle
# This is adapted from Box 12 on page 115 of Indra's Pearls

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig, (ax0, ax1, ax2) = plt.subplots(1,3)
plt.subplots_adjust(right=2, top=1, wspace=0.25)

def legend_without_duplicate_labels(ax):      # a method for removing duplicate labels
    handles, labels = ax.get_legend_handles_labels()
    unique = [(h, l) for i, (h, l) in enumerate(zip(handles, labels)) if l not in labels[:i]]
    ax.legend(*zip(*unique))

def PlotTri(T, lev, col):        # transform triangle by T and fill
    if (np.mod(lev,2) == 1):     # if level is odd then conjugate curve
        z = np.conjugate(C)
    else:
        z = C
    TC = (T[0,0]*z + T[0,1])/(T[1,0]*z + T[1,1])  # apply the map
    ax2.fill(TC.real, TC.imag, col, label=str(lev))  # color and label (for legend) by level

# parametrize and bundle the three sides of the triangle

s = np.sqrt(3)
t = np.linspace(-1/2,1/2,100)
C1 = t
x = np.arange(1/2, 0, -1/100)
C2 = x + 1j*s*(1/2 - x)
x = np.arange(0, -1/2, -1/1000)
C3 = x + 1j*s*(1/2 + x)
C = np.hstack([C1, C2, C3])

ax0.fill(C.real, C.imag, 'w', edgecolor = 'k', hatch = '///')   # draw and label original triangle
ax0.text(0, -.07, '$L_1$', fontsize=12)
ax0.text(.3, .4, '$L_2$', fontsize=12)
ax0.text(-.35, .4, '$L_3$', fontsize=12)
ax0.text(0.45, -0.05, '$v_1$', fontsize=12)
ax0.text(0, s/2+.025, '$v_2$', fontsize=12)
ax0.text(-.5, -.05, '$v_3$', fontsize=12)
ax0.axis('equal')
ax0.axis('off')

# build and bundle the three inversions

I1 = np.array([[1,0],[0,1]])
b = -s + 1j
c = s
I2 = np.array([[-np.conjugate(b),-c],[0,b]])
b = s + 1j
c = s
I3 = np.array([[-np.conjugate(b),-c],[0,b]])

gens = np.array([I1, I2, I3])   #   (3,2,2)  

colormap = ['r','g','b','c','orange','m','k']

bigN = 800  # number of mobius transformations

ga = np.zeros(bigN)   # graph angles and coordinates
gx = np.zeros(bigN)
gy = np.zeros(bigN)

group = np.zeros((bigN,2,2), dtype=complex)   # storage for mobius transformations
tag = np.zeros(bigN, dtype=int)       # storage for tags (1, 2, or 3)
levmax = 7
num = np.zeros(levmax+1, dtype=int)   # storage for num (number of transforms per level)

r = 14/15   # length of graph edge

for j in range(3):
    group[j] = gens[j]    # start with the generators
    tag[j] = j
    ga[j] = 2*j*np.pi/3
    gx[j] = r*np.cos(ga[j])
    gy[j] = r*np.sin(ga[j])
    ax1.plot([0, gx[j]], [0, gy[j]], colormap[0])
    ax1.text(gx[j]/2, gy[j]/2, str(j+1))
    
num[0] = 0
num[1] = 3

for lev in range(1,levmax):   # build the transformations and graph coords
    inew = num[lev]
    r = r*r                  # decreasing radius in graph
    for iold in range(num[lev-1], num[lev]):
        for j in range(1,3):
            k = np.mod(tag[iold] + j, 3) # never follow an inversion with itself
            group[inew] = gens[k] @ np.conjugate(group[iold])
            ga[inew] = ga[iold] + (-1)**j*(np.pi/4.5)*(1-lev/(levmax+1)) # decreasing aperture
            gx[inew] = gx[iold] + r*np.cos(ga[inew])
            gy[inew] = gy[iold] + r*np.sin(ga[inew])
            ax1.plot([gx[iold], gx[inew]], [gy[iold], gy[inew]], colormap[np.mod(lev,7)])
            if (lev<4): ax1.text((gx[iold]+gx[inew])/2, (gy[iold]+gy[inew])/2, str(k+1))   # only label first few levels
            tag[inew] = k
            inew = inew + 1
    num[lev+1] = inew
    
# work backwards and paint by level (working backwards offers a nicer pattern - reverse and see for yourself)

for lev in range(levmax,0,-1):
    for val in range(num[lev],num[lev-1],-1):
            PlotTri(group[val-1], lev, colormap[np.mod(lev-1,7)]) 
                     
ax2.fill(C.real, C.imag, 'w', hatch = '/////')  # repaint level 0

ax1.axis('equal')
ax1.axis('off')

legend_without_duplicate_labels(ax2)
                 
ax2.axis('equal')
ax2.axis('off');
```

\begin{myfig}\label{fig:TriTile} Tiling the plane with equilateral triangles. **Left:** the hatched triangle with labeled sides and vertices. **Middle:** A graphical representation of the composition of reflections. Starting from the center, the red edges labeled 1, 2, and 3, signify the application of $\cI_1$, $\cI_2$, and $\cI_3$ to the original, hatched, triangle. Following the red edge 1 we see green edges labeled 2 and 3. These indicate the application of $\cI_2\circ\cI_1$ and $\cI_3\circ\cI_1$ to the original, hatched, triangle. **Right:** The resulting images of the hatched triangle under the maps built (and colored by level) according to the recipe in our Middle graph.\end{myfig}

The final tiling at right reveals structure not present in our simple tree representation. In particular, the graph has 12 blue edges but our tiling has only 9 blue tiles. This is a result of the identities

\begin{equation}\label{eq:3Ids}

\end{equation}

### Fixed Points and Orbits

We now return to our classification

 (a) When $|\tr(M)| = 2$ then $\mu$ has one real fixed point and is call **parabolic**. If $\alpha$ is the fixed point of $\mu$ then, $S(z) = 1/(z-\alpha)$ maps $\alpha$ to $\infty$ and so $W=S\mu S^{-1}$ is parabolic and takes $\infty$ to
$\infty$ and so $W(z) = z + t$ for some nonzero $t\in\maR$. With $V(z)=z/|t|$ we find that $VWV^{-1}$ is $z\pm 1$. We will see in the exercise below that $z+1$ is *not* conjugate to $z-1$.
 
 (b) When $|\tr(M)| > 2$ then $\mu$ has two real fixed points and is call **hyperbolic**. If $\alpha$ and $\beta$ are the two fixed points we note that $S(z) \equiv (z-\alpha)/(z-\beta)$ obeys $S(\alpha)=0$ and $S(\beta)=\infty$ and so $W=S\mu S^{-1}$ is hyperbolic and fixes $0$ and $\infty$ and so $W(z)=tz$ for some $t\in\maR\setminus\{0,1\}$. With $B(z)=-1/z$ then $(BWB^{-1})(z) = (1/t)z$. We will see in the exercise below that $tz$ is *not* conjugate to $sz$ unless $t=s$ or $t=1/s$.

 (c) When $|\tr(M)| < 2$ then $\mu$ has two complex (conjugate) fixed points and is call **elliptic**.


We can now return to unfinished business, namely the two complex fixed points of 

\begin{equation}\label{eq:Ktz}
  K_\theta(z) = \frac{\cos(\theta)z - \sin(\theta)}{\sin(\theta)z + \cos(\theta)}
\end{equation}

This has the fixed points $\pm i$ and so its orbits are circles about $i$ and $-i$. As such, each orbit can be distinguished by a real $\alpha$ of magnitude greater than one, and $(K_\theta)^{(n)}(\alpha i) = K_{n\theta}(\alpha i)$.

### Circle Pairing Transformations

We denote by $C_r(c)$ the circle of radius $r$ and center $c$ and consider the M&ouml;bius transformation

\end{exercise}

We next label the inverses of (\ref{eq:mua}) and (\ref{eq:mub}) as $\mu_A=\mu_a^{-1}$ and $\mu_B=\mu_b^{-1}$ and explore compositions of the four maps, $\mu_a,\mu_A, \mu_b, \mu_B$. 

```python
# Plot 4 theta-Schottky disks with theta = pi/4
# along with their images under 2 pair of mobius maps,
# where each pair of maps "pair" a pair of disks
#
# Based on Figure 4.4 in Indra's Pearls by Mumford, Series & Wright

import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
%matplotlib inline

fig, axs = plt.subplots(1, 3)
plt.subplots_adjust(right=2, top=1, wspace=0.25)

def PlotCirc(Cc, Cr, colormap, ax): 
    r = np.linspace(0, Cr, 50)
    a = np.linspace(0, 2*np.pi, 500) 
    R, A = np.meshgrid(r,a)
    X = Cc.real + R*np.cos(A)
    Y = Cc.imag + R*np.sin(A)
    Z = R**2
    axs[ax].contourf(X, Y, Z, 40, cmap=colormap)   # give depth to the disk
    
def MobiusOnCircle(T, Cc, Cr):  # Box 10, page 91, Indra
    z = Cc - Cr**2/np.conjugate(T[1,1]/T[1,0] + Cc)
    Dc = MobiusOnPoint(T, z)
    Dr = abs(Dc - MobiusOnPoint(T, Cc+Cr))
    return Dc, Dr

def MobiusOnPoint(T, z):
    return (T[0,0]*z + T[0,1])/(T[1,0]*z + T[1,1])

theta = np.pi/4   # build the maps and circles, 
s = np.sin(theta)  # from Project 4.2, page 118, Indra
c = np.cos(theta)

a = np.array([[1, 1j*c],[-1j*c, 1]])/s
A = inv(a)

b = np.array([[1, c], [c, 1]])/s
B = inv(b)

gens = np.array([a, b, A, B])   #   shape is (4,2,2)  
invind = np.array([2, 3, 0, 1])     # indicies of inverses in gens

r = 1./np.cos(theta)
Ccenb = r 
Crb = np.tan(theta) 
Ccena = 1j*r 
Cra = np.tan(theta) 
CcenB = -r 
CrB = np.tan(theta) 
CcenA = -1j*r 
CrA = np.tan(theta) 

circ = np.array([[Ccena, Cra], [Ccenb, Crb], [CcenA, CrA], [CcenB, CrB]])

colormap = ['Reds', 'Greens', 'Blues', 'Wistia']
ecolors = ['r', 'g', 'b', 'orange']
Dname = ['$D_a$', '$D_b$', '$D_A$', '$D_B$']
name = ['a', 'b', 'A', 'B']

for j in range(4):   # draw initial disks
    PlotCirc(circ[j,0], circ[j,1], colormap[j], 0) 
    axs[0].text(1.1*np.real(circ[j,0]), 1.1*np.imag(circ[j,0]), Dname[j], fontsize = 10)

    
for j in range(4): # draw next generation
    for k in range(4):
        if k != invind[j]:
            G = gens[k,:]
            [Dcen, Drad] = MobiusOnCircle(G, circ[j,0], circ[j,1])
            PlotCirc(Dcen, Drad, colormap[j], 0)
            axs[0].text(np.real(Dcen)-Drad/1.5, np.imag(Dcen)-Drad/5, name[k]+Dname[j], fontsize = 10)

axs[0].plot([-.4, .4, .4, -.4, -.4],[0.6, 0.6, 1.4, 1.4, 0.6], 'k--')
axs[0].axis('equal')
axs[0].axis('off')

# now for the zoom

bigN = 400

ga = np.zeros(bigN)   # graph angles and coordinates
gx = np.zeros(bigN)
gy = np.zeros(bigN)

group = np.zeros((bigN, 2, 2), dtype=complex)  
tag = np.zeros(bigN, dtype=int)
levmax = 3
num = np.zeros(levmax+1, dtype=int)
r = 5/6
for j in range(4):
    group[j] = gens[j]  
    tag[j] = j
    PlotCirc(circ[j,0], circ[j,1], colormap[j], 2) 
    ga[j] = j*np.pi/2
    gx[j] = r*np.cos(ga[j])
    gy[j] = r*np.sin(ga[j])
    axs[1].plot([0, gx[j]], [0, gy[j]], ecolors[j])
    axs[1].text(gx[j]/2, gy[j]/2, name[j])
    
num[0] = 0
num[1] = 4

for lev in range(1,levmax):
    inew = num[lev]
    r = r*r
    for iold in range(num[lev-1], num[lev]):
        inc = -1   # angle incrementer for graph branch points
        for j in range(4):
            if j != invind[tag[iold]]:
                group[inew] = group[iold] @ gens[j]  
                tag[inew] = j
                ga[inew] = ga[iold] + inc*(np.pi/3)*(1-lev/(levmax+1)) # decreasing aperture
                gx[inew] = gx[iold] + r*np.cos(ga[inew])
                gy[inew] = gy[iold] + r*np.sin(ga[inew])
                axs[1].plot([gx[iold], gx[inew]], [gy[iold], gy[inew]], ecolors[j]) #colormap[lev])
                if (lev<3): axs[1].text((gx[iold]+gx[inew])/2, (gy[iold]+gy[inew])/2, name[j])
                inew = inew + 1
                inc = inc + 1
    num[lev+1] = inew

for i in range(num[levmax]):
    for j in range(4):    # apply ith transform to 3 circles
        if j != invind[tag[i]]:
            Ncen, Nrad = MobiusOnCircle(group[i], circ[j,0], circ[j,1]) 
            PlotCirc(Ncen, Nrad, colormap[j], 2)

axs[1].axis('equal')
axs[1].axis('off')
axs[2].axis('equal')
axs[2].set_xlim(-0.4,0.4)
axs[2].set_ylim(0.6, 1.4)
axs[2].axis('off')
plt.show();
```

\end{myfig}
