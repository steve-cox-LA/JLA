# Complex Integration  


Our goal here is to develop the main results of complex integration theory
and to apply these to partial fraction expansion of the resolvent of a matrix
and to the Inverse Laplace Transform of a rational function. With these tools
and applications we will have placed the dynamics work in \S\subsecref{lap} 
on a solid foundation.

These tools will also permit us to solve the challenging Causal Wiener Filter 
problem set in the previous chapter, and to develop the tools needed for a 
careful study of the eigenvalue perturbation problem.

 

We will be integrating complex functions over complex curves. Such a curve
is parametrized by one complex valued or, equivalently, two real valued, 
function(s) of a real parameter (typically denoted by $t$). More
precisely,

\begin{equation*}
  C \equiv \{z(t)=x(t)+iy(t):t_1\le t\le t_2\}.
\end{equation*}

For example, if $x(t)=y(t)=t$ from $t_1=0$ to $t_2=1$, then $C$ is the line
segment joining $0+i0$ to $1+i$.  We now define

\begin{equation}
  \boxed{\iC{f(z)} \equiv \int_{t_1}^{t_2} f(z(t))z'(t)\,dt.}  \label{eq:ICdef}
\end{equation}

For example, if $C=\{t+it:0\le t\le 1\}$ as above and $f(z)=z$ then

\begin{equation*}
  \iC{z} = \int_0^1(t+it)(1+i)\,dt = \int_0^1\{(t-t)+i2t\}\,dt = i,
\end{equation*}

while if $C$ is the unit circle $\{\exp(it):0\le t\le 2\pi\}$ then

\begin{equation*}
  \iC{z} = \int_0^{2\pi}\exp(it)i\exp(it)\,dt = i\int_0^{2\pi}\exp(i2t)\,dt  =
  i\int_0^{2\pi}\{\cos(2t)+i\sin(2t)\}\,dt = 0.
\end{equation*}

Remaining with the unit circle but now integrating $f(z)=1/z$ we find

\begin{equation*}
  \iC{z^{-1}} = \int_0^{2\pi}\exp(-it)i\exp(it)\,dt = 2\pi i.
\end{equation*}

We generalize this calculation to arbitrary (integer) powers over arbitrary
circles. More precisely, for integer $m$ and fixed complex $a$ we integrate
$(z-a)^m$ over

\begin{equation}
  C(a,\rho) \equiv\{a+\rho\exp(it):0\le t\le 2\pi\},  \label{eq:carho}
\end{equation}

the circle of radius $\rho$ centered at $a$. We find

\begin{equation}
  \eqalign{ \int_{C(a,\rho)}(z-a)^m\,dz 
  &= \int_0^{2\pi}(a+\rho\exp(it)-a)^m \rho i\exp(it)\,dt \cr
  & = i\rho^{m+1}\int_0^{2\pi}\exp(i(m+1)t)\,dt \cr
  & = i\rho^{m+1}\int_0^{2\pi}\{\cos((m+1)t)+i\sin((m+1)t)\}\,dt \cr
  &= \begin{cases}2\pi i & \text{if $m=-1$,} \cr 0 & \text{otherwise,}\end{cases} \cr} 
     \label{eq:res0}
\end{equation}

*regardless*  of the size of $\rho$!

When integrating more general functions it is often convenient to 
express the integral in terms of its real and imaginary parts. More precisely

\begin{equation}
   \iC{f(z)} = \int_C  f(x+iy)(dx+idy) = 
               \int_C  f(x+iy)\,dx + i\int_C  f(x+iy)\,dy.  \label{eq:pregreen}
\end{equation}

This representation is specially suited to one of the cornerstones of
multivariable calculus. In particular, Green's Theorem permits us to express 
our contour integral, \eq(\ref{eq:pregreen}), as an especially 
convenient integral over space.


\begin{proposition}\label{prop:ci1} **Green's Theorem.** If $C$ is a closed curve and $M$ and $N$ are
continuously differentiable real-valued functions on $C_{in}$, the region 
enclosed by $C$, then

\begin{equation*}
  \int_C  M(x,y)\,dx + \int_C N(x,y)\,dy = 
  \int\!\!\!\int_{C_{in}}\left(\px{N}-\py{M}\right)\,dxdy
\end{equation*}

\end{proposition}

**Proof:** We suppose that $C$ encloses the lens shaped region like that of 
Figure \ref{fig:green} with the dual description

\begin{equation*}
  \eqalign{
  C_{in} &= \{(x,y):y_1\le y\le y_2,\ x_a(y)\le x\le x_b(y)\} \cr
         &= \{(x,y):x_1\le x\le x_2,\ y_a(x)\le y\le y_b(x)\}. \cr}
\end{equation*}

\centerline{\psfig{figure=c8/green.eps,width=3in} \quad
            \psfig{figure=c8/green2.eps,width=3in}}

\centerline{\Fig\ref{fig:green}. (A) A lens shaped region.
(B) A union of lenses.}

 The advantage of such a region is that it permits us to express
integrals over $(x,y)$ as sequential, or iterated, integrals over $x$ and $y$
individually. When our integrand is a partial derivative with respect to $x$ or
$y$ we may then invoke the fundamental theorem of calculus, relegating the
final integral to the boundary. More precisely

\begin{equation*}
  \eqalign{\int\!\!\!\int_{C_{in}}\px{N(x,y)}\,dxdy
  &= \int_{y_1}^{y_2}\int_{x_a(y)}^{x_b(y)}\px{N(x,y)}\,dxdy \cr
  &= \int_{y_1}^{y_2}(N(x_b(y),y)-N(x_a(y),y))\,dy 
  = \int_C N(x,y)\,dy \cr}
\end{equation*}

where the second equality is a direct application of the fundamental
theorem of calculus. To see the third equality use Figure \ref{fig:green} as
a guide and note that integrating up along $x_b$ agrees with the orientation
of $C$ while integrating up along $x_a$ is counter to $C$'s orientation, but
as this is prefaced with a minus sign we arrive at integration of $N$ round
the full contour. Similarly

\begin{equation*}
  \eqalign{\int\!\!\!\int_{C_{in}}\py{M(x,y)}\,dxdy
  &= \int_{x_1}^{x_2}\int_{y_a(x)}^{y_b(x)}\py{M(x,y)}\,dydx \cr
  &= \int_{x_1}^{x_2}(M(x,y_b(x))-M(x,y_a(y)))\,dy 
  = -\int_C M(x,y)\,dx \cr}
\end{equation*}

Regarding the final equality, note that integrating rightward along $y_b$
is counter to $C$'s orientation while integrating rightward along $y_a$ 
conforms to $C$'s orientation.  

To complete the proof we return to a general smooth closed curve $C$ and
express $C_{in}$ as a union of lens shaped regions. Take for example the
region of Figure \ref{fig:green}(B) where $C_{in}=R_1\cup R_2$. Now

\begin{equation}
  \eqalign{\int\!\!\!\int_{C_{in}}\px{N(x,y)}\,dxdy
  &= \int\!\!\!\int_{R_1}\px{N(x,y)}\,dxdy + \int\!\!\!\int_{R_2}\px{N(x,y)}\,dxdy \cr
  &= \int_{C_1} N(x,y)\,dy + \int_{C_2} N(x,y)\,dy \cr} \label{eq:genR}
\end{equation}

where $C_j$ is the counterclockwise contour enclosing $R_j$. As $C_1$ and
$C_2$ have opposite orientations along their common (dashed) segment these
contributions to \eq(\ref{eq:genR}) cancel and we find

\begin{equation*}
  \int\!\!\!\int_{C_{in}}\px{N(x,y)}\,dxdy = \int_C N(x,y)\,dy.
\end{equation*}

The same argument applies to the $M$ integral. It also applies to all
finite unions of lens shaped regions. **End of Proof.**

Applying this proposition to \eq(\ref{eq:pregreen}), we find, so long as $C$ 
is closed, that

\begin{equation*}
  \iC{f(z)} = 
  -\int\!\!\!\int_{C_{in}}\left(\px{v}+\py{u}\right)\,dxdy
  +i\int\!\!\!\int_{C_{in}}\left(\px{u}-\py{v}\right)\,dxdy.
\end{equation*}

At first glance it appears that Green's Theorem only serves to muddy the
waters. Recalling the Cauchy-Riemann equations however we find that each
of these double integrals is in fact identically zero! In brief, we have
proven

\begin{proposition}\label{prop:ci2}**Cauchy's Theorem.** If $f$ is differentiable on and in the closed curve $C$ then

\begin{equation*}
  \iC{f(z)} = 0.
\end{equation*}

\end{proposition}

Strictly speaking, in order to invoke Green's Theorem we require not only that
$f$ be differentiable but that its derivative in fact be continuous.
This however is simply a limitation of our simple mode
of proof, Cauchy's Theorem is true as stated.

This theorem, together with (\ref{eq:res0}), permits us to integrate every
proper rational function. More precisely, if $r=f/g$ where 
$f$ is a polynomial of degree at most $m-1$ and $g$ is an $m$th degree
polynomial with $h$ distinct zeros at $\{\la_j\}_{j=1}^h$ with respective
multiplicities of $\{p_j\}_{j=1}^h$ we found that

\begin{equation}
  r(z) = \sum_{j=1}^h\sum_{k=1}^{p_j}\frac{r_{j,k}}{(z-\la_j)^k}.
  \label{eq:pfe1}
\end{equation}

Observe now that if we choose the radius $\rho_j$ so small that 
$\la_j$ is the only zero of 
$g$ encircled by $C_j\equiv C(\la_j,\rho_j)$ then by Cauchy's Theorem

\begin{equation*}
  \iCj{r(z)} = \sum_{k=1}^{p_j}r_{j,k}\iCj{\frac{1}{(z-\la_j)^k}}.
\end{equation*}

In (\ref{eq:res0}) we found that each, save the first, of the integrals under the 
sum is in fact zero. Hence

\begin{equation}
  \iCj{r(z)} = 2\pi i r_{j,1}. \label{eq:res1}
\end{equation}

With $r_{j,1}$ in hand, say from (\ref{eq:pfe2}) or  *residue*  , one may view
(\ref{eq:res1}) as a means for computing the indicated integral. The opposite
reading, i.e., that the integral is a convenient means of expressing $r_{j,1}$,
will prove just as useful. With that in mind, we note that the remaining
residues may be computed as integrals of the product of $r$ and the appropriate
factor. More precisely,

\begin{equation}
  \iCj{r(z)(z-\la_j)^{k-1}} = 2\pi i r_{j,k}. \label{eq:resk}
\end{equation}

It is a simple, but highly important, matter to extend this representation to a matrix of 
rational functions.
More precisely, if $R(z)\equiv (zI-B)^{-1}$ is the resolvent 
associated with $B$ then (\ref{eq:pfe1} and \Eqref{resk}) state that

\begin{equation*}
  R(z) = \sum_{j=1}^h\sum_{k=1}^{p_j}\frac{R_{j,k}}{(z-\la_j)^k}
\end{equation*}

where

\begin{equation}
  R_{j,k} = \frac{1}{2\pi i} \iCj{R(z)(z-\la_j)^{k-1}}. \label{eq:Resk}
\end{equation}

Lets consider these in the concrete setting of critically damped single mass. The resolvent in that case, recall (\ref{eq:kmdres2}), can then be expressed

\begin{equation}
   (sI-B)^{-1} = \frac{1}{(s+1)^2}\begin{bmatrix} s+2 & 1 \\ -1 & s\end{bmatrix}
   = \frac{R_{1,1}}{s+1} + \frac{R_{1,2}}{(s+1)^2}  \label{eq:kmdres3p5}
\end{equation}

where

\begin{equation}
   R_{1,1} = \frac{1}{2\pi i}\begin{bmatrix}\int_{C_1}\frac{z+2}{(z+1)^2}\,dz & 
    \int_{C_1}\frac{1}{(z+1)^2}\,dz  \cr 
  \int_{C_1}\frac{-1}{(z+1)^2}\,dz & \int_{C_1}\frac{z}{(z+1)^2}\,dz 
    \end{bmatrix}  \and
R_{1,2} = \frac{1}{2\pi i}\begin{bmatrix}\int_{C_1}\frac{z+2}{z+1}\,dz & 
    \int_{C_1}\frac{1}{z+1}\,dz  \cr 
  \int_{C_1}\frac{-1}{z+1}\,dz & \int_{C_1}\frac{z}{z+1}\,dz 
    \end{bmatrix}  \label{eq:kmdres4}
\end{equation} 

and $C_1$ encloses the pole $z=-1$. The off-diagonal terms of both matrices
may be computed directly from our bare-handed result, (\ref{eq:res0}). Evaluation
of the diagonal terms will follow from the theory built in the next section.

After (\ref{eq:res1} and \ref{eq:resk})
perhaps the most useful consequence of Cauchy's Theorem is the freedom it grants one
to choose the most advantageous curve over which to integrate. More precisely,


\begin{proposition}\label{prop:ci3}Suppose that $C_2$ is a closed curve that
lies inside the region encircled by the closed curve $C_1$. If $f$ is
differentiable in the annular region outside $C_2$ and inside $C_1$ then

\begin{equation*}
  \int_{C_1}f(z)\,dz = \int_{C_2}f(z)\,dz.
\end{equation*}

\end{proposition}

**Proof:** With reference to the figure below we introduce two vertical 
segments and
define the closed curves $C_3=abcda$ (where the $bc$ arc is clockwise and the
$da$ arc is counter-clockwise) and $C_4=adcba$ 
(where the $ad$ arc is counter-clockwise and the $cb$ arc
is clockwise). By merely following the arrows we learn that

\begin{equation*}
  \int_{C_1}f(z)\,dz = \int_{C_2}f(z)\,dz + \int_{C_3}f(z)\,dz + 
                       \int_{C_4}f(z)\,dz.
\end{equation*}

As Cauchy's Theorem implies that the integrals over $C_3$ and $C_4$ each
vanish, we have our result. **End of Proof.**


\centerline{\psfig{figure=c8/cif2.eps,angle=270,height=1.25in}}

\centerline{\Fig\ref{fig:cif2}. The Curve Replacement Lemma.}

 
As an example, recalling (\ref{eq:pfe1} and \ref{eq:res1}), we may express the integral of
a rational function around a curve that encircles  **all**  of its poles
as a sum of residues

\begin{equation}
  \iC{r(z)} = \sum_{j=1}^h\sum_{k=1}^{p_j}
                          \int_{C_j}\frac{r_{j,k}}{(z-\la_j)^k}\,dz =
              2\pi i\sum_{j=1}^h r_{j,1}. \label{eq:Resq}
\end{equation}

To take a slightly more complicated example let us integrate $f(z)/(z-a)$
over some closed curve $C$ inside of which $f$ is differentiable and $a$
resides. Our Curve Replacement Lemma now permits us to claim that

\begin{equation*}
  \int_C\frac{f(z)}{z-a}\,dz = \int_{C(a,\rho)}\frac{f(z)}{z-a}\,dz.
\end{equation*}

It appears that one can go no further without specifying $f$. The alert
reader however recognizes that the integral over $C(a,\rho)$ is independent of
$\rho$ and so proceeds to let $\rho\to 0$, in which case $z\to a$ and $f(z)\to f(a)$.
Computing the integral of $1/(z-a)$ along the way we are lead to the
hope that

\begin{equation*}
  \int_C\frac{f(z)}{z-a}\,dz = f(a)2\pi i.
\end{equation*}

In support of this conclusion we note that

\begin{equation*}
  \eqalign{ \int_{C(a,\rho)}\frac{f(z)}{z-a}\,dz 
  &= \int_{C(a,\rho)}\left\{\frac{f(z)}{z-a}+\frac{f(a)}{z-a}-\frac{f(a)}{z-a}
                         \right\}\,dz \cr
  &= f(a)\int_{C(a,\rho)}\frac{1}{z-a}\,dz + 
      \int_{C(a,\rho)}\frac{f(z)-f(a)}{z-a}\,dz.  \cr}
\end{equation*}

Now the first term is $f(a)2\pi i$ regardless of $\rho$ while, as $\rho\to 0$, 
the
integrand of the second term approaches $f'(a)$ and the region of integration
approaches the point $a$. Regarding this second term, as the integrand remains
bounded (in fact it tends to $f'(a)$) as the region of integration shrinks to a point, the
integral must tend to zero. We have just proven


\begin{proposition}\label{prop:ci4} **Cauchy's Integral Formula.** If $f$ is differentiable on and in the closed curve $C$ then

\begin{equation}
  f(a) = \frac{1}{2\pi i}\int_C\frac{f(z)}{z-a}\,dz  \label{eq:cif0}
\end{equation}

for each $a$ lying inside $C$.

\end{proposition}

The consequences of such a formula run far and deep. We shall delve into only
one or two. First, we note that, as $a$ does not lie on $C$, the right hand 
side of (\ref{eq:cif0}) is a perfectly smooth function
of $a$. Hence, differentiating each side, we find

\begin{equation}
  f'(a) = \frac{df(a)}{da} = \frac{1}{2\pi i}\int_C\frac{d}{da}\frac{f(z)}{z-a}\,dz 
  = \frac{1}{2\pi i}\int_C\frac{f(z)}{(z-a)^2}\,dz \label{eq:cif1}
\end{equation}

for each $a$ lying inside $C$. Applying this reasoning $n$ times we arrive at
a formula for the $n$th derivative of $f$ at $a$,

\begin{equation}
  \frac{d^nf}{da^n}(a) = \frac{n!}{2\pi i}\int_C\frac{f(z)}{(z-a)^{1+n}}\,dz
  \label{eq:cifn}
\end{equation}

for each $a$ lying inside $C$. The upshot is that once $f$ is shown to be
differentiable it must, in fact, be infinitely differentiable.
Regarding concrete examples, the diagonal terms in (\ref{eq:kmdres4}),
where $C_1$ is a circle centered at $z=-1$, may be evaluated by
(\ref{eq:cif0} and \Eqref{cifn}) respectively. In particular,

\begin{equation}
  \frac{1}{2\pi i}\int_C\frac{s}{s+1} = -1 \and
 \frac{1}{2\pi i}\int_C\frac{s}{(s+1)} = 1.    \label{eq:kmdres5}
\end{equation}

On substiution into (\ref{eq:kmdres4}) we find

\begin{equation*}
  R_{1,1} = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} \and
  R_{1,2} = \begin{bmatrix} 1 & 1 \\ -1 & -1 \end{bmatrix}
\end{equation*}

which on substitution in (\ref{eq:kmdres3p5}) delivers

\begin{equation}
  (sI-B)^{-1} = \frac{1}{s+1}\begin{bmatrix} 1 & 0 \\ 0 & 1\end{bmatrix} 
  + \frac{1}{(s+1)^2}\begin{bmatrix} 1 & 1 \\ -1 & -1\end{bmatrix}
    \label{eq:kmdres3again}
\end{equation}

in complete agreement with (\ref{eq:kmdres3}).
  
As a second example let us consider

\begin{equation*}
  \frac{1}{2\pi i}\int_C\frac{f(z)}{(z-\la_1)(z-\la_2)^2}\,dz
\end{equation*}

where $f$ is differentiable on and in $C$ and $C$ encircles 
both $\la_1$ and $\la_2$. By the curve replacement lemma this integral is the
sum

\begin{equation*}
  \frac{1}{2\pi i}\int_{C_1}\frac{f(z)}{(z-\la_1)(z-\la_2)^2}\,dz +
  \frac{1}{2\pi i}\int_{C_2}\frac{f(z)}{(z-\la_1)(z-\la_2)^2}\,dz
\end{equation*}

where $\la_j$ now lies in only $C_j$. As $f(z)/(z-\la_2)$ is well behaved in 
$C_1$ we may use (\ref{eq:cif0}) to conclude that

\begin{equation*}
  \frac{1}{2\pi i}\int_{C_1}\frac{f(z)}{(z-\la_1)(z-\la_2)^2}\,dz = 
   \frac{f(\la_1)}{(\la_1-\la_2)^2}.
\end{equation*}

Similarly, As $f(z)/(z-\la_1)$ is well behaved in 
$C_2$ we may use (\ref{eq:cif1}) to conclude that

\begin{equation*}
  \frac{1}{2\pi i}\int_{C_2}\frac{f(z)}{(z-\la_1)(z-\la_2)^2}\,dz = 
   \frac{d}{da}\frac{f(a)}{(a-\la_1)}\bigg|_{a=\la_2}.
\end{equation*}

These calculations can be read as a concrete instance of

\begin{proposition}\label{prop:ci5}**The Second Residue Theorem.** If $g$ is a polynomial with roots
$\{\la_j\}_{j=1}^h$ of degree $\{p_j\}_{j=1}^h$ and $C$ is a closed
curve encircling each of the $\la_j$ and $f$ is differentiable on and in $C$
then

\begin{equation*}
  \iC{\frac{f(z)}{g(z)}} = 2\pi i\sum_{j=1}^h{\rm res}(f/g,\la_j)
\end{equation*}

where

\begin{equation}
  {\rm res}(f/g,\la_j) = \lim_{z\to\la_j}\frac{1}{(p_j-1)!}
    \frac{d^{p_j-1}}{dz^{p_j-1}}\left((z-\la_j)^{p_j}\frac{f(z)}{g(z)}\right) \label{eq:res95}
\end{equation}

is called the  **residue** of $f/g$ at $\la_j$ by extension of (\ref{eq:pfe2}).

\end{proposition}

The generality of this statement, and the notation required to specify the residue, should
not obscure the fact that it permits us to compute important integrals by merely *evaluating the good parts at the bad places.*  The bad places are of course the
poles of the integrand and the good part of the integrand is what remains after
multiplying by the offending factor, and perhaps taking a few derivatives. 
For example, the integrands in

\begin{equation*}
  \frac{1}{2\pi i} \int_{C(0,1)}\frac{\exp(z)}{z^p}\,dz, \hskip 0.25in p=1,2,\ldots
\end{equation*}

have $z=0$ as a pole of order $p$ and so we must take $p-1$ derivatives of the
good part, $\exp(z)$, at $z=0$, and finally divide by $(p-1)!$. Hence,

\begin{equation*}
  \frac{1}{2\pi i} \int_{C(0,1)}\frac{\exp(z)}{z^p}\,dz = \frac{1}{(p-1)!}.
\end{equation*}

One of the most useful applications of the Second Residue Theorem is the formula
for the inverse Laplace transform of a rational function.

If $r$ is a rational function with poles $\{\la_j\}_{j=1}^h$ then the
inverse Laplace transform of $r$ is

\begin{equation}
  \boxed{(\cL^{-1}r)(t) \equiv \frac{1}{2\pi i}\int_C r(z)\exp(zt)\,dz} \label{eq:i1}
\end{equation}

where $C$ is a curve that encloses each of the poles of $r$. As a result

\begin{equation}
  (\cL^{-1} r)(t) = \sum_{j=1}^h{\rm res}(r(z)\exp(zt),\la_j).  \label{eq:ilap}
\end{equation}

Let us put this lovely formula to the test.
We take our examples from dynamical systems of Chapter \secref{dyn}. 
According to (\ref{eq:ilap}) the inverse Laplace Transform of 

\begin{equation*}
  r(z) = \frac{1}{(z+1)^2}.
\end{equation*}

is simply the residue of $r(z)\exp(zt)$ at $z=-1$, i.e.,

\begin{equation*}
  {\rm res}(r(z)\exp(zt),-1) = \lim_{z\to -1}\frac{d}{dz}\exp(zt) = t\exp(-t).
\end{equation*}

This closes the circle on the example begun in \S\secref{dyn}.3 and 
continued in Exer.~\secref{dyn}.1.

For our next example we return to (\ref{eq:res1}) and take the inverse Laplace 
transform of the constituents of the resolvent

\begin{equation*}
 (zI-B)^{-1} = \frac{1}{z^2+1}\begin{bmatrix}z & 1 \cr -1 & z \end{bmatrix} \quad\hbox{of}\quad
  B = \begin{bmatrix}0 & 1 \cr -1 & 0 \end{bmatrix}.
\end{equation*}

In particular, we compute

\begin{equation*}
  \cL^{-1}\frac{z}{z^2+1} = \frac{z\exp(zt)(z+i)}{z^2+1}\bigg|_{z=-i} +
   \frac{z\exp(zt)(z-i)}{z^2+1}\bigg|_{z=i} = \exp(-it)/2 + \exp(it/2) = \cos(t)
\end{equation*}

and

\begin{equation*}
  \cL^{-1}\frac{1}{z^2+1} = \frac{\exp(zt)(z+i)}{z^2+1}\bigg|_{z=-i} +
   \frac{\exp(zt)(z-i)}{z^2+1}\bigg|_{z=i} = i\exp(-it)/2 -i\exp(it)/2 = \sin(t)
\end{equation*}

and recall the mysterious (\ref{eq:expvialap}) en route to the lovely

\begin{equation*}
  \exp(Bt) = \cL^{-1}(sI-B)^{-1} = \begin{bmatrix}\cos(t) & \sin(t) \\ -\sin(t) & \cos(t)\end{bmatrix}.
\end{equation*}

You may wish to confirm that this matrix indeed obeys $(\exp(Bt))'=B\exp(Bt)$.

The curve replacement lemma of course gives us considerable freedom in our choice of the
curve $C$ used to define the inverse Laplace transform, (\ref{eq:i1}). As in applications the
poles of $r$ are typically in the left half of the complex plane (why?) it is common to choose
$C$ to be the half circle, see Figure \ref{fig:invlapcontour},

\begin{equation}
  C = C_L(\rho) \cup C_A(\rho), \label{eq:ilapcon0}
\end{equation}

comprised of the line segment, $C_L$, and arc, $C_A$,
  
\begin{equation*}  
  C_L(\rho) \equiv \{i\om:-\rho\le \om\le \rho\} \and 
  C_A(\rho) \equiv \{\rho\exp(i\theta):\pi/2\le \theta\le 3\pi/2\},
\end{equation*}

where $\rho$ is chosen large enough to encircle the poles of $r$. With this concrete choice, 
(\ref{eq:i1}) takes the form

\begin{equation}
  \eqalign{(\cL^{-1}r)(t) 
  &=\frac{1}{2\pi i}\int_{C_L} r(z)\exp(zt)\,dz + \frac{1}{2\pi i}\int_{C_A} r(z)\exp(zt)\,dz \cr
  &=\frac{1}{2\pi}\int_{-\rho}^\rho r(i\om)\exp(i\om t)\,d\om + 
      \frac{\rho}{2\pi}\int_{\pi/2}^{3\pi/2} r(\rho\exp(i\theta))\exp(\rho \exp(i\theta)t)\exp(i\theta)\,d\theta. \cr}  \label{eq:laparc}
\end{equation}

Although this second term appears unwieldy it can be shown to vanish as $\rho\to\infty$, in
which case we arrive at

\begin{equation}
  \boxed{(\cL^{-1}r)(t) = \frac{1}{2\pi}\int_{-\infty}^\infty r(i\om)\exp(i\om t)\,d\om,}
   \label{eq:ilapcon}
\end{equation}

the conventional definition of the inverse Laplace transform.

 
We return to the prediction problem of chapter 6. $\{x_n\}$ is a stationary timeseries (on the event space $\Omega$ with probability density function $p$) and mean 0 and covariances

\begin{equation*}
  c_k = \|x_{j+k}\ov x_j\|^2 \equiv \int_\Omega x_{j+k}(\om)x_j(\om)p(\om)\,d\om
\end{equation*}

If we assemble these into a Fourier Series,

\begin{equation*}
  f(\theta) = \sum_{n=-\infty}^\infty c_n\re^{in\theta}, \hskip 0.25in c_n = \frac{1}{2\pi}\int_{-\pi}^\pi f(\theta)\re^{-in\theta}\,d\theta.
\end{equation*}

Examples:::

The prediction problem is to minimize, over $a\in\maR^n$,

\begin{equation*}
  \eqalign{ \|x_1 - \sum_{k=0}^{n-1} a_kx_{-k}\|^2
   &= \int_\Omega \left(x_1(\om) - \sum_{k=0}^{n-1} a_kx_{-k}(\om)\right)^2p(\om)\,d\om \cr
   &= \frac{1}{2\pi}\int_{-\pi}^\pi \left|\re^{i\theta}-\sum_{k=0}^{n-1} a_k\re^{-ik\theta}\right|^2 f(\theta)\,d\theta \cr}
\end{equation*}

From here we state that two polynomials are orthogonal on the unit circle with weight $f$ if

\begin{equation*}
  \int_{C_1} p_1(z)\ov p_2(z) \frac{w(z)}{z}\, dz
\end{equation*}

Our interest is conditions on $f$ that assure causality, i.e.,

\begin{equation}
  \int_0^1 f(\om)\exp(2k\pi i\om)\,d\om = 0 \hskip 0.25in\forall\ k=-1,-2,\ldots.
  \label{eq:caus0}
\end{equation}

Suppose $f$ is a rational function of $\exp(2\pi i\om)$, i.e.,

\begin{equation*}
  f(\om) = \frac{p(\exp(2\pi i\om))}{q(\exp(2\pi i\om))}
\end{equation*}

where $p$ and $q$ are polynomials.
The change of variable $z=\exp(2\pi i\om)$ will reduce (\ref{eq:caus0})
to a residue problem. In particular, as $dz=2\pi i z d\om$ we find

\begin{equation*}
  \eqalign{\int_0^1 \frac{p(\exp(2\pi i\om))}{q(\exp(2\pi i\om))}\exp(-2k\pi i\om)\,d\om 
  &= \frac{1}{2\pi i}\int_{C_1}\frac{p(z)}{q(z)}z^{-k-1}\,dz \cr
  &= \sum_{j=1}^{h}\sum_{m=1}^{m_j}\frac{r_{j,m}}{2\pi i}\int_{C_1}
    \frac{dz}{(z-\la_j)^mz^{k+1}} \cr}
\end{equation*}

where the last equality follows from the partial fraction expansion of $p/q$.
We now note that if $|\la_j|<1$ then the residues at $z=0$ and 
$z=\la_j$ are equal and opposite

\begin{equation*}
  \eqalign{ \frac{1}{2\pi i}\int_{C_1}(z-\la)^{-m}z^{-k-1}\,dz 
  &= \frac{1}{(m-1)!}\frac{d^{m-1}}{dz^{m-1}}z^{-k-1}\bigg |_{z=\la}  +
   \frac{1}{k!}\frac{d^k}{dz^k}(z-\la)^{-m}\bigg |_{z=0}  \cr
  &= \frac{(m-1+k)!}{(m-1)!k!}(-1)^{m-1}z^{-m-k}\bigg |_{z=\la}  +
   \frac{(m-1+k)!}{k!(m-1)!}(-1)^k(z-\la)^{-m-k}\bigg |_{z=0} \cr
  &= \frac{(m-1+k)!}{k!(m-1)!}\{(-1)^{m-1} + (-1)^m\}\la^{-m-k} = 0. \cr}
\end{equation*}

This provides the clue to revising the IIR filter (\ref{eq:IIRa})
$\hat a(\om) = \hat c_{xy}(\om)/\hat c_{xx}(\om)$. The obvious guess of
simply keeping the good part is too naive. Instead we establish the





\begin{proposition}\label{prop:ci6}. If $f(\om)> 0$ and integrable and is
a rational function of $\exp(2\pi i\om)$ then $f$ may be factored as

\begin{equation}
  f(\om) = L(\exp(2\pi i\om))\ov{L(\exp(2\pi i\om))} \label{eq:specfrac}
\end{equation}

where all poles of $L$ lie strictly inside the unit disk.

\end{proposition}

**Proof:** To begin

\begin{equation*}
  f(\om) = c\frac{\prod_{j=1}^n(\exp(2\pi i\om)-\mu_j)}
       {\prod_{j=1}^d(\exp(2\pi i\om)-\la_j)}
\end{equation*}

where $c\ne 0$ and no $\mu_j$ nor $\la_j$ is zero. Also, no $\la_j$
has magnitude 1, for this would make $f$ nonintegrable. And no $\mu_j$
has magnitude 1, for $f$ never vanishes.

As $f$ is real we equate conjugates and find

\begin{equation*}
  \eqalign{
  c\frac{\prod_{j=1}^n(\exp(2\pi i\om)-\mu_j)}
       {\prod_{j=1}^d(\exp(2\pi i\om)-\la_j)} &=
  \ov c\frac{\prod_{j=1}^n(\exp(-2\pi i\om)-\ov\mu_j)}
       {\prod_{j=1}^d(\exp(-2\pi i\om)-\ov\la_j)} \cr
  &= \ov c\exp(2\pi i (d-n)\omega)
  \frac{\prod_{j=1}^n(1/\ov\mu_j-\exp(2\pi i\om))\ov\mu_j}
       {\prod_{j=1}^d(1/\ov\la_j-\exp(2\pi i\om))\ov\la_j} \cr}
\end{equation*}

and so for pole $\la_j$ we find that $1/\ov \la_j$ is also a pole.
As such we may partition the poles into those lying within the unit
circle $\{\la_j^+:j=1,\ldots,d/2\}$ and their reflections across
the unit circle  $\{1/\ov \la_j^+:j=1,\ldots,d/2\}$.
Similar, we partition the zeros into those lying within the unit
circle $\{\mu_j^+:j=1,\ldots,n/2\}$ and their reflections across
the unit circle  $\{1/\ov \mu_j^+:j=1,\ldots,n/2\}$. With this we may
define

\begin{equation*}
  L(\om) \equiv \frac{\prod_{j=1}^{n/2}(\exp(2\pi i\om)-\mu_j^+)}
       {\prod_{j=1}^{d/2}(\exp(2\pi i\om)-\la_j^+)},
\end{equation*}

note that its zeros and poles lie strictly inside the unit circle
and compute

\begin{equation*}
  \eqalign{L(\om)\ov{L(\om)} 
  &= \frac{\prod_{j=1}^{n/2}(\exp(2\pi i\om)-\mu_j^+)(\exp(-2\pi i\om)-\ov\mu_j^+)}
   {\prod_{j=1}^{d/2}(\exp(2\pi i\om)-\la_j^+)(\exp(-2\pi i\om)-\ov\la_j^+)} \cr
  &= \exp(\pi i(d-n)\om)\frac{\prod_{j=1}^{n/2}(\exp(2\pi i\om)-\mu_j^+)(1/\ov\mu_j^+ - \exp(2\pi i\om))\ov\mu_j^+}
   {\prod_{j=1}^{d/2}(\exp(2\pi i\om)-\la_j^+)(1/\ov\la_j^+ - \exp(2\pi i\om))\ov\la_j^+} \cr
  &= \exp(\pi i(d-n)\om)(-1)^{(n-d)/2}
   \frac{\prod_{j=1}^{n/2}\mu_j^+}{\prod_{j=1}^{d/2}\la_j^+}
   \frac{f(\om)}{c}.\cr}
\end{equation*}

As both the left hand side and $f$ are nonnegative we can take magnitudes of both
sides and conclude that

\begin{equation*}
  L(\om)\ov{L(\om)} = \rho f(\om) \where 
  \rho = \left|\frac{\prod_{j=1}^{n/2}\mu_j^+}{c\prod_{j=1}^{d/2}\la_j^+}\right|
\end{equation*}

On setting $L=L/\sqrt\rho$ we conclude (\ref{eq:specfrac}). **End of Proof.**
  
We assume that $c_{xx}$ is rational and exploit its factorization in the
the associated error spectrum, recall (\ref{eq:IIRerrs}), takes the
form

\begin{equation*}
  \eqalign{\hat c_{\eps\eps}(\om) 
  &= \hat c_{yy}(\om)- \ov{\hat a}(\om)\hat c_{xy}(\om) -
     \hat a(\om) \ov{\hat c}_{xy}(\om) + |\hat a(\om)|^2L(\exp(2\pi i\om))\ov{L(\exp(2\pi i\om))}   \cr
  &= \hat c_{yy}(\om) + 
    |\hat a(\om) L(\exp(2\pi i\om)) - \hat c_{xy}/\ov{L(\exp(2\pi i\om))}|^2 - 
      |\hat c_{xy}(\om)|^2/c_{xx}(\om) \cr}
\end{equation*}

Now, as we wish to determine the best causal $a$ we expect $\hat a$ to have all
of its poles within $C_1$. As the poles of $L$ are also there then the best we
can do is to match the causal part of scaled cross-spectrum. More precisely, we
suppose that $\hat c_{xy}$ is also rational then use partial fractions to write

\begin{equation*}
  \frac{\hat c_{xy}}{\ov{L(\exp(2\pi i\om))}} =
  \left\{\frac{\hat c_{xy}}{\ov{L(\exp(2\pi i\om))}}\right\}_+ +
  \left\{\frac{\hat c_{xy}}{\ov{L(\exp(2\pi i\om))}}\right\}_-
\end{equation*}

as a sum of terms with poles within and without (respectively) of $C_1$. With
this notation we may solve the causal Wiener filter for rational spectra via

\begin{equation}
  \hat a(\om) = \frac{1}{L(\exp(2\pi i\om))}
     \left\{\frac{\hat c_{xy}}{\ov{L(\exp(2\pi i\om))}}\right\}_+. \label{eq:Wcaus}
\end{equation}

We show how to structure an integrand so that the Second Residue Theorem returns
the number of zeros of a pair of functions within a chosen region. 
This in turn leads to a beautiful comparison theorem that permits us to 
equate the number of zeros of a "hard" function with that of an "easy" 
function. This in turn permits us to prove the Fundamental Theorem of
Algebra, i.e., the statement that every $n$th order polynomial has $n$ zeros. 
And finally, by a very similar argument we prove a useful perturbation result. 
Namely, we show that the zeros of a polynomial don't move drastically when one 
perturbs the polynomial coefficients. We begin with the zero counter.

\begin{proposition}\label{prop:ci7}.  **The Argument Principle.**  If $r=f/g$ 
is the ratio of two differentiable functions on and in the simple closed 
curve $C$ and neither $f$ nor $g$ have zeros on $C$ then

\begin{equation}
  \frac{1}{2\pi i}\int_C \frac{r'(z)}{r(z)}\,dz = Z(f,C) - Z(g,C),  
  \label{eq:ArgP}
\end{equation}
where $Z(f,C)$ is the number (counting multiplicity) of zeros of $f$ in $C$.
 
\end{proposition}

**Proof:** From $r(z)=f(z)/g(z)$ comes

\begin{equation*}
  \frac{r'(z)}{r(z)} = \frac{g(z)f'(z)-f(z)g'(z)}{f(z)g(z)}
\end{equation*}
and so each pole of $r'/r$ is a zero of either $f$ or $g$. 
We take these up separately.

If $\la$ is a zero of $f$ of order $k$ then $r(z)=(z-\la)^kq(z)$ where 
$q(\la)\ne 0$.
It follows that $r'(z)=k(z-\la)^{k-1}q(z) + (z-\la)^kq'(z)$ and so

\begin{equation*}
  \frac{r'(z)}{r(z)} = \frac{k}{z-\la} + \frac{q'(z)}{q(z)}.
\end{equation*}
As the latter term is well behaved at $z=\la$ it follows that 
${\rm res}(r'/r,\la)=k$.

If $\mu$ is a zero of $g$ of order $m$ then $r(z)=(z-\mu)^{-m}p(z)$ where 
$p(\mu)\ne 0$.
It follows that $r'(z)=-m(z-\mu)^{-m-1}p(z) + (z-\mu)^{-m}p'(z)$ and so

\begin{equation*}
  \frac{r'(z)}{r(z)} = \frac{-m}{z-\mu} + \frac{p'(z)}{p(z)}.
\end{equation*}
As the latter term is well behaved at $z=\mu$ it follows that 
${\rm res}(r'/r,\mu)=-m$.

Combining these two residue calculations, the Second Residue Theorem delivers 
(\ref{eq:ArgP}). **End of Proof.**


From this we establish the very useful comparison principle.


\begin{proposition}\label{prop:ci8}.  **Rouch\'e's Theorem.**  If $f$ and $g$ are
two complex differentiable functions on and in the simple closed curve $C$ and

\begin{equation}
  |f(z)-g(z)| < |g(z)| \hskip 0.25in \forall \ z\in C  \label{eq:Rhyp}
\end{equation}

then $f$ and $g$ have the same number of zeros in $C$.
 
\end{proposition}

**Proof:** We define $r\equiv f/g$ and deduce from (\ref{eq:Rhyp}) that $r$ has neither zeros
nor poles on $C$. As such we may read from the Argument Principle that

\begin{equation*}
  \frac{1}{2\pi i}\int_C \frac{r'(z)}{r(z)}\,dz = Z(f,C) - Z(g,C),
\end{equation*}

and so it remains to show that this integral is zero. To wit, we note
that

\begin{equation*}
  F(t) = \frac{1}{2\pi i}\int_C \frac{r'(z)}{r(z)+t}\,dz = Z(f+tg,C) - Z(g,C)
\end{equation*}

is both integer valued and continuous in $t$. This implies that $F(t)$ is constant.
This constant reveals itself on noting that

\begin{equation*}
  |F(t)| \le \frac{|C|}{2\pi}\frac{\max\{|r'(z)|:z\in C\}}{t-\max\{|r(z)|:z\in C\}}.
\end{equation*}

This implies that $F(t)\to 0$ as $t\to\infty$. Hence, $0=F(0)=Z(f,C) - Z(g,C)$
as claimed. **End of Proof.**



This has many subtle applications, perhaps the simplest being


\begin{proposition}\label{prop:ci9}.  **Fundamental Theorem of Algebra.**  If $f(z)$ is
a polynomial of degree $n$ the $f$ has precisely $n$ zeros.

\end{proposition}
 
**Proof:** From $f(z)=f_0+f_1z+\cdots+f_nz^n$ we construct $g(z)=f_nz^n$ and $C=C(0,R)$ where

\begin{equation*}
  R = 1 + \frac{1}{|f_n|}\sum_{j=0}^{n-1}|f_j|.
\end{equation*}

It now follows from the triangle inequality that for $z\in C$,

\begin{equation*}
  |f(z)-g(z)|=|f_0+f_1z+\cdots+f_{n-1}z^{n-1}| \le R^{n-1}\sum_{j=0}^{n-1}|f_j|
   < |f_n|R^n = |f_nz^n|=|g(z)|. 
\end{equation*}

It now follows from Rouch\'e's Theorem that $f$ and $g$ have the same number of zeros
in $C$. As $g$ has $n$ zeros there then so too does $f$. **End of Proof.**

By a very similar argument we can show that the roots of polynomial are continuous
functions of its coefficients. We will use this result to build a quantitative
perturbation theory in Chapter \secref{sym}.


\begin{proposition}\label{prop:ci10}. Suppose
$f(\eps,z)=f_0(\eps)+f_1(\eps)z + f_2(\eps)z^2 + \cdots + f_n(\eps)z^n$ 
where each $f_j$ is a continuous complex function of the complex parameter $\eps$
in some ball about $\eps=0$. If $\la$ is a zero of order $k$ of $z\mapsto f(0,z)$ 
then there exists an $\rho>0$ and $\eps_0>0$ such that $z\mapsto f(\eps,z)$ has precisely
$k$ zeros in $C(\la,\rho)$ when $|\eps|<\eps_0$.
\end{proposition}

**Proof:** Pick $\rho>0$ so that no other zeros of $z\mapsto f(0,z)$ lie in $C=C(\la,\rho)$
and record 

\begin{equation*}
  F = \min_{z\in C}|f(0,z)|  \and R = 1 + \max_{z\in C}|z|.
\end{equation*}

As each $f_j$ is continuous we may choose an $\eps_0$ small enough to guarantee that

\begin{equation*}
  \max_j|f_j(\eps)-f_j(0)| < \frac{F}{(n+1)R^n}, 
          \hskip 0.25in \forall\ |\eps|<\eps_0.
\end{equation*}

This permits us to establish the bound

\begin{equation*}
  |f(\eps,z)-f(0,z)| \le \sum_{j=0}^n |f_j(\eps)-f_j(0)||z^j| < F \le |f(0,z)|,
\end{equation*}

and so conclude from Rouch\'e's Theorem that $z\mapsto f(\eps,z)$ has the same number of
zeros in $C$ as $z\mapsto f(0,z)$ for each $|\eps|<\eps_0$. **End of Proof.**
